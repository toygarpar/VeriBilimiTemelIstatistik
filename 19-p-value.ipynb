## 🔍 P-Değeri Nedir?

**p-değeri**, sıfır hipotezi (H₀) doğruysa,  
gözlemlediğiniz farkın ya da daha uç bir farkın  
rastgele oluşma olasılığıdır.

📌 Basitçe:
> “Eğer aslında fark yoksa,  
bu kadar fark göreceğimiz ne kadar olasıydı?”




P-değeri (olasılık değeri), bir hipotez testinde gözlemlenen sonuçların (veya daha uç sonuçların), sıfır hipotezi doğruyken elde edilme olasılığıdır. Başka bir deyişle, eğer sıfır hipotezi doğruysa, elde ettiğimiz verilerin (veya daha uç verilerin) rastlantısal olarak ortaya çıkma olasılığını gösterir.

**P-değerinin Anlamı ve Yorumlanması:**

- **Düşük p-değeri (genellikle ≤ 0.05):** Gözlemlenen sonuçların sıfır hipotezi doğruyken rastlantısal olarak ortaya çıkma olasılığının düşük olduğunu gösterir. Bu, sıfır hipotezine karşı güçlü kanıt olarak yorumlanır ve genellikle sıfır hipotezinin reddedilmesine yol açar.
- **Yüksek p-değeri (genellikle > 0.05):** Gözlemlenen sonuçların sıfır hipotezi doğruyken rastlantısal olarak ortaya çıkma olasılığının yüksek olduğunu gösterir. Bu, sıfır hipotezini reddetmek için yeterli kanıt olmadığını gösterir. Bu, sıfır hipotezinin _doğru olduğu_ anlamına gelmez, sadece elimizdeki verilerin sıfır hipotezini çürütmek için yeterli kanıt sağlamadığı anlamına gelir.

## 🧠 Neden Önemlidir?

| Sebep | Açıklama |
|--------|----------|
| ✅ **Karar verebilme gücü** | "Bu fark gerçek mi?" sorusunun cevabını verir |
| ✅ **Yanılma riskini ölçer** | Tip I hatayı kontrol edebilirsin |
| ✅ **Modelleme sürecinde kullanılır** | Regresyon katsayıları için test yapılır |
| ✅ **A/B testlerinde yön verir** | Yeni tasarım gerçekten iyi mi? |

📌 p-değeri, **karar alırken kullandığın pusuladır.**


P değeri (p-value), bir istatistiksel test sonucunda elde edilen verilerin, test edilen hipotezin doğruluğu ile çelişip çelişmediğini değerlendirmek için kullanılan bir ölçüdür. P değeri, null hipotezi (H0) reddetme kararını vermede önemli bir rol oynar.

  

Temel olarak, p değeri şu soruya cevap verir: "Eğer null hipotezi doğruysa, gözlemlenen verilerin bu hipotez altında bu kadar aşırı veya daha fazla olma olasılığı nedir?"


Hipotez testlerinin sonuçlarını değerlendirmek üzere programlar tarafından p-value değeri verilir. Bu değer üzerinden kolayca yorum yapabiliriz.


## ⚖️ p-değeri ve Anlamlılık Düzeyi (α)

| Karşılaştırma | Anlamı |
|----------------|---------|
| **p ≤ α** → `H₀ reddedilir` | İddia çürütüldü |
| **p > α** → `H₀ reddedilemez` | Kanıt yeterli değil |

📌 Genelde `α = 0.05` alınır.  
Yani %5 yanılma payı kabul edilir.

---

İşte p değerinin bazı temel özellikleri:

  

1. **P Değeri ve Null Hipotezi:**

- **Null hipotezi genellikle şu şekildedir: "İki grup arasında bir fark yoktur" veya "Bir etki yoktur."

- Küçük bir p değeri, null hipotezinin reddedilmesine ve iki grup arasında anlamlı bir farkın veya bir etkinin olduğuna işaret edebilir.

  

2. **P Değeri ve Anlamlılık Düzeyi (α):**

- P değeri, anlamlılık düzeyi (α) ile karşılaştırılır.

- Anlamlılık düzeyi, bir testin yanlış pozitif sonuç verme olasılığını kontrol eder. Tip I hatasının kabul edilebilir seviyede tutulması önemlidir.

- Yaygın olarak kullanılan anlamlılık düzeyleri arasında α = 0.05 ve α = 0.01 bulunur.


**Anlamlılık Düzeyi (α):**

Anlamlılık düzeyi (alfa), hipotez testinde önceden belirlenmiş bir eşik değerdir. Genellikle 0.05 olarak seçilir, bu da %5 yanılma riskini (I. tip hata yapma riskini) kabul ettiğimiz anlamına gelir.

**P-değeri ve Anlamlılık Düzeyi Karşılaştırması:**

Bir hipotez testinde karar vermek için p-değeri anlamlılık düzeyi (α) ile karşılaştırılır:

- **p-değeri ≤ α:** Sıfır hipotezi reddedilir. Sonuçlar istatistiksel olarak anlamlıdır.
- **p-değeri > α:** Sıfır hipotezi reddedilmez. Sonuçlar istatistiksel olarak anlamlı değildir.


**Örnek:**

Bir ilaç firması, yeni bir ilacın kan basıncını düşürdüğünü iddia etmektedir. Bu iddiayı test etmek için bir deney yaparlar ve p-değerini 0.03 olarak bulurlar. Anlamlılık düzeyini 0.05 olarak belirlediysek:

- p-değeri (0.03) < α (0.05) olduğundan, sıfır hipotezi (ilacın kan basıncını düşürmediği) reddedilir. Bu, ilacın kan basıncını düşürdüğüne dair istatistiksel olarak anlamlı kanıt olduğunu gösterir.

Başka bir örnekte, aynı deneyde p-değeri 0.12 olarak bulunmuş olsaydı:

- p-değeri (0.12) > α (0.05) olduğundan, sıfır hipotezi reddedilmez. Bu, ilacın kan basıncını düşürmediği anlamına gelmez, sadece elimizdeki verilerin bu iddiayı desteklemek için yeterli kanıt sağlamadığı anlamına gelir.

## 📈 p-değeri Nasıl Hesaplanır?

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns


# Örnek veri oluşturma (boy ölçümleri)
np.random.seed(42)
orneklem = np.random.normal(loc=174, scale=5, size=100)  # cm cinsinden

# Test parametreleri
mu_0 = 172  # H0 hipotezi: Ortalama boy = 172 cm
alpha = 0.05  # Anlamlılık düzeyi
```

```python
# Betimsel istatistikler
orneklem_ort = np.mean(orneklem)
orneklem_std = np.std(orneklem, ddof=1)  # Örneklem standart sapması
n = len(orneklem)

# 1. Z-testi ile p-değeri hesaplama (n > 30 olduğu için)
z_skor = (orneklem_ort - mu_0) / (orneklem_std / np.sqrt(n))
p_deger_z = 2 * (1 - stats.norm.cdf(abs(z_skor)))  # İki yönlü test

# 2. t-testi ile p-değeri hesaplama (daha doğru)
t_skor = (orneklem_ort - mu_0) / (orneklem_std / np.sqrt(n))
p_deger_t = 2 * (1 - stats.t.cdf(abs(t_skor), df=n-1))  # İki yönlü test

# 3. Scipy'nin hazır fonksiyonu ile
t_test_sonuc = stats.ttest_1samp(orneklem, popmean=mu_0)

# Sonuçları yazdırma
print(f"Örneklem Ortalaması: {orneklem_ort:.2f} cm")
print(f"Standart Sapma: {orneklem_std:.2f} cm")
print(f"\nZ-testi:")
print(f"  Z-skoru: {z_skor:.4f}")
print(f"  p-değeri: {p_deger_z:.6f}")
print(f"\nt-testi:")
print(f"  t-skoru: {t_skor:.4f}")
print(f"  p-değeri: {p_deger_t:.6f}")
print(f"\nScipy t-test fonksiyonu:")
print(f"  t-skoru: {t_test_sonuc.statistic:.4f}")
print(f"  p-değeri: {t_test_sonuc.pvalue:.6f}")
```

```python
# Karar verme
if p_deger_t < alpha:
    print(f"\nSonuç: p < {alpha} → H0 reddedilir (ortalama {mu_0} cm değil)")
else:
    print(f"\nSonuç: p ≥ {alpha} → H0 reddedilemez (ortalama {mu_0} cm olabilir)")
```

```python
# Görselleştirme
plt.figure(figsize=(12, 6))

# 1. Örneklem dağılımı
plt.subplot(1, 2, 1)
sns.histplot(orneklem, kde=True, color='skyblue')
plt.axvline(orneklem_ort, color='red', linestyle='--', label=f'Örneklem ort. ({orneklem_ort:.1f} cm)')
plt.axvline(mu_0, color='green', linestyle='-', label=f'H0 değeri ({mu_0} cm)')
plt.title('Örneklem Dağılımı')
plt.xlabel('Boy (cm)')
plt.legend()

# 2. t-dağılımı ve p-değeri
plt.subplot(1, 2, 2)
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df=n-1)
plt.plot(x, y, 'b-', label=f't-dağılımı (df={n-1})')

# Kritik bölge ve p-değeri
t_kritik = stats.t.ppf(1-alpha/2, df=n-1)
plt.axvline(t_kritik, color='red', linestyle=':', label=f'Kritik değer (±{t_kritik:.2f})')
plt.axvline(-t_kritik, color='red', linestyle=':')
plt.axvline(t_skor, color='purple', linestyle='-', label=f't-skoru ({t_skor:.2f})')

# p-değerini gölgelendirme
if t_skor > 0:
    p_x = np.linspace(t_skor, 4, 100)
else:
    p_x = np.linspace(-4, t_skor, 100)
plt.fill_between(p_x, stats.t.pdf(p_x, df=n-1), color='purple', alpha=0.3, label='p-değeri alanı')

plt.title('t-Dağılımı ve p-değeri')
plt.xlabel('t Değeri')
plt.ylabel('Yoğunluk')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

- Genellikle, α'dan daha küçük bir p değeri, sonuçların istatistiksel olarak anlamlı olduğunu düşündürecektir.

## 📐 p-değeri ve Hipotez Türleri Bağlantısı

| Hipotez Türü | p-değeri Yorumu |
|---------------|------------------|
| **Çift yönlü (two-tailed)** | Fark var ama yönü önemli değil |
| **Tek yönlü (one-tailed)** | Fark belirli yönde arandı |
| **Korelasyon testi** | Değişkenler ilişkili mi? |
| **Regresyon analizi** | Katsayı anlamlı mı? |
| **ANOVA** | Gruplar arasında fark var mı? |

📌 Hangi test yapılırsa yapılsın,  
p-değeri her zaman **kanıt kalitesini** gösterir.

📈 Görselleştirme: p-değeri ve Red Bölgesi

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Set parameters
alpha = 0.05  # Significance level
z_skoru = 2.5  # Example z-score (replace with your calculated value)

# Generate data for standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

# Create figure with improved styling
plt.figure(figsize=(12, 6))

# Plot the standard normal curve
plt.plot(x, y, label="Standard Normal Distribution", linewidth=2, color='navy')

# Calculate critical values
kritik_deger = norm.ppf(1 - alpha/2)

# Add critical value lines
plt.axvline(kritik_deger, color='red', linestyle='--', 
           linewidth=1.5, label=f'Critical Value (±{kritik_deger:.2f})')
plt.axvline(-kritik_deger, color='red', linestyle='--', linewidth=1.5)

# Add z-score line
plt.axvline(z_skoru, color='green', linestyle='-', 
           linewidth=2, label=f'Test Statistic (z = {z_skoru:.2f})')

# Shade rejection regions
plt.fill_between(x, y, where=(x >= kritik_deger), 
                color='red', alpha=0.2, label='Rejection Region')
plt.fill_between(x, y, where=(x <= -kritik_deger), 
                color='red', alpha=0.2)


# p-değeri (iki yönlü test)
p_degeri = 2 * (1 - norm.cdf(abs(z_skoru)))
print(f"p-değeri: {p_degeri:.6f}")


# Add p-value area
if z_skoru > 0:
    p_x = np.linspace(z_skoru, 4, 100)
else:
    p_x = np.linspace(-4, z_skoru, 100)
plt.fill_between(p_x, norm.pdf(p_x), color='green', alpha=0.1, label='p-value area')

# Add decision text
decision = "Reject H₀" if abs(z_skoru) > kritik_deger else "Fail to reject H₀"
plt.text(0, 0.1, f'Decision: {decision}', ha='center', va='center',
         bbox=dict(facecolor='white', alpha=0.8), fontsize=12)

# Formatting
plt.title("Hypothesis Testing Visualization (α = 0.05)", pad=20, fontsize=14)
plt.xlabel("Standardized Z Score", labelpad=10)
plt.ylabel("Probability Density", labelpad=10)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout
plt.tight_layout()
plt.show()
```

This code now properly shows:

The standard normal distribution curve

The critical values at ±1.96 (for α=0.05)

The test statistic (z-score)

The rejection regions (red)

The p-value area (green)

The statistical decision based on the comparison


📌 Bu grafikte:
- Yeşil çizgi → z-skorunuz
- Kırmızı alan → red bölgesi
- Eğer yeşil çizgi kırmızının içindeyse → H₀ reddedilir

**Dikkat Edilmesi Gerekenler:**

- P değeri, gözlemlenen farkın büyüklüğü veya bilimsel önemi hakkında bir bilgi vermez. Sadece istatistiksel anlamlılığı değerlendirir.

  

Sonuç olarak, p değeri, bir istatistiksel testin sonuçlarını değerlendirmek ve null hipotezi hakkında bir çıkarım yapmak için önemli bir araçtır. Ancak, p değerinin tek başına yeterli bir ölçü olmadığı, sonuçların bağlamı, hipotezin kendisi ve diğer faktörler de dikkate alınmalıdır.


----



**P-değerinin Yanlış Yorumlanması:**

P-değeri sıklıkla yanlış yorumlanabilir. İşte bazı yaygın yanlış anlamalar:

- **P-değeri, sıfır hipotezinin doğru olma olasılığı değildir.** P-değeri, _eğer_ sıfır hipotezi doğruysa, gözlemlenen sonuçların (veya daha uç sonuçların) ortaya çıkma olasılığıdır.
- **İstatistiksel anlamlılık, pratik anlamlılık anlamına gelmez.** İstatistiksel olarak anlamlı bir sonuç, gerçek dünyada önemli bir etki anlamına gelmeyebilir. Örneğin, bir ilacın kan basıncını istatistiksel olarak anlamlı bir şekilde düşürdüğü bulunabilir, ancak bu düşüş klinik olarak önemsiz olabilir.
- **Yüksek p-değeri, sıfır hipotezinin doğru olduğunu kanıtlamaz.** Sadece elimizdeki verilerin sıfır hipotezini çürütmek için yeterli kanıt sağlamadığı anlamına gelir.



Dağılım testlerinde H0 reddedilmek istenmez. Çünü H0 "örnek dağılım ile teorik dağılım arasında fark yoktur" der.

## 🧪 Gerçek Hayat Senaryosu: A/B Testi

Yeni tasarım vs eski tasarım → hangisi daha iyi?

```python
from statsmodels.stats.proportion import proportions_ztest

# Simülasyon: dönüşüm oranları
kontrol_don = [1] * 120 + [0] * 880  # %12
deney_don = [1] * 150 + [0] * 850    # %15

_, p_degeri = proportions_ztest(count=[sum(kontrol_don), sum(deney_don)],
                                 nobs=[len(kontrol_don), len(deney_don)])
print(f"A/B Testi p-değeri: {p_degeri:.4f}")
```

📌 p-değeri < 0.05 → fark anlamlı  
Yani yeni tasarım daha iyi!
