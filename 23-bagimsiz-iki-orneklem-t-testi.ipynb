# Bağımsız İki Örneklem T Testi Teorisi

## <span style ="color:red">**Bağımsız iki örneklem T-Testi (AB Testi)**</span>


- **İki grup ortalaması arasında karşılaştırma yapılmak istenildiğinde kullanılır.**

  

'Bağımsız İki Örneklem T-Testi', aynı özelliğe sahip iki farklı grup arasında istatistiksel olarak anlamlı bir fark olup olmadığını değerlendiren bir hipotez testidir. Bu test, genellikle bir değişiklik veya müdahalenin (örneğin, bir reklam kampanyasının etkisi) etkisini ölçmek için kullanılır. Bu test bazen "AB Testi" olarak da adlandırılır.



Bağımsız İki Örneklem T Testi (Independent Samples t-test veya Student's t-test olarak da bilinir), iki bağımsız grubun ortalamaları arasında istatistiksel olarak anlamlı bir fark olup olmadığını belirlemek için kullanılan bir parametrik testtir. "Bağımsız" terimi, iki örneklem grubunun birbirinden bağımsız olduğu anlamına gelir; yani, bir gruptaki bir gözlem diğer gruptaki hiçbir gözlemi etkilemez. Bu test, özellikle iki farklı grubun belirli bir özellik açısından karşılaştırılması gerektiğinde yaygın olarak kullanılır.



**Bağımsız İki Örneklem T Testi Ne Zaman Kullanılır?**

- İki farklı grubun ortalamalarını karşılaştırmak istediğinizde. Örneğin, erkeklerin ve kadınların boy ortalamalarını, iki farklı öğretim yönteminin öğrencilerin başarıları üzerindeki etkisini veya iki farklı ilacın etkinliklerini karşılaştırmak gibi.
- Örneklem büyüklüğü küçük olduğunda (genellikle n < 30) ve popülasyon standart sapmaları bilinmediğinde. Ancak, örneklem büyüklüğü arttıkça, normal dağılım varsayımına olan bağımlılık azalır (Merkezi Limit Teoremi).
- Verilerin normal dağılıma yakın olduğu varsayıldığında.
- Verilerin aralıklı veya oransal ölçekte ölçüldüğü durumlarda.




**Bağımsız İki Örneklem T Testinin Varsayımları:**

- **Bağımsızlık:** Her iki örneklemdeki gözlemler birbirinden bağımsız olmalıdır. Yani, bir gruptaki bir gözlem diğer gruptaki hiçbir gözlemi etkilememelidir.
- **Normallik:** Her iki popülasyondan alınan örneklemlerin normal dağılıma sahip olması varsayılır. Ancak, örneklem büyüklükleri yeterince büyük olduğunda, bu varsayımın ihlali testin sonuçlarını önemli ölçüde etkilemeyebilir (Merkezi Limit Teoremi). Normallik varsayımı test edilebilir (örneğin, Shapiro-Wilk testi veya Kolmogorov-Smirnov testi ile). Normallik sağlanmıyorsa, Mann-Whitney U testi gibi parametrik olmayan alternatifler düşünülebilir.
- **Varyansların Homojenliği (Eşit Varyanslar):** İki grubun varyanslarının eşit olduğu varsayılır. Bu varsayımın kontrol edilmesi önemlidir çünkü testin hesaplama yöntemi bu varsayıma göre değişir. Varyansların homojenliği Levene testi ile kontrol edilebilir.





### **Bağımsız İki Örneklem T-Testi'ni uygulamak için aşağıdaki adımları takip edebilirsiniz:

  

1. **Hipotez Kurma:**

- **Null Hipotezi (H0):** İki grup arasında bir fark yoktur.

- **Alternatif Hipotez (H1):** İki grup arasında bir fark vardır.

  

2. **Veri Toplama:**

- İki bağımsız grup için verileri topla. Örneğin, bir ürünün eski ve yeni bir tasarımıyla ilgili kullanıcıların memnuniyet skorları gibi.

  

3. **Test İstatistiğini Hesaplama:**

- İki grup arasındaki ortalama farkı ve grupların varyanslarını kullanarak t-istatistiğini hesapla.

## $t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$

- Burada:

- ${X_1}$ ve ${X_2}$, sırasıyla grup 1 ve grup 2'nin ortalamalarıdır.

- $s_1^2$ ve $s_2^2$, sırasıyla grup 1 ve grup 2'nin varyanslarıdır.

- $n_1$ ve $n_2$, sırasıyla grup 1 ve grup 2'nin örneklem büyüklükleridir.

  

4. **P Değerini Hesaplama:**

- Hesaplanan t-istatistiği üzerinden p değerini belirle.

  

5. **Karar Verme:**

- Belirlenen anlamlılık düzeyi örneğin, alpha = 0.05 üzerinden p değerini karşılaştır ve null hipotezini reddet veya reddetmeme kararı ver.

  

Bağımsız İki Örneklem T-Testi, özellikle bir müdahalenin etkisini değerlendirmek, iki farklı durumu karşılaştırmak veya gruplar arasında bir fark olup olmadığını anlamak için yaygın olarak kullanılan bir istatistiksel testtir.

  
  

- **A / B testi (bölme testi veya kova testi olarak da bilinir) hangisinin daha iyi performans gösterdiğini belirlemek için bir web sayfasının veya uygulamanın iki sürümünü birbiriyle karşılaştırmanın bir yöntemidir.**














### **Bağımsız İki Örneklem T Testi Nasıl Yapılır?**

1. **Sıfır Hipotezi (H₀) ve Alternatif Hipotezi (H₁) Belirleyin:**
    
    - H₀: İki grup ortalaması arasında fark yoktur (μ₁ = μ₂).
    - H₁: İki grup ortalaması arasında fark vardır (μ₁ ≠ μ₂) (İki yönlü test). Veya, bir grup ortalaması diğerinden büyüktür (μ₁ > μ₂ veya μ₁ < μ₂) (Tek yönlü test).
2. **Anlamlılık Düzeyini (α) Belirleyin:** Genellikle α = 0.05 kullanılır.
    
3. **Varyansların Homojenliğini Test Edin (Levene Testi):** Bu testin sonucuna göre hangi t testinin kullanılacağına karar verilir.
    
4. **T İstatistiğini Hesaplayın:**
    
    - **Eşit Varyanslar Durumu:**
        
        ```
        t = (x̄₁ - x̄₂) / Sp√((1/n₁) + (1/n₂))
        ```
        
        Burada:
        
        - x̄₁ ve x̄₂: İki grubun örneklem ortalamaları
        - Sp: Birleştirilmiş standart sapma
        - n₁ ve n₂: İki grubun örneklem büyüklükleri
    - **Eşit Olmayan Varyanslar Durumu (Welch's t-test):** Farklı bir formül kullanılır ve serbestlik derecesi de farklı hesaplanır. Formül biraz daha karmaşıktır ve genellikle istatistiksel yazılımlar tarafından otomatik olarak hesaplanır.
        
5. **Serbestlik Derecesini (sd) Hesaplayın:**
    
    - **Eşit Varyanslar Durumu:** sd = n₁ + n₂ - 2
    - **Eşit Olmayan Varyanslar Durumu (Welch's t-test):** Farklı bir formül kullanılır ve genellikle ondalıklı bir değerdir.
6. **Kritik T Değerini Bulun:** T dağılım tablosundan veya istatistiksel yazılımdan, belirlenen anlamlılık düzeyi (α) ve serbestlik derecesi (sd) ile kritik t değerini bulunur.
    
7. **Karar Verin:**
    
    - Eğer hesaplanan t istatistiğinin mutlak değeri, kritik t değerinden büyükse, sıfır hipotezi reddedilir. Bu, iki grup ortalaması arasında istatistiksel olarak anlamlı bir fark olduğu anlamına gelir.
    - Eğer hesaplanan t istatistiğinin mutlak değeri, kritik t değerinden küçükse, sıfır hipotezi reddedilmez.


## 📐 Formülü:

$$
t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
$$

| Sembol | Açıklama |
|--------|----------|
| $\bar{x}_1, \bar{x}_2$ | İki grubun ortalamaları |
| $s_1, s_2$ | Standart sapmalar |
| $n_1, n_2$ | Örneklem büyüklükleri |

📌 Eğer varyanslar eşitse → daha basit formül  
Yoksa → Welch’s yöntemi tercih edilir

## 🛠️ Adım Adım Uygulama: Gerçek Hayat Senaryosu

### Örnek:
Senin Tasarım (Grup A): 50 kişi → ortalama dönüşüm skoru 7.8  
Rakip Tasarım (Grup B): 50 kişi → ortalama dönüşüm skoru 6.2

#### Soru:
“Tasarım A, B’den daha iyi mi?”

---

### Adım 1: Hipotez Kurulumu

```python
print("H₀: μ_A = μ_B → Tasarımlar arasında fark yok")
print("H₁: μ_A ≠ μ_B → Tasarımlar istatistiksel olarak farklı")
```

✅ Null hipotezi → fark yok  
Alternatif → fark var

### Adım 2: Anlamlılık Düzeyini Ayarla

````python
alpha = 0.05
print(f"Anlamlılık düzeyi (α): {alpha}")
````

📌 %5 yanılma payı kabul ettik.

### Adım 3: Sample Veri Üret

````python
import numpy as np

np.random.seed(42)

grup_A = np.random.normal(loc=7.8, scale=1.5, size=50)  # Tasarım A
grup_B = np.random.normal(loc=6.2, scale=1.5, size=50)  # Tasarım B

print("Grup A Ortalaması:", grup_A.mean().round(2))
print("Grup B Ortalaması:", grup_B.mean().round(2))
````

### Adım 4: Levene Testi ile Varyans Kontrolü

````python
from scipy.stats import levene

stat, p = levene(grup_A, grup_B)
print(f"Levene Testi p-değeri: {p:.4f}")

if p > alpha:
    print("✅ Eşit varyans varsayımı sağlanıyor")
else:
    print("❌ Varyanslar eşit değil → Welch’s t-testi yapılacak")
````

### Adım 5: T-Testi Yap

````python
from scipy.stats import ttest_ind

t_skoru, p_degeri = ttest_ind(grup_A, grup_B)
print(f"T Skoru: {t_skoru:.2f}, p-değeri: {p_degeri:.4f}")
````


### Adım 6: Karar Ver

````python
if p_degeri < alpha:
    print("✅ H₀ reddedildi → Tasarımlar arasında fark var")
else:
    print("❌ H₀ reddedilemedi → Fark tesadüfi olabilir")
````

📌 Yani:
Grup A’nın skoru istatistiksel olarak daha yüksek.

### **Örnek:**

İki farklı öğretim yönteminin öğrencilerin sınav başarıları üzerindeki etkisini karşılaştırmak için bir araştırma yapılıyor. Bir gruba geleneksel yöntemle (Grup 1), diğer gruba ise yeni bir yöntemle (Grup 2) ders veriliyor. Her iki gruptan da 20 öğrenci seçiliyor. Sınav sonuçları aşağıdaki gibi olsun:

- Grup 1 (Geleneksel Yöntem): Ortalama = 75, Standart Sapma = 10
- Grup 2 (Yeni Yöntem): Ortalama = 80, Standart Sapma = 8

α = 0.05 alarak bu iki yöntem arasında istatistiksel olarak anlamlı bir fark olup olmadığını test edelim.

Bu örneğin detaylı hesaplaması, varyansların homojenliği testinin yapılması ve doğru t testinin seçilmesini gerektirir. Bu hesaplamalar genellikle istatistiksel yazılımlar (SPSS, R, Python vb.) ile yapılır.

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Stil ayarları

sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (14, 6)

def bagimsiz_iki_orneklem_t_testi(grup1, grup2, alpha=0.05):
    """
    Bağımsız İki Örneklem t-Testi Uygulaması
    
    Parametreler:
        grup1 (array): Birinci gruba ait veriler
        grup2 (array): İkinci gruba ait veriler
        alpha (float): Anlamlılık düzeyi
    """
    # Betimsel istatistikler
    n1, n2 = len(grup1), len(grup2)
    mean1, mean2 = np.mean(grup1), np.mean(grup2)
    std1, std2 = np.std(grup1, ddof=1), np.std(grup2, ddof=1)
    
    # Varyans homojenliği testi (Levene)
    levene_test = stats.levene(grup1, grup2)
    
    # Varyanslara göre uygun t-testi seçimi
    if levene_test.pvalue > alpha:
        # Varyanslar homojen
        t_stat, p_value = stats.ttest_ind(grup1, grup2, equal_var=True)
        test_turu = "Varyanslar homojen (Student t-testi)"
    else:
        # Varyanslar heterojen
        t_stat, p_value = stats.ttest_ind(grup1, grup2, equal_var=False)
        test_turu = "Varyanslar heterojen (Welch t-testi)"
    
    # Etki büyüklüğü (Cohen's d)
    pooled_std = np.sqrt(((n1-1)*std1**2 + (n2-1)*std2**2) / (n1+n2-2))
    cohen_d = (mean1 - mean2) / pooled_std
    
    # Güven aralığı
    diff_mean = mean1 - mean2
    margin_of_error = stats.t.ppf(1 - alpha/2, df=n1+n2-2) * np.sqrt(std1**2/n1 + std2**2/n2)
    ci_lower = diff_mean - margin_of_error
    ci_upper = diff_mean + margin_of_error
    
    # Sonuçları yazdırma
    print("📊 BAĞIMSIZ İKİ ÖRNEKLEM t-TESTİ SONUÇLARI")
    print("----------------------------------------")
    print(f"Grup 1 (n={n1}): Ortalama = {mean1:.2f}, Standart Sapma = {std1:.2f}")
    print(f"Grup 2 (n={n2}): Ortalama = {mean2:.2f}, Standart Sapma = {std2:.2f}")
    print(f"\nVaryans Homojenliği Testi (Levene):")
    print(f"  F-istatistiği = {levene_test.statistic:.4f}, p-değeri = {levene_test.pvalue:.4f}")
    print(f"  Sonuç: {test_turu}")
    print(f"\nt-istatistiği: {t_stat:.4f}")
    print(f"p-değeri: {p_value:.6f}")
    print(f"Etki Büyüklüğü (Cohen's d): {cohen_d:.3f}")
    print(f"%95 Güven Aralığı (Ortalama Fark): ({ci_lower:.2f}, {ci_upper:.2f})")
    
    # Karar ve yorum
    print("\n🔍 YORUM:")
    if p_value < alpha:
        print(f"✅ p < {alpha} → H₀ REDDEDİLİR (Gruplar arasında anlamlı fark var)")
        if mean1 > mean2:
            print(f"   Grup 1 ortalaması ({mean1:.2f}), Grup 2 ortalamasından ({mean2:.2f}) ANLAMLI ŞEKİLDE YÜKSEK")
        else:
            print(f"   Grup 2 ortalaması ({mean2:.2f}), Grup 1 ortalamasından ({mean1:.2f}) ANLAMLI ŞEKİLDE YÜKSEK")
    else:
        print(f"❌ p ≥ {alpha} → H₀ REDDEDİLEMEZ (Gruplar arasında anlamlı fark YOK)")
    
    # Görselleştirme
    plt.figure(figsize=(14, 6))
    
    # 1. Kutu grafiği
    plt.subplot(1, 2, 1)
    sns.boxplot(data=[grup1, grup2], palette=['skyblue', 'lightgreen'])
    plt.xticks([0, 1], ['Grup 1 (Geleneksel)', 'Grup 2 (Yeni)'])
    plt.title('Grupların Sınav Puan Dağılımı')
    plt.ylabel('Puan')
    
    # 2. t-dağılımı ve test sonucu
    plt.subplot(1, 2, 2)
    df = n1 + n2 - 2  # Serbestlik derecesi
    x = np.linspace(-4, 4, 1000)
    y = stats.t.pdf(x, df)
    plt.plot(x, y, label=f't-dağılımı (df={df})')
    
    # Kritik bölgeler
    t_critical = stats.t.ppf(1 - alpha/2, df)
    plt.axvline(t_critical, color='red', linestyle=':', label=f'Kritik değer (±{t_critical:.2f})')
    plt.axvline(-t_critical, color='red', linestyle=':')
    plt.fill_between(x, y, where=(x >= t_critical), color='red', alpha=0.2, label='Red bölgesi')
    plt.fill_between(x, y, where=(x <= -t_critical), color='red', alpha=0.2)
    
    # Test istatistiği
    plt.axvline(t_stat, color='green', linestyle='-', label=f't-istatistiği ({t_stat:.2f})')
    
    plt.title('t-Testi Sonuçları')
    plt.xlabel('t Değeri')
    plt.ylabel('Yoğunluk')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

````

````python
# Örnek veri oluşturma (verilen örnek için)
np.random.seed(42)
grup1 = np.random.normal(loc=75, scale=10, size=20)  # Geleneksel yöntem
grup2 = np.random.normal(loc=80, scale=8, size=20)   # Yeni yöntem

# Testi çalıştırma
print("🎓 ÖĞRETİM YÖNTEMLERİNİN KARŞILAŞTIRILMASI")
bagimsiz_iki_orneklem_t_testi(grup1, grup2, alpha=0.05)
````

### Örnek uygulamalar:

  

1. **İki Reklam Stratejisi Karşılaştırması:**

- Problem: Bir şirket, iki farklı reklam stratejisi uyguluyor ve hangisinin daha etkili olduğunu öğrenmek istiyor.

- Gruplar: İlk grup, Reklam Stratejisi A'nın maruz kaldığı kişiler, ikinci grup ise Reklam Stratejisi B'nin maruz kaldığı kişiler.

- Hipotez: Reklam Stratejisi A ve B arasında satışlarda anlamlı bir fark var mı?

  



````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown

# Stil ayarları

sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (14, 6)

def reklam_stratejisi_testi(strateji_a, strateji_b, alpha=0.05):
    """
    İki reklam stratejisinin performansını karşılaştıran t-testi
    
    Parametreler:
        strateji_a (array): Strateji A'nın satış verileri
        strateji_b (array): Strateji B'nin satış verileri
        alpha (float): Anlamlılık düzeyi
    """
    # Betimsel istatistikler
    n_a, n_b = len(strateji_a), len(strateji_b)
    mean_a, mean_b = np.mean(strateji_a), np.mean(strateji_b)
    std_a, std_b = np.std(strateji_a, ddof=1), np.std(strateji_b, ddof=1)
    
    # Varyans homojenliği testi (Levene)
    levene_test = stats.levene(strateji_a, strateji_b)
    
    # Varyanslara göre uygun t-testi seçimi
    if levene_test.pvalue > alpha:
        t_stat, p_value = stats.ttest_ind(strateji_a, strateji_b, equal_var=True)
        test_turu = "Varyanslar homojen (Student t-testi)"
    else:
        t_stat, p_value = stats.ttest_ind(strateji_a, strateji_b, equal_var=False)
        test_turu = "Varyanslar heterojen (Welch t-testi)"
    
    # Etki büyüklüğü (Cohen's d)
    pooled_std = np.sqrt(((n_a-1)*std_a**2 + (n_b-1)*std_b**2) / (n_a+n_b-2))
    cohen_d = (mean_a - mean_b) / pooled_std
    
    # Güven aralığı
    diff_mean = mean_a - mean_b
    margin_of_error = stats.t.ppf(1 - alpha/2, df=n_a+n_b-2) * np.sqrt(std_a**2/n_a + std_b**2/n_b)
    ci_lower = diff_mean - margin_of_error
    ci_upper = diff_mean + margin_of_error
    
    # Sonuç raporu
    display(Markdown(f"""
    ### 📊 Reklam Stratejileri Karşılaştırma Sonuçları
    
    | Özellik | Strateji A | Strateji B |
    |---------|------------|------------|
    | Örneklem Büyüklüğü | {n_a} | {n_b} |
    | Ortalama Satış | {mean_a:.2f} | {mean_b:.2f} |
    | Standart Sapma | {std_a:.2f} | {std_b:.2f} |
    
    **Varyans Homojenliği Testi:**  
    Levene F = {levene_test.statistic:.3f}, p = {levene_test.pvalue:.3f} → *{test_turu}*
    
    **t-Test Sonuçları:**  
    t({n_a+n_b-2}) = {t_stat:.3f}, p = {p_value:.4f}  
    Cohen's d = {cohen_d:.3f}  
    %95 GA = [{ci_lower:.2f}, {ci_upper:.2f}]
    """))
    
    # Karar ve yorum
    if p_value < alpha:
        if mean_a > mean_b:
            sonuc = f"✅ Strateji A, Strateji B'den anlamlı derecede BAŞARILI (p = {p_value:.4f})"
        else:
            sonuc = f"✅ Strateji B, Strateji A'dan anlamlı derecede BAŞARILI (p = {p_value:.4f})"
    else:
        sonuc = f"❌ İki strateji arasında anlamlı fark YOK (p = {p_value:.4f})"
    
    display(Markdown(f"**Sonuç:** {sonuc}"))
    
    # Görselleştirme
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    
    # 1. Kutu grafiği ve swarm plot
    sns.boxplot(data=[strateji_a, strateji_b], ax=ax1, width=0.5,
                palette=["skyblue", "lightgreen"])
    sns.swarmplot(data=[strateji_a, strateji_b], ax=ax1,
                  palette=["darkblue", "darkgreen"], size=5)
    ax1.set_xticks([0, 1], ['Strateji A', 'Strateji B'])
    ax1.set_ylabel('Satış Miktarı')
    ax1.set_title('Satış Dağılımları')
    
    # 2. Etki büyüklüğü görselleştirme
    ax2.barh(['Ortalama Fark'], [diff_mean], color='salmon')
    ax2.errorbar(x=[diff_mean], y=[0], xerr=[margin_of_error], 
                 fmt='o', color='darkred', capsize=5)
    ax2.axvline(0, color='gray', linestyle='--')
    ax2.set_title('Ortalama Fark ve %95 Güven Aralığı')
    ax2.set_xlabel('Strateji A - Strateji B')
    ax2.set_yticks([])
    
    plt.tight_layout()
    plt.show()

# Simüle edilmiş reklam verileri (örnek)
np.random.seed(42)
strateji_a = np.random.normal(loc=120, scale=15, size=30).astype(int)  # Strateji A
strateji_b = np.random.normal(loc=130, scale=20, size=35).astype(int)  # Strateji B

# Testi çalıştırma
reklam_stratejisi_testi(strateji_a, strateji_b)
````

2. **Ürün Fiyatlandırma Analizi:**

- Problem: Bir e-ticaret şirketi, iki farklı fiyatlandırma stratejisini test etmek istiyor.

- Gruplar: İlk grup, Yüksek Fiyatlı Ürünü gören kişiler, ikinci grup ise Düşük Fiyatlı Ürünü gören kişiler.

- Hipotez: İki fiyatlandırma stratejisi arasında müşteri satın alma davranışlarında anlamlı bir fark var mı?

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from statsmodels.stats.power import TTestIndPower

# Veri oluşturma (simülasyon)
np.random.seed(42)
n = 50  # Her grup için örneklem büyüklüğü

# Yüksek fiyat stratejisi (499 TL) - %8 dönüşüm
yuksek_fiyat = np.random.binomial(1, 0.08, size=n)

# Düşük fiyat stratejisi (299 TL) - %12 dönüşüm
dusuk_fiyat = np.random.binomial(1, 0.12, size=n)

# Betimsel istatistikler
df = pd.DataFrame({
    'Strateji': ['Yüksek Fiyat']*n + ['Düşük Fiyat']*n,
    'Satın Alma': np.concatenate([yuksek_fiyat, dusuk_fiyat])
})
````

````python
# Bağımsız iki örneklem t-testi (proportion test)
def proportion_test(group1, group2, alpha=0.05):
    n1, n2 = len(group1), len(group2)
    p1, p2 = group1.mean(), group2.mean()
    
    # Pooled proportion
    p_pooled = (group1.sum() + group2.sum()) / (n1 + n2)
    
    # Z-test istatistiği
    z = (p1 - p2) / np.sqrt(p_pooled*(1-p_pooled)*(1/n1 + 1/n2))
    
    # p-değeri (iki yönlü)
    p_value = 2 * (1 - stats.norm.cdf(abs(z)))
    
    # Etki büyüklüğü (Cohen's h)
    h = 2 * (np.arcsin(np.sqrt(p1)) - np.arcsin(np.sqrt(p2)))
    
    # Güç analizi
    effect_size = abs(p1 - p2)
    power_analysis = TTestIndPower()
    power = power_analysis.solve_power(effect_size=effect_size, nobs1=n1, alpha=alpha)
    
    return z, p_value, h, power
````

````python
# Testi uygula
z_score, p_val, cohen_h, test_power = proportion_test(yuksek_fiyat, dusuk_fiyat)
````

3. **Eğitim Programları Karşılaştırması:**

- Problem: Bir okul, iki farklı eğitim programını değerlendirmek istiyor.

- Gruplar: İlk grup, Eğitim Programı A'ya katılan öğrenciler, ikinci grup ise Eğitim Programı B'ye katılan öğrenciler.

- Hipotez: İki eğitim programı arasında öğrenci başarı düzeylerinde anlamlı bir fark var mı?

````python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.power import TTestIndPower

# Veri oluşturma (simülasyon)
np.random.seed(42)
n_a, n_b = 30, 35  # Örneklem büyüklükleri

# Program A: Ortalama 75 puan, Standart sapma 10
program_a = np.random.normal(loc=75, scale=10, size=n_a).clip(0, 100)

# Program B: Ortalama 82 puan, Standart sapma 8
program_b = np.random.normal(loc=82, scale=8, size=n_b).clip(0, 100)

# Veri çerçevesi oluşturma
df = pd.DataFrame({
    'Program': ['A']*n_a + ['B']*n_b,
    'Puan': np.concatenate([program_a, program_b]),
    'ÖğrenciID': np.arange(n_a + n_b)
})

# Normallik testi (Shapiro-Wilk)
normal_a = stats.shapiro(program_a).pvalue > 0.05
normal_b = stats.shapiro(program_b).pvalue > 0.05

# Varyans homojenliği testi (Levene)
levene_test = stats.levene(program_a, program_b)

# Parametrik test (t-test) veya non-parametrik test (Mann-Whitney U) seçimi
if normal_a and normal_b:
    # Varyans homojenliğine göre uygun t-test seçimi
    if levene_test.pvalue > 0.05:
        t_test = stats.ttest_ind(program_a, program_b, equal_var=True)
        test_adi = "Bağımsız İki Örneklem t-Testi (Student)"
    else:
        t_test = stats.ttest_ind(program_a, program_b, equal_var=False)
        test_adi = "Bağımsız İki Örneklem t-Testi (Welch)"
    test_istatistik, p_value = t_test.statistic, t_test.pvalue
else:
    # Non-parametrik test
    test_istatistik, p_value = stats.mannwhitneyu(program_a, program_b)
    test_adi = "Mann-Whitney U Testi"
````

````python
# Etki büyüklüğü hesaplama
def calculate_effect_size(group1, group2):
    n1, n2 = len(group1), len(group2)
    pooled_std = np.sqrt(((n1-1)*np.std(group1, ddof=1)**2 + (n2-1)*np.std(group2, ddof=1)**2) / (n1+n2-2))
    return (np.mean(group1) - np.mean(group2)) / pooled_std

cohen_d = calculate_effect_size(program_b, program_a)

# Güç analizi
effect_size = abs(cohen_d)
power_analysis = TTestIndPower()
power = power_analysis.solve_power(effect_size=effect_size, nobs1=n_a, alpha=0.05)
````

````python
# Görselleştirme
plt.figure(figsize=(14, 6))

# 1. Kutu grafiği ve swarm plot
plt.subplot(1, 2, 1)
sns.boxplot(x='Program', y='Puan', data=df, hue='Program', palette=["#1f77b4", "#ff7f0e"], legend=False)
sns.swarmplot(x='Program', y='Puan', data=df, color='.25', size=5)
plt.title('Eğitim Programlarına Göre Başarı Puanları\n')
plt.ylabel('Başarı Puanı')
plt.ylim(0, 100)

# 2. Etki büyüklüğü görselleştirme
plt.subplot(1, 2, 2)
effect_sizes = np.linspace(0.1, 1.5, 50)
power_values = power_analysis.solve_power(effect_size=effect_sizes, nobs1=n_a, alpha=0.05)
plt.plot(effect_sizes, power_values, color='#2ca02c')
plt.axvline(x=effect_size, color='red', linestyle='--', label=f'Mevcut Etki: {effect_size:.2f}')
plt.title('Test Gücü Analizi\n')
plt.xlabel("Etki Büyüklüğü (Cohen's d)")
plt.ylabel('Test Gücü (1 - β)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

4. **Web Sitesi Kullanılabilirliği Karşılaştırması:**

- Problem: Bir şirket, iki farklı web sitesi tasarımını test etmek istiyor.

- Gruplar: İlk grup, Web Sitesi A'yı kullanan kullanıcılar, ikinci grup ise Web Sitesi B'yi kullanan kullanıcılar.

- Hipotez: İki web sitesi arasında kullanılabilirlik performansında anlamlı bir fark var mı?

````python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.power import TTestIndPower
import pingouin as pg  # İleri istatistiksel analizler için

# Veri oluşturma (simülasyon)
np.random.seed(42)
n_a, n_b = 45, 50  # Kullanıcı sayıları

# Web Sitesi A: Ortalama görev tamamlama süresi (saniye)
site_a = np.random.normal(loc=120, scale=25, size=n_a).clip(60, 180)

# Web Sitesi B: Ortalama görev tamamlama süresi (saniye)
site_b = np.random.normal(loc=95, scale=20, size=n_b).clip(50, 150)

# Kullanıcı memnuniyet puanları (1-10 arası)
memnuniyet_a = np.random.normal(loc=6.5, scale=1.5, size=n_a).clip(1, 10)
memnuniyet_b = np.random.normal(loc=8.2, scale=1.2, size=n_b).clip(1, 10)

# Veri çerçevesi oluşturma
df = pd.DataFrame({
    'Site': ['A']*n_a + ['B']*n_b,
    'Görev_Süresi': np.concatenate([site_a, site_b]),
    'Memnuniyet': np.concatenate([memnuniyet_a, memnuniyet_b]),
    'KullanıcıID': np.arange(n_a + n_b)
})

# Normallik testi
normal_a = stats.shapiro(site_a).pvalue > 0.05
normal_b = stats.shapiro(site_b).pvalue > 0.05

# Varyans homojenliği testi
levene_test = pg.homoscedasticity(df, dv='Görev_Süresi', group='Site')

# Uygun test seçimi ve uygulanması
if normal_a and normal_b:
    if levene_test['pval'].iloc[0] > 0.05:
        t_test = pg.ttest(df[df['Site']=='A']['Görev_Süresi'],
                         df[df['Site']=='B']['Görev_Süresi'],
                         correction=False)
        test_adi = "Bağımsız t-Testi (Student)"
    else:
        t_test = pg.ttest(df[df['Site']=='A']['Görev_Süresi'],
                         df[df['Site']=='B']['Görev_Süresi'],
                         correction=True)
        test_adi = "Bağımsız t-Testi (Welch)"
else:
    t_test = pg.mwu(df[df['Site']=='A']['Görev_Süresi'],
                   df[df['Site']=='B']['Görev_Süresi'])
    test_adi = "Mann-Whitney U Testi"

# Etki büyüklüğü hesaplama
cohen_d = pg.compute_effsize(df[df['Site']=='A']['Görev_Süresi'],
                            df[df['Site']=='B']['Görev_Süresi'],
                            eftype='cohen')

# Memnuniyet puanları için test
memnuniyet_test = pg.mwu(df[df['Site']=='A']['Memnuniyet'],
                        df[df['Site']=='B']['Memnuniyet'])

# Güç analizi
power_analysis = TTestIndPower()
# Etki büyüklüğünü doğrudan Cohen's d değerinden alıyoruz
power = power_analysis.solve_power(effect_size=abs(cohen_d), # Etki büyüklüğü pozitif olmalı
                                  nobs1=n_a,
                                  alpha=0.05)

# Görselleştirme
plt.figure(figsize=(16, 6))

# 1. Görev tamamlama süreleri
plt.subplot(1, 3, 1)
sns.violinplot(x='Site', y='Görev_Süresi', data=df,hue='Site', palette=["#1f77b4", "#ff7f0e"])
sns.swarmplot(x='Site', y='Görev_Süresi', data=df, color='.25', size=3, alpha=0.7)
plt.title('Görev Tamamlama Süreleri\n')
plt.ylabel('Süre (saniye)')

# 2. Memnuniyet puanları
plt.subplot(1, 3, 2)
sns.boxplot(x='Site', y='Memnuniyet', data=df, hue='Site', palette=["#1f77b4", "#ff7f0e"])
plt.title('Kullanıcı Memnuniyet Puanları\n')
plt.ylabel('Memnuniyet (1-10)')
plt.ylim(0, 10)

# 3. Süre-Memnuniyet ilişkisi
plt.subplot(1, 3, 3)
sns.scatterplot(x='Görev_Süresi', y='Memnuniyet', hue='Site', data=df,
                palette=["#1f77b4", "#ff7f0e"], alpha=0.7)
plt.title('Süre-Memnuniyet İlişkisi\n')
plt.xlabel('Görev Süresi (saniye)')
plt.ylabel('Memnuniyet Puanı')

plt.tight_layout()
plt.show()

# Sonuç raporu
print(f"""
📊 WEB SİTESİ KULLANILABİLİRLİK ANALİZ SONUÇLARI
--------------------------------------------

⏱️ GÖREV TAMAMLAMA SÜRELERİ:
---------------------------------
Web Sitesi A (n={n_a}):
- Ortalama: {np.mean(site_a):.1f} saniye
- Standart Sapma: {np.std(site_a, ddof=1):.1f} saniye
- Normallik: {'Normal' if normal_a else 'Normal Değil'}

Web Sitesi B (n={n_b}):
- Ortalama: {np.mean(site_b):.1f} saniye
- Standart Sapma: {np.std(site_b, ddof=1):.1f} saniye
- Normallik: {'Normal' if normal_b else 'Normal Değil'}

📈 İSTATİSTİKSEL TEST SONUÇLARI (SÜRELER):
---------------------------------
Test Türü: {test_adi}
Test İstatistiği: {t_test['U'].iloc[0] if test_adi.startswith('Mann') else t_test['T'].iloc[0]:.3f}
p-değeri: {t_test['p-val'].iloc[0]:.4f}
Etki Büyüklüğü (Cohen's d): {cohen_d:.3f}
Test Gücü (1-β): {power:.2%}

💡 YORUM (SÜRELER):
{'✅' if t_test['p-val'].iloc[0] < 0.05 else '❌'} {'Fark istatistiksel olarak ANLAMLI' if t_test['p-val'].iloc[0] < 0.05 else 'Fark istatistiksel olarak ANLAMSIZ'}
(p = {t_test['p-val'].iloc[0]:.4f}, α = 0.05)

🔍 PRATİK ETKİ (SÜRELER):
Web Sitesi B, Web Sitesi A'ya göre ortalama {np.mean(site_a)-np.mean(site_b):.1f} saniye daha hızlı.
Bu fark, Cohen's d standartlarına göre {'küçük' if abs(cohen_d)<0.5 else 'orta' if abs(cohen_d)<0.8 else 'büyük'} bir etki büyüklüğüne sahip.

😊 KULLANICI MEMNUNİYET SONUÇLARI:
---------------------------------
Web Sitesi A:
- Ortalama Memnuniyet: {np.mean(memnuniyet_a):.1f}/10

Web Sitesi B:
- Ortalama Memnuniyet: {np.mean(memnuniyet_b):.1f}/10

Mann-Whitney U Testi:
- p-değeri: {memnuniyet_test['p-val'].iloc[0]:.4f}
- Sonuç: {'Memnuniyetler arasında anlamlı fark var' if memnuniyet_test['p-val'].iloc[0] < 0.05 else 'Memnuniyetler arasında anlamlı fark yok'}

📌 TAVSİYELER:
1. {'Web Sitesi B açıkça daha iyi performans gösteriyor.' if t_test['p-val'].iloc[0] < 0.05 else 'Performans farkı istatistiksel olarak anlamlı değil.'}
2. Kullanıcı memnuniyeti Web Sitesi B'de {'önemli ölçüde daha yüksek' if memnuniyet_test['p-val'].iloc[0] < 0.05 else 'benzer'}.
3. Hız ve memnuniyet arasındaki negatif korelasyon {'doğrulanmıştır' if np.corrcoef(df['Görev_Süresi'], df['Memnuniyet'])[0,1] < -0.3 else 'zayıftır'}.
""")
````

````python
import numpy as np
from scipy import stats

# Örnek veriler (Web Sitesi A ve B için performans süreleri ve memnuniyet puanları)
# Bu verileri kendi gerçek verilerinizle değiştirin
siteA_performance = np.array([85, 78, 92, 65, 70, 88, 75, 82, 79, 91])  # saniye
siteB_performance = np.array([60, 55, 58, 63, 52, 59, 61, 57, 56, 54])  # saniye

siteA_satisfaction = np.array([3.2, 3.5, 3.0, 3.7, 3.1, 3.4, 3.3, 3.6, 3.2, 3.5])  # 5 üzerinden
siteB_satisfaction = np.array([4.8, 4.7, 4.9, 4.5, 5.0, 4.6, 4.8, 4.7, 4.9, 4.8])  # 5 üzerinden

# Performans için t-testi
t_stat_perf, p_value_perf = stats.ttest_ind(siteA_performance, siteB_performance, equal_var=False)

# Memnuniyet için t-testi
t_stat_satis, p_value_satis = stats.ttest_ind(siteA_satisfaction, siteB_satisfaction, equal_var=False)

# Sonuçları yazdır
print("PERFORMANS KARŞILAŞTIRMASI (Web Sitesi A vs B)")
print(f"Ortalama performans süresi - A: {np.mean(siteA_performance):.1f} saniye")
print(f"Ortalama performans süresi - B: {np.mean(siteB_performance):.1f} saniye")
print(f"Performans farkı: {np.mean(siteA_performance)-np.mean(siteB_performance):.1f} saniye")
print(f"t-istatistiği: {t_stat_perf:.3f}, p-değeri: {p_value_perf:.5f}")

print("\nMEMNUNİYET KARŞILAŞTIRMASI (Web Sitesi A vs B)")
print(f"Ortalama memnuniyet puanı - A: {np.mean(siteA_satisfaction):.1f}")
print(f"Ortalama memnuniyet puanı - B: {np.mean(siteB_satisfaction):.1f}")
print(f"Memnuniyet farkı: {np.mean(siteB_satisfaction)-np.mean(siteA_satisfaction):.1f}")
print(f"t-istatistiği: {t_stat_satis:.3f}, p-değeri: {p_value_satis:.5f}")

# Anlamlılık değerlendirmesi
alpha = 0.05
print("\nSONUÇ:")
if p_value_perf < alpha:
    print("Performans süreleri arasındaki fark istatistiksel olarak anlamlıdır (p < 0.05).")
else:
    print("Performans süreleri arasındaki fark istatistiksel olarak anlamlı değildir.")

if p_value_satis < alpha:
    print("Memnuniyet puanları arasındaki fark istatistiksel olarak anlamlıdır (p < 0.05).")
else:
    print("Memnuniyet puanları arasındaki fark istatistiksel olarak anlamlı değildir.")
````

5. **İki Ürün Serisi Karşılaştırması:**

- Problem: Bir şirket, iki farklı ürün serisini piyasaya sürmek istiyor.

- Gruplar: İlk grup, Ürün Serisi A'yı satın alan müşteriler, ikinci grup ise Ürün Serisi B'yi satın alan müşteriler.

- Hipotez: İki ürün serisi arasında müşteri memnuniyetinde anlamlı bir fark var mı?

Ürün Serisi A ve Ürün Serisi B'nin müşteri memnuniyet puanlarını karşılaştırmak için bağımsız iki örneklem t-testi uygulayan Python kodu:


````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Örnek veriler (Ürün Serisi A ve B için müşteri memnuniyet puanları)
# Gerçek verilerinizle bu kısmı değiştirin
seriA_memnuniyet = np.array([4.2, 4.5, 3.8, 4.1, 4.3, 3.9, 4.0, 4.4, 4.1, 4.6, 3.7, 4.3])
seriB_memnuniyet = np.array([4.6, 4.8, 4.5, 4.9, 4.7, 4.4, 4.8, 4.6, 4.7, 4.5, 4.9, 4.7])

# Normallik testi (Shapiro-Wilk)
_, pA = stats.shapiro(seriA_memnuniyet)
_, pB = stats.shapiro(seriB_memnuniyet)

print("NORMALLİK TESTİ SONUÇLARI:")
print(f"Ürün Serisi A p-değeri: {pA:.4f}")
print(f"Ürün Serisi B p-değeri: {pB:.4f}")
````

````python
# Varyans homojenliği testi (Levene testi)
_, p_var = stats.levene(seriA_memnuniyet, seriB_memnuniyet)

print(f"\nVARYANS HOMOJENLİĞİ TESTİ (Levene) p-değeri: {p_var:.4f}")
````

````python
# Bağımsız iki örneklem t-testi
# equal_var=False (Welch'in t-testi) varyanslar eşit değilse otomatik olarak kullanılır
t_stat, p_value = stats.ttest_ind(seriA_memnuniyet, seriB_memnuniyet, equal_var=False)

# Betimsel istatistikler
meanA, meanB = np.mean(seriA_memnuniyet), np.mean(seriB_memnuniyet)
stdA, stdB = np.std(seriA_memnuniyet, ddof=1), np.std(seriB_memnuniyet, ddof=1)
nA, nB = len(seriA_memnuniyet), len(seriB_memnuniyet)

print("\nBETİMSEL İSTATİSTİKLER:")
print(f"Ürün Serisi A (n={nA}): Ortalama = {meanA:.2f}, Standart Sapma = {stdA:.2f}")
print(f"Ürün Serisi B (n={nB}): Ortalama = {meanB:.2f}, Standart Sapma = {stdB:.2f}")
print(f"Ortalama farkı (B - A): {meanB-meanA:.2f}")

print("\nT-TESTİ SONUÇLARI:")
print(f"t-istatistiği: {t_stat:.3f}")
print(f"p-değeri: {p_value:.5f}")
````

````python
# Anlamlılık değerlendirmesi
alpha = 0.05
print("\nSONUÇ:")
if p_value < alpha:
    print(f"İki ürün serisi arasında müşteri memnuniyeti açısından istatistiksel olarak anlamlı bir fark vardır (p < {alpha}).")
    print(f"Ürün Serisi B ({meanB:.2f}), Ürün Serisi A'ya ({meanA:.2f}) göre anlamlı derecede {'daha yüksek' if meanB > meanA else 'daha düşük'} memnuniyet puanına sahiptir.")
else:
    print(f"İki ürün serisi arasında müşteri memnuniyeti açısından istatistiksel olarak anlamlı bir fark yoktur (p > {alpha}).")
````

````python
# Görselleştirme
plt.figure(figsize=(10, 6))
sns.boxplot(data=[seriA_memnuniyet, seriB_memnuniyet], palette=["lightblue", "lightgreen"])
plt.xticks([0, 1], ['Ürün Serisi A', 'Ürün Serisi B'])
plt.ylabel('Memnuniyet Puanı (1-5)')
plt.title('İki Ürün Serisi Müşteri Memnuniyeti Karşılaştırması')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

Kodun Açıklaması:
Veri Hazırlığı: Her iki ürün serisi için müşteri memnuniyet puanları numpy dizileri olarak tanımlanır.

Normallik Testi: Shapiro-Wilk testi ile verilerin normal dağılıma uygunluğu kontrol edilir.

Varyans Homojenliği Testi: Levene testi ile varyansların homojenliği kontrol edilir.

t-Testi: Welch'in t-testi (varyanslar eşit değilse otomatik olarak kullanılır) uygulanır.

Betimsel İstatistikler: Her iki grup için ortalama, standart sapma ve örneklem büyüklükleri hesaplanır.

Sonuçların Yorumlanması: 0.05 anlamlılık düzeyine göre sonuçlar yorumlanır.

## İş Uygulaması : ML Modelinin Başarı Testi (AB Testi)

Problem : 

Bir ML projesine yatırım yapılmış. Ürettiği tahminler neticesinde oluşan gelir ile eski sistemin ürettiği gelirler karşılaştırılıp anlamlı farklılık olup olmadığı test edilmek isteniyor.

Detaylar:

Model geliştirilmiş ve web sitesine entefre edilmiş.

Site kullanıcıları belirli bir kurala göre, online bir örnekleme tekniği izlenerek web sitesinde kullanıclar ikiye bölünmüş olsun.

A Grubu - Eski sisteme göre ürünler gösteriliyor

B Grubu - Yeni sisteme göre ürünler gösteriliyor

Gelir anlamında anlamlı bir iş yapılıp yapılmadığı test edilmek isteniyor.


Attığımız taş ürküttüğümüz kurbağaya değdi mi? Verielen emeğin karşılığı var mı?


H₀: İki grup ortalaması arasında fark yoktur (μ₁ = μ₂).
H₁: İki grup ortalaması arasında fark vardır (μ₁ ≠ μ₂) 


````python
import numpy as np
import pandas as pd


#veri tipi 1 - excel , csv veya veritababından okuduğumuzu varsayalım
#veri bu şekilde elimize geldiği için veri tiplerini dönüştürme işlemi yapmamız gerekiyor




A=pd.DataFrame([30,27,21,27,29,30,20,20,27,32,35,22,24,23,25,27,23,27,23,25,21,18,24,26,33,26,27,28,19,25])
B=pd.DataFrame([37,39,31,31,34,38,30,36,29,28,38,28,37,37,30,32,31,31,27,32,33,33,33,31,32,33,26,32,33,29])
A_B=pd.concat([A,B],axis=1)
A_B.columns=['A','B']
A_B.head()
````

````python
#veri tipi 2 - elimizdeki veri bu şekilde ise


#A ve B gruplarının karşılaştırılması
#A ve B gruplarının ortalamaları arasında istatistiksel olarak anlamlı bir fark olup olmadığını test etmek istiyoruz.
#Bu test için bağımsız iki örneklem t testi kullanacağız.
#H0: M1 = M2 (iki grup ortalamaları arasında istatistiksel olarak anlamlı bir fark yoktur)
#H1: M1 != M2 (iki grup ortalamaları arasında istatistiksel olarak anlamlı bir fark vardır)


#A pandas DataFrame is created containing a single column of numeric values. These represent hypothetical income data (labeled "gelir").
A=pd.DataFrame([30,27,21,27,29,30,20,20,27,32,35,22,24,23,25,27,23,27,23,25,21,18,24,26,33,26,27,28,19,25])
B=pd.DataFrame([37,39,31,31,34,38,30,36,29,28,38,28,37,37,30,32,31,31,27,32,33,33,33,31,32,33,26,32,33,29])

#A ve A'nın grubu
GRUP_A =np.arange(len(A))   #Creating a Group Identifier , creates an array of integers from 0 to len(A)-1 (0 to 29 in this case, as A has 30 rows
GRUP_A=pd.DataFrame(GRUP_A)  # GRUP_A is converted to a DataFrame
GRUP_A[:]='A'               # All values in GRUP_A are set to 'A'
A=pd.concat([A,GRUP_A],axis=1)  # A and GRUP_A are concatenated along axis 1 (i.e. columns)

#B ve B'nın grubu
GRUP_B =np.arange(len(B))
GRUP_B=pd.DataFrame(GRUP_B)
GRUP_B[:]='B'
B=pd.concat([B,GRUP_B],axis=1)


#Combining A and B into a Single Dataset
#tum veri
AB=pd.concat([A,B])   # Both A and B are concatenated into a single DataFrame AB
AB.columns=['gelir','GRUP']    # renames the columns to gelir (income) and GRUP (group).
#AB[:10]
print(AB.head())
print(AB.tail())

"""
Hipotez Testi Adımları  
"""
````

## <span style ="color:red">**Hipotez Testi**</span>

````python
stats.ttest_ind(A_B['A'], A_B['B'], equal_var = True) #bağımsız iki örneklem t testi uyguluyoruz, bu fonsiyonun içerisine testini yapmak istediğimiz iki değişkeni ve varyans homojenliği varsayımını giriyoruz. sol taraftaki değer test istatistiğini, sağ taraftaki değer ise p-value değerini verir
````

`stats.ttest_ind` fonksiyonu, iki bağımsız örneklem arasında istatistiksel olarak anlamlı bir fark olup olmadığını test etmek için kullanılır. Bu fonksiyon, Student'un t-testini uygular.

İki ana parametre alır:

1. **İlk grup verileri (`A_B['A']`):** Bu, karşılaştırma yapmak istediğiniz birinci grup verileridir.
   
2. **İkinci grup verileri (`A_B['B']`):** Bu, karşılaştırma yapmak istediğiniz ikinci grup verileridir.

`equal_var` parametresi, varsayılan olarak `True` olarak ayarlanmıştır. Bu durumda, iki grup için varyansın eşit olduğunu varsayar. Eğer `equal_var=False` olarak ayarlanırsa, varyansların eşit olmadığı kabul edilir ve Welch's t-testi uygulanır.

Sonuçlar arasında bir t-test istatistiği ve p değeri bulunur. P değeri, iki grup arasındaki farkın istatistiksel olarak anlamlı olup olmadığını belirler. Düşük bir p değeri (< 0.05 genellikle kullanılır) genellikle iki grup arasındaki farkın istatistiksel olarak anlamlı olduğunu gösterir.

Bu kod, `A_B` veri çerçevesindeki 'A' ve 'B' grupları arasındaki t-test istatistiğini ve p değerini hesaplar.

````python
#A_B veri setinde A ve B grupları
t_statistic, p_value = stats.ttest_ind(A_B['A'], A_B['B'], equal_var=True)
print(f"T-Statistic: {t_statistic:.4f}, P-Value: {p_value:.9f}")

print('p<0.05 ,H0 hipotezini reddettik yani ortalamaları eşit değil')
````

**Dolayısı ile eski sistem ile yeni sistem arasında gelirler göz önüne alındığında istatistiki olarak anlamlı bir farklılık vardır ve bu yüzden H0 reddedilir. Bu fark yeni sistemin lehinedir. Ortalamaları incelediğimizde de zaten bunu gözlemlemiştik.**

## <span style ="color:red">**Nonparametrik Bağımsız İki Örneklem Testi**</span>


- **Varsayımlar gerçekleşmediği zaman yapılır.**

`stats.mannwhitneyu` fonksiyonu, Mann-Whitney U testini uygular. Mann-Whitney U testi, iki bağımsız örneklem arasında medyan değerlerinin istatistiksel olarak farklı olup olmadığını test etmek için kullanılır. Bu test, özellikle bağımsız örneklem t-testinin varsayımlarının karşılanmadığı durumlarda (örneğin, veriler normal dağılmıyorsa) kullanışlıdır.

Fonksiyon, iki ana parametre alır:

1. **İlk grup verileri (`A_B['A']`):** Bu, karşılaştırma yapmak istediğiniz birinci grup verileridir.
   
2. **İkinci grup verileri (`A_B['B']`):** Bu, karşılaştırma yapmak istediğiniz ikinci grup verileridir.

Sonuç olarak, Mann-Whitney U istatistiği ve p değeri döner. P değeri, iki grup arasındaki medyan farkının istatistiksel olarak anlamlı olup olmadığını belirler. Düşük bir p değeri (< 0.05 genellikle kullanılır) genellikle iki grup arasındaki medyan farkının istatistiksel olarak anlamlı olduğunu gösterir.

````python
#varsayalım kigerçekleştirdiğim varsayım testleri sonucu negatif oldu ve normal dağılım ve varyans homojenliği varsayımlarının sağlandığını göremedik

#Bu durumda non-parametrik bağımsız iki örneklem t testi yapmamız gerekmektedir

stats.mannwhitneyu(A_B['A'], A_B['B']) #non-parametrik bağımsız iki örneklem t testi uyguluyoruz, bu fonsiyonun içerisine testini yapmak istediğimiz iki değişkeni giriyoruz. sol taraftaki değer test istatistiğini, sağ taraftaki değer ise p-value değerini verir
````

````python
# A_B veri çerçevesinde 'A' ve 'B' grupları
u_statistic, p_value = stats.mannwhitneyu(A_B['A'], A_B['B'])

print(f"Mann-Whitney U Statistic: {u_statistic:.4f}, P-Value: {p_value:.9f}")


# Bu kod, `A_B` veri çerçevesindeki 'A' ve 'B' grupları arasında Mann-Whitney U istatistiğini ve p değerini hesaplar.
````
