## 🔍 Tek Örneklem Oran Testi Nedir?

**Tek örneklem oran testi**, kategorik verilerdeki oranı hipotezle karşılaştırır.

📌 Örneğin:
- Yeni kampanya dönüşüm oranı arttı mı?
- Bir ilacı alan hastaların oranı beklenen orandan farklı mı?
- Anket sonucu popülasyonun beklentisinden sapmış mı?

📌 Bu testte kullandığımız dağılım: **normal dağılım (Z-testi)**

## <span style ="color:red">**Tek Örneklem Oran Testi**</span>

Oransal bir ifade test edilmek istenildiğinde kullanılır.

Tek Örneklem Oran Testi (One-Sample Proportion Test), bir popülasyondaki belirli bir özelliğin oranının, hipotez edilen bir değere eşit olup olmadığını test etmek için kullanılan bir istatistiksel testtir. Başka bir deyişle, bir örneklemdeki belirli bir özelliğin görülme sıklığının, popülasyonda beklenen orandan farklı olup olmadığını inceler.

**Tek Örneklem Oran Testi Ne Zaman Kullanılır?**

- Bir popülasyondaki bir oranın bilinen veya hipotez edilen bir değerle karşılaştırılması gerektiğinde.
- Veriler kategorik olduğunda (yani, iki veya daha fazla kategoriye ayrılabilir olduğunda). Örneğin, "evet/hayır", "başarılı/başarısız", "kız/erkek" gibi.
- Örneklem büyüklüğü yeterince büyük olduğunda.

**Tek Örneklem Oran Testinin Varsayımları:**

- **Rastgele Örnekleme:** Örneklem, popülasyondan rastgele seçilmelidir.
- **Bağımsızlık:** Her bir gözlem diğerlerinden bağımsız olmalıdır.
- **Yeterince Büyük Örneklem:** Başarı ve başarısızlık sayılarının her ikisi de en az 10 olmalıdır (bazı kaynaklarda 5 olarak da belirtilir). Bu koşul, normal yaklaşımın kullanılabilmesi için gereklidir. Bu koşul sağlanmıyorsa, binom testi gibi alternatif testler kullanılmalıdır.
- **n > 30** İlgilendiğimiz kitlede toplamış olduğumuz örneklem 30'dan büyük ise bu durumda bu testi gerçekleştirebilriiz.







**Tek Örneklem Oran Testi Nasıl Yapılır?**

1. **Sıfır Hipotezi (H₀) ve Alternatif Hipotezi (H₁) Belirleyin:**
    
    - H₀: Popülasyon oranı, hipotez edilen değere eşittir (p = p₀).
    - H₁: Popülasyon oranı, hipotez edilen değerden farklıdır (p ≠ p₀) (İki yönlü test). Veya, popülasyon oranı hipotez edilen orandan büyüktür (p > p₀) veya küçüktür (p < p₀) (Tek yönlü test).
2. **Anlamlılık Düzeyini (α) Belirleyin:** Genellikle α = 0.05 kullanılır.
    
3. **Test İstatistiğini Hesaplayın:**
    
    ```
    z = (p̂ - p₀) / √(p₀(1 - p₀) / n)
    ```
    
    Burada:
    
    - p̂: Örneklem oranı (başarı sayısı / örneklem büyüklüğü)
    - p₀: Hipotez edilen popülasyon oranı
    - n: Örneklem büyüklüğü
4. **Kritik Z Değerini Bulun:** Standart normal dağılım (Z) tablosundan veya istatistiksel yazılımdan, belirlenen anlamlılık düzeyi (α) için kritik z değerini bulunur. İki yönlü testlerde, α/2 kullanılarak iki kritik değer bulunur (pozitif ve negatif).
    
5. **Karar Verin:**
    
    - Eğer hesaplanan z istatistiğinin mutlak değeri, kritik z değerinden büyükse, sıfır hipotezi reddedilir. Bu, örneklem oranının hipotez edilen popülasyon oranından istatistiksel olarak anlamlı bir şekilde farklı olduğu anlamına gelir.
    - Eğer hesaplanan z istatistiğinin mutlak değeri, kritik z değerinden küçükse, sıfır hipotezi reddedilmez. Bu, örneklem oranının hipotez edilen popülasyon oranından farklı olmadığı anlamına gelmez, sadece elimizdeki verilerin bu farkı göstermek için yeterli kanıt sağlamadığı anlamına gelir.


## 📐 Formülü:

$$
z = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0(1 - p_0)}{n}}}
$$

| Sembol | Açıklama |
|--------|----------|
| $\hat{p}$ | Örneklem oranı (başarı sayısı / örneklem büyüklüğü) |
| $p_0$ | Popülasyon oranı (hipotezdeki değer) |
| $n$ | Örneklem büyüklüğü |

📌 Eğer `|z| > 1.96` ve `p < 0.05` → H₀ reddedilir

## 🧠 Neden Önemlidir?

| Sebep | Açıklama |
|--------|----------|
| ✅ **Kategorik veride karar verme** | Evet/hayır, başarılı/başarısız gibi durumlarda kullanılır |
| ✅ **A/B testlerinde** | Yeni tasarım daha iyi mi? Yoksa eski versiyon mu doğru? |
| ✅ **Finans ve pazarlama** | Dönüşüm oranı, müşteri memnuniyeti vs. |
| ✅ **Ankette güvenilirlik** | Sonuçlar gerçekten anlamlı mı? |

📌 Her oran testi, aslında bir **karar verme sürecidir.**

## 🛠️ Adım Adım Uygulama: Gerçek Hayat Senaryosu

### Örnek:
Bir siyasi partinin şehirdeki desteği %60 iddiası var.  
Sen 200 kişiye soruyorsun → 100 kişi destekliyor → `%50`

### **Örnek:**

Bir anketör, bir şehirdeki yetişkinlerin %60'ının bir siyasi partiyi desteklediğini iddia etmektedir. Rastgele seçilen 200 yetişkinden oluşan bir örneklemde, 100 kişinin bu partiyi desteklediği bulunmuştur. Bu iddiayı test etmek için Tek Örneklem Oran Testi kullanalım (α = 0.05):

1. **Hipotezler:**
    
    - H₀: p = 0.60
    - H₁: p ≠ 0.60
2. **Örneklem Oranı:**
    
    p̂ = 100 / 200 = 0.50
    
3. **Test İstatistiği:**
    
    ```
    z = (0.50 - 0.60) / √(0.60 * (1 - 0.60) / 200)
      = -0.10 / √(0.24 / 200)
      = -0.10 / 0.0346
      ≈ -2.89
    ```
    
4. **Kritik Z Değeri:**
    
    α = 0.05 için iki yönlü testte kritik z değerleri ±1.96'dır.
    
5. **Karar:**
    
    Hesaplanan z istatistiğinin mutlak değeri (2.89), kritik z değerinden (1.96) büyüktür. Bu nedenle, sıfır hipotezi reddedilir. Bu, örneklemdeki %50'lik oranın, anketörün iddiası olan %60'lık orandan istatistiksel olarak anlamlı bir şekilde farklı olduğunu gösterir.

#### Soru:
%50’lik oran, %60’lık iddiayı çürütmeye yetecek kadar mı uzak?

### Adım 1: Hipotez Kurulumu

```python
print("H₀: Destek oranı = 0.60")
print("H₁: Destek oranı ≠ 0.60")
```

✅ Basitçe:
- H₀ → Parti destek oranı gerçekten %60
- H₁ → Destek oranı farklı

Adım 2: Anlamlılık Düzeyini Belirle

````python
alpha = 0.05
print(f"Anlamlılık düzeyi (α): {alpha}")
````

📌 %5 yanılma payı kabul ettik

Adım 3: Sample Veri Üret

````python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Hipotez edilen oran
p_0 = 0.60
orneklem_buyuklugu = 200
basari_sayisi = 100  # %50 oran
````

📌 Bu veri üzerinden oran testi yapacağız.

Adım 4: Z-Testi Yap

````python
z_skoru, p_degeri = proportions_ztest(count=basari_sayisi, nobs=orneklem_buyuklugu, value=p_0)
print(f"Z Skoru: {z_skoru:.2f}, p-değeri: {p_degeri:.4f}")
````

📌 Bu değerler, iddianın gerçekliği hakkında ipucu veriyor.

Adım 5: Karar Ver

````python
if p_degeri < alpha:
    print("✅ H₀ reddedildi → Oran istatistiksel olarak farklı")
else:
    print("❌ H₀ reddedilemedi → Oran tahmin edilen değerden çok uzak değil")
````

📌 T-testi → ortalama fark  
Z-testi → oran farkı



### **Örnek:**

Örneğin, bir üniversite mezunlarının iş bulma oranı hakkında bir popülasyon tahmini yapmak istediğinizi düşünelim. Varsayalım ki, genel popülasyonun iş bulma oranının %75 olduğunu düşünüyorsunuz. Ancak, sadece belirli bir örnekleme dayalı olarak elde ettiğiniz iş bulma oranının genel popülasyondan istatistiksel olarak farklı olup olmadığını test etmek istiyorsunuz.

  

Bu durumu değerlendirmek için Tek Örneklem Oran Testi şu adımları takip eder:

  

1. **Hipotez Kurma:**

- **Null Hipotezi (H0):** Örneklem oranı, popülasyon oranı ile aynıdır.

- **Alternatif Hipotez (H1):** Örneklem oranı, popülasyon oranından istatistiksel olarak farklıdır.

  

2. **Veri Toplama:**

- Belirli bir örneklemden ilgili kategorik özellik (örneğin, iş bulma durumu) verilerini topla.

Oransal bir ifade test edilmek istediğinde kullanılır.

  

Tek Örneklem Oran Testi'nin formülü şu şekildedir:

  
  
  

## $z = \frac{\hat{p} - p_0}{\sqrt{\frac{\hat{p}(1-\hat{p})}{n}}}$

  

Bu formülde:

  

- $z$: Test istatistiğini temsil eder.

- $p$: Örneklem oranını temsil eder.

- $p0$: Popülasyon oranını temsil eder (null hipotezde belirtilen değer).

- $n$: Örneklem büyüklüğünü temsil eder.

  

4. **P Değerini Hesaplama:**

- Hesaplanan z-istatistiği üzerinden p değerini belirle.

  

5. **Karar Verme:**

- Belirlenen anlamlılık düzeyi (örneğin, \(\alpha = 0.05\)) üzerinden p değerini karşılaştır ve null hipotezini reddet veya reddetmeme kararı ver.

  

Tek Örneklem Oran Testi, özellikle popülasyon oranına ilişkin bir önerme üzerinde istatistiksel çıkarımlar yapmak istediğiniz durumlarda kullanılır.

  

Bu formül, örneklem oranının popülasyon oranından ne kadar farklı olduğunu ölçen bir z-skoru üretir. Bu z-skorunu kullanarak, hipotez testi yapılır ve null hipotezi reddedilip edilmeyeceği belirlenir.>

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Stil ayarları

sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)

def tek_oran_testi(sample_success, sample_size, pop_prop, alpha=0.05, alternative='two-sided'):
    """
    Tek Örneklem Oran Testi Uygulaması
    
    Parametreler:
        sample_success (int): Örneklemdeki başarı sayısı
        sample_size (int): Örneklem büyüklüğü
        pop_prop (float): Popülasyon oranı (H0 değeri)
        alpha (float): Anlamlılık düzeyi
        alternative (str): 'two-sided', 'greater', 'less'
    """
    # Örneklem oranı
    sample_prop = sample_success / sample_size
    
    # Test istatistiği (z-score)
    z = (sample_prop - pop_prop) / np.sqrt((pop_prop * (1 - pop_prop)) / sample_size)
    
    # p-değeri hesaplama
    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(abs(z)))
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z)
    else:  # 'less'
        p_value = stats.norm.cdf(z)
    
    # Kritik değerler
    if alternative == 'two-sided':
        critical_value = stats.norm.ppf(1 - alpha/2)
    else:
        critical_value = stats.norm.ppf(1 - alpha)
    
    # Güven aralığı (Wald yöntemi)
    margin_of_error = stats.norm.ppf(1 - alpha/2) * np.sqrt((sample_prop * (1 - sample_prop)) / sample_size)
    ci_lower = sample_prop - margin_of_error
    ci_upper = sample_prop + margin_of_error
    
    # Sonuçları yazdırma
    print("📊 TEK ÖRNEKLEM ORAN TESTİ SONUÇLARI")
    print("-----------------------------------")
    print(f"Örneklem Başarı Sayısı: {sample_success}")
    print(f"Örneklem Büyüklüğü (n): {sample_size}")
    print(f"Örneklem Oranı (p̂): {sample_prop:.4f}")
    print(f"Popülasyon Oranı (p₀): {pop_prop:.4f}")
    print(f"Test İstatistiği (z): {z:.4f}")
    print(f"Kritik z-değeri: ±{critical_value:.4f}" if alternative == 'two-sided' else f"Kritik z-değeri: {critical_value:.4f}")
    print(f"p-değeri: {p_value:.6f}")
    print(f"%{(1-alpha)*100} Güven Aralığı: ({ci_lower:.4f}, {ci_upper:.4f})")
    
    # Karar ve yorum
    print("\n🔍 YORUM:")
    if p_value < alpha:
        print(f"✅ p < {alpha} → H₀ REDDEDİLİR")
        if alternative == 'two-sided':
            if sample_prop > pop_prop:
                print(f"   Örneklem oranı ({sample_prop:.4f}), popülasyon oranından ({pop_prop:.4f}) ANLAMLI ŞEKİLDE YÜKSEK")
            else:
                print(f"   Örneklem oranı ({sample_prop:.4f}), popülasyon oranından ({pop_prop:.4f}) ANLAMLI ŞEKİLDE DÜŞÜK")
        elif alternative == 'greater':
            print(f"   Örneklem oranı ({sample_prop:.4f}), popülasyon oranından ({pop_prop:.4f}) ANLAMLI ŞEKİLDE YÜKSEK")
        else:
            print(f"   Örneklem oranı ({sample_prop:.4f}), popülasyon oranından ({pop_prop:.4f}) ANLAMLI ŞEKİLDE DÜŞÜK")
    else:
        print(f"❌ p ≥ {alpha} → H₀ REDDEDİLEMEZ")
        print(f"   Örneklem oranının ({sample_prop:.4f}) popülasyon oranından ({pop_prop:.4f}) farklı olduğuna dair YETERLİ KANIT YOK")
    
    # Görselleştirme
    plt.figure(figsize=(12, 6))
    
    # 1. Oranların karşılaştırması
    plt.subplot(1, 2, 1)
    plt.bar(['Popülasyon (p₀)', 'Örneklem (p̂)'], [pop_prop, sample_prop], color=['red', 'green'], alpha=0.7)
    plt.errorbar(x=1, y=sample_prop, yerr=margin_of_error, fmt='o', color='black', 
                 capsize=5, label='Güven Aralığı')
    plt.title('Oran Karşılaştırması')
    plt.ylabel('Oran')
    plt.ylim(0, 1)
    plt.legend()
    
    # 2. z-dağılımı ve test sonucu
    plt.subplot(1, 2, 2)
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)
    plt.plot(x, y, label='Standart Normal Dağılım')
    
    # Kritik bölgeler
    if alternative == 'two-sided':
        plt.axvline(critical_value, color='red', linestyle=':', label=f'Kritik değer (±{critical_value:.2f})')
        plt.axvline(-critical_value, color='red', linestyle=':')
        plt.fill_between(x, y, where=(x >= critical_value), color='red', alpha=0.2, label='Red bölgesi')
        plt.fill_between(x, y, where=(x <= -critical_value), color='red', alpha=0.2)
    elif alternative == 'greater':
        plt.axvline(critical_value, color='red', linestyle=':', label=f'Kritik değer ({critical_value:.2f})')
        plt.fill_between(x, y, where=(x >= critical_value), color='red', alpha=0.2, label='Red bölgesi')
    else:
        plt.axvline(-critical_value, color='red', linestyle=':', label=f'Kritik değer (-{critical_value:.2f})')
        plt.fill_between(x, y, where=(x <= -critical_value), color='red', alpha=0.2, label='Red bölgesi')
    
    # Test istatistiği
    plt.axvline(z, color='green', linestyle='-', label=f'z-istatistiği ({z:.2f})')
    
    plt.title('z-Testi Sonuçları')
    plt.xlabel('z Değeri')
    plt.ylabel('Yoğunluk')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

````

````python
# Üniversite mezunları iş bulma oranı örneği
print("🎓 ÜNİVERSİTE MEZUNLARI İŞ BULMA ORANI ANALİZİ")
tek_oran_testi(sample_success=85, sample_size=100, pop_prop=0.75, alpha=0.05, alternative='two-sided')
````

## İş Uygulaması : Dönüşüm Oran Testi

Problem: 

Bir yazılım ile bir mecrada reklam verilmiş ve bu reklama ilişkin yazılım tarafından 0.125 dönüşüm oranı elde edildiği ifade edilmiş. Fakat bu durum kontrol edilmek isteniyor. Çünkü bu yüksek bir oran ve gelirler incelendiğinde örtüşmüyor.


Detaylar: 

500 kişi dış mecrada reklamlara tıklamış, 40 tanesini sitemize gelip alışveriş yapmış.

Örnek üzerinden elde edilen dönüşüm oranı: 40/500 = 0.08




## Varsayımlar

- **n>30 olmalı p^ :dönüşüm oranı P0 test edilmek istenen değer**

- **HO: p=0.125**
- **H1: p!=0.0125**

````python
from statsmodels.stats.proportion import proportions_ztest #iki oran arasında karşılaştırma yapmak için gerekli modülü import edelim
````

````python
count = 40 #başarı sayısı
nobs = 500 #gözlem sayısı
value = 0.125 #karşılaştırılacak değer
````

````python
proportions_ztest(count, nobs, value) #iki oran arasında karşılaştırma yapmak için bu fonksiyonu kullanıyoruz, bu fonksiyonun içerisine başarı sayısı, gözlem sayısı ve karşılaştırılacak değeri giriyoruz. sol taraftaki değer test istatistiğini, sağ taraftaki değer ise p-value değerini verir  
````

P-value değeri < alfa 0.05 değerinden daha küçük olduğundan dolayı H0 hipotezini reddediyoruz.
