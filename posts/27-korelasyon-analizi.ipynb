## 🔍 Korelasyon Nedir?

İki değişken arasında doğrusal ilişkinin **yönünü** ve **şiddetini** ölçen istatistiksel yöntemdir.

📌 Örnek:
- `fiyat` vs `satış_adedi`
- `çalışma_saati` vs `not`
- `reklam_harcaması` vs `tıklama`

📌 Korelasyon katsayısı `-1` ile `+1` arasında değer alır:
| Aralık | Anlam |
|--------|-------|
| **+1** | Mükemmel pozitif ilişki |
| **0** | İlişki yok |
| **-1** | Mükemmel negatif ilişki |

📌 Yüksek mutlak korelasyon = güçlü ilişki  
Yüksek olmayan = zayıf ilişki


## <span style ="color:red">**Kolerasyon Analizi**</span>

Korelasyon analizi, iki değişken arasındaki doğrusal ilişkinin yönünü ve gücünü ölçmek için kullanılan bir istatistiksel yöntemdir. Başka bir deyişle, iki değişkenin birlikte ne kadar değiştiğini ve bu değişimin yönünü inceler. Korelasyon, bir neden-sonuç ilişkisi göstermez, sadece iki değişken arasında bir ilişki olup olmadığını ve bu ilişkinin ne kadar güçlü olduğunu gösterir.

- Değişkenler arasındaki ilişki,bu ilişkinin yönü ve şiddeti ile ilgili bilgiler sağlayan istatistiksel bir yöntemdir.

- Korelasyon analizinde genellikle test edilen temel hipotez, iki değişken arasında bir ilişki olup olmadığıdır. Korelasyon analizi genellikle Pearson Korelasyon Katsayısı veya Spearman Sıra Korelasyon Katsayısı kullanılarak yapılır. İşte korelasyon analizi hipotezleri:

**Korelasyonun Temel Kavramları:**

- **Değişken:** Ölçülebilen ve farklı değerler alabilen özelliktir. Örneğin, boy, kilo, sıcaklık, gelir birer değişkendir.
- **Doğrusal İlişki:** İki değişken arasındaki ilişkinin bir doğru ile temsil edilebilmesidir.
- **Korelasyon Katsayısı (r):** İki değişken arasındaki doğrusal ilişkinin gücünü ve yönünü gösteren bir sayıdır. -1 ile +1 arasında değer alır.


**Korelasyon Katsayısının Yorumlanması:**

- **r = +1:** Mükemmel pozitif korelasyon. Bir değişken arttıkça, diğer değişken de aynı oranda artar.
- **0 < r < +1:** Pozitif korelasyon. Bir değişken arttıkça, diğer değişken de artma eğilimindedir.
- **r = 0:** Korelasyon yok. İki değişken arasında doğrusal bir ilişki yoktur.
- **-1 < r < 0:** Negatif korelasyon. Bir değişken arttıkça, diğer değişken azalma eğilimindedir.
- **r = -1:** Mükemmel negatif korelasyon. Bir değişken arttıkça, diğer değişken aynı oranda azalır.


**Korelasyon Türleri:**

- **Pearson Korelasyon Katsayısı (r):** En yaygın kullanılan korelasyon türüdür. İki sürekli değişken arasındaki doğrusal ilişkiyi ölçer. Verilerin normal dağılıma sahip olması varsayılır.
- **Spearman Sıra Korelasyon Katsayısı (ρ veya rs):** Sürekli veya sıralı veriler için kullanılır. Verilerin normal dağılım göstermesi gerekmez. İki değişken arasındaki sıralamaların benzerliğini ölçer.
- **Kendall Tau Korelasyon Katsayısı (τ):** Spearman korelasyonuna benzerdir, ancak farklı bir hesaplama yöntemi kullanır. Özellikle küçük örneklemlerde ve sıralı verilerde daha uygundur.

**Korelasyon Analizi Nasıl Yapılır?**

Korelasyon analizi genellikle istatistiksel yazılımlar (SPSS, R, Excel vb.) kullanılarak yapılır. Temel adımlar şunlardır:

1. **Verileri Hazırlama:** İki değişken için veri setini oluşturun.
2. **Korelasyon Katsayısını Hesaplama:** Uygun korelasyon katsayısını (Pearson, Spearman veya Kendall) seçerek hesaplayın.
3. **Anlamlılık Testi Yapma:** Korelasyon katsayısının istatistiksel olarak anlamlı olup olmadığını test edin. Bu genellikle bir hipotez testi (t-testi) ile yapılır.


Korelasyon analizi hipotezleri, iki değişken arasında doğrusal bir ilişki olup olmadığını test etmek için kullanılır. Temel olarak iki tür hipotez vardır: sıfır hipotezi ve alternatif hipotez.

**1. Sıfır Hipotezi (H₀):**

Sıfır hipotezi, iki değişken arasında _doğrusal bir ilişki olmadığını_ varsayar. Başka bir deyişle, korelasyon katsayısının (r) sıfıra eşit olduğunu iddia eder.

- **H₀: ρ = 0** (Popülasyon korelasyon katsayısı sıfırdır)
- **H₀: r = 0** (Örneklem korelasyon katsayısı sıfırdır)

Burada ρ (rho) popülasyon korelasyon katsayısını, r ise örneklem korelasyon katsayısını temsil eder.

**2. Alternatif Hipotez (H₁ veya Hₐ):**

Alternatif hipotez, iki değişken arasında _doğrusal bir ilişki olduğunu_ varsayar. Üç farklı şekilde formüle edilebilir:

- **Çift Yönlü Alternatif Hipotez:** İki değişken arasında bir ilişki _olduğunu_ ancak bu ilişkinin yönünün (pozitif veya negatif) belirtilmediğini varsayar.
    
    - **H₁: ρ ≠ 0** (Popülasyon korelasyon katsayısı sıfıra eşit değildir)
    - **H₁: r ≠ 0** (Örneklem korelasyon katsayısı sıfıra eşit değildir)
- **Tek Yönlü Alternatif Hipotez (Pozitif Korelasyon):** İki değişken arasında _pozitif bir ilişki olduğunu_ varsayar. Bir değişken arttıkça diğerinin de artacağını iddia eder.
    
    - **H₁: ρ > 0** (Popülasyon korelasyon katsayısı sıfırdan büyüktür)
    - **H₁: r > 0** (Örneklem korelasyon katsayısı sıfırdan büyüktür)
- **Tek Yönlü Alternatif Hipotez (Negatif Korelasyon):** İki değişken arasında _negatif bir ilişki olduğunu_ varsayar. Bir değişken arttıkça diğerinin azalacağını iddia eder.
    
    - **H₁: ρ < 0** (Popülasyon korelasyon katsayısı sıfırdan küçüktür)
    - **H₁: r < 0** (Örneklem korelasyon katsayısı sıfırdan küçüktür)



**Korelasyon Analizi Hipotez Testi Nasıl Yapılır?**

Korelasyon analizinde hipotez testi genellikle bir t-testi kullanılarak yapılır. Test istatistiği aşağıdaki formülle hesaplanır:

```
t = r / √[(1 - r²) / (n - 2)]
```

Burada:

- r: Örneklem korelasyon katsayısı
- n: Örneklem büyüklüğü

Hesaplanan t değeri, belirli bir serbestlik derecesi (n-2) ve anlamlılık düzeyi (α) için t dağılım tablosundaki kritik değerle karşılaştırılır.

- **Çift yönlü testte:** Hesaplanan |t| değeri, kritik değerden büyükse, sıfır hipotezi reddedilir ve iki değişken arasında istatistiksel olarak anlamlı bir ilişki olduğu sonucuna varılır.
- **Tek yönlü testte:** Hesaplanan t değeri, uygun yöndeki kritik değerden büyükse, sıfır hipotezi reddedilir ve iki değişken arasında belirtilen yönde istatistiksel olarak anlamlı bir ilişki olduğu sonucuna varılır.

Ayrıca, p-değeri de kullanılabilir. p-değeri anlamlılık düzeyinden (α) küçükse, sıfır hipotezi reddedilir.


## 📐 Korelasyon Türleri

### 1. **Pearson Korelasyon Katsayısı**
Sürekli ve normal dağılan veriler için uygun.

````python
import numpy as np
from scipy.stats import pearsonr

# Simülasyon: reklam harcaması ve satış adedi
np.random.seed(42)
reklam = np.random.normal(loc=50, scale=10, size=100)
satis = reklam * 1.3 + np.random.normal(0, 5, size=100)  # Pozitif ilişki

r, p_degeri = pearsonr(reklam, satis)
print(f"Pearson r: {r:.2f}, p-değeri: {p_degeri:.4f}")
````

📌 Yorum:
- Güçlü pozitif ilişki
- p-değeri < 0.05 → anlamlı

### 2. **Spearman Sıra Korelasyonu**
Veriler sıralı ya da normal dağılmıyorsa kullanılır.

````python
from scipy.stats import spearmanr

rho, p_val_rho = spearmanr(reklam, satis)
print(f"Spearman ρ: {rho:.2f}, p-değeri: {p_val_rho:.4f}")
````

📌 Spearman, ranklar üzerinden hesaplandığı için aykırı değerlere daha dayanıklıdır.

### 3. **Kendall’s Tau Korelasyonu**
Özellikle küçük örneklemlerde tercih edilir.

````python
from scipy.stats import kendalltau

tau, p_val_tau = kendalltau(reklam, satis)
print(f"Kendall τ: {tau:.2f}, p-değeri: {p_val_tau:.4f}")
````

📌 Kendall’s Tau, Pearson’a göre daha az duyarlıdır ama daha güvenilirdir.


✅ Null hipotezi → ilişki yok  
Alternatif → ilişki var

### Adım 2: Anlamlılık Düzeyini Belirle

````python
alpha = 0.05
print(f"Anlamlılık düzeyi (α): {alpha}")
````

### Adım 3: Sample Veri Üzerinden Bakalım

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "reklam": reklam,
    "satis": satis
})

print("Veri seti örnek:")
print(df.head())
````

### Adım 4: Korelasyon Matrisi Oluştur

````python
corr_matrix = df.corr(method='pearson')
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pearson Korelasyon Matrisi")
plt.show()
````


📌 Bu ısı haritasında:
- 0.89 → güçlü pozitif ilişki
- P-değeri çok düşük → anlamlı

### Adım 5: Korelasyon Testi Yap

````python
from scipy.stats import ttest_1samp

# Pearson r'den t-testi yap
n = len(reklam)
t_skoru = r / np.sqrt((1 - r**2) / (n - 2))
print(f"T Skoru: {t_skoru:.2f}")

# Serbestlik derecesi ve test
_, p_degeri = ttest_1samp(t_skoru, popmean=0)
print(f"p-değeri: {p_degeri:.4f}")






#  T-skorunu hesapla (bu formül doğru)
# RuntimeWarning'dan kaçınmak için (1 - r**2) nin 0'a yakın olmamasını kontrol edin,
# ki bu da korelasyonun 1 veya -1 olması durumudur.
# Eğer r tam olarak 1 veya -1 ise, (1-r**2) sıfır olur ve t_skoru tanımsız olur.
# Gerçek verilerde bu nadirdir.
if np.isclose(r**2, 1.0) and n == 2: # n=2 ise r=1 veya -1 olacaktır
    t_skoru = np.inf * np.sign(r) # r'nin işaretine göre inf veya -inf
elif np.isclose(r**2, 1.0): # n > 2 ise r'nin tam 1 veya -1 olması imkansızdır
    # Çok nadir durumda r^2 tam olarak 1 ise, buraya düşmez
    # ama yine de t_skoru formülü tanımsız hale gelir.
    t_skoru = np.inf * np.sign(r) # Veya duruma göre hata fırlatılabilir
else:
    t_skoru = r * np.sqrt((n - 2) / (1 - r**2))

print(f"Hesaplanan T Skoru: {t_skoru:.4f}")
````

### Adım 6: Karar Ver

````python
# if p_degeri < alpha:
#     print("✅ H₀ reddedildi → Korelasyon anlamlı")
# else:
#     print("❌ H₀ reddedilemedi → Korelasyon tesadüfi olabilir")


# --- Sonuç Yorumlama ---
if not np.isnan(p_degeri):
    if p_degeri < alpha:
        print(f"\nSonuç: p-değeri ({p_degeri:.4f}) < alpha ({alpha}) olduğundan,")
        print("Pearson korelasyon katsayısı istatistiksel olarak **anlamlıdır**.")
        print("Yani, reklam ve satışlar arasında anlamlı bir doğrusal ilişki vardır.")
    else:
        print(f"\nSonuç: p-değeri ({p_degeri:.4f}) >= alpha ({alpha}) olduğundan,")
        print("Pearson korelasyon katsayısı istatistiksel olarak **anlamlı değildir**.")
        print("Yani, reklam ve satışlar arasında anlamlı bir doğrusal ilişki olduğuna dair yeterli kanıt yoktur.")
else:
    print("\nSonuç: p-değeri hesaplanamadığı için anlamlılık yorumu yapılamıyor.")
````

### Örnek:
Bir ürün satıcısı iki değişkeni incelemek istiyor:
- `reklam_harcamasi`
- `satis_adedi`

#### Soru:
“Reklam harcaması artarsa, satış adedi de artıyor mu?”

````python
import numpy as np
from scipy import stats

# --- Simülasyon Verisi Oluşturma (Örnek Amaçlı) ---
# Gerçek senaryoda bu verileriniz olurdu
np.random.seed(42)
reklam_harcamasi = np.random.normal(loc=100, scale=20, size=50) # Örn. reklam harcaması
satislar = 0.8 * reklam_harcamasi + np.random.normal(loc=0, scale=10, size=50) # Reklamla ilişkili satışlar

# --- Pearson Korelasyon Katsayısını Hesaplama ---
r = np.corrcoef(reklam_harcamasi, satislar)[0, 1]
print(f"Hesaplanan Pearson Korelasyon Katsayısı (r): {r:.4f}")
````

````python
# --- Pearson r'nin Anlamlılığını Test Etme ---
# Adım 1: T-skorunu hesapla
n = len(reklam_harcamasi) # Örneklem büyüklüğü
# Korelasyon katsayısı için t-skoru formülü
t_skoru = r * np.sqrt((n - 2) / (1 - r**2))
print(f"Hesaplanan T Skoru: {t_skoru:.4f}")
````

````python
# Adım 2: Serbestlik derecesini belirle
df = n - 2 # Pearson korelasyonu için serbestlik derecesi
````

````python
# Adım 3: T-dağılımını kullanarak p-değerini bul
# stats.t.sf(abs(t_skoru), df) tek kuyruk p-değerini verir.
# Çift kuyruk (iki taraflı) test için 2 ile çarpılır.
p_degeri = 2 * stats.t.sf(abs(t_skoru), df)
print(f"p-değeri (iki taraflı): {p_degeri:.4f}")
````

````python
# --- Sonuç Yorumlama ---
alpha = 0.05 # Anlamlılık düzeyi

if p_degeri < alpha:
    print(f"\nSonuç: p-değeri ({p_degeri:.4f}) < alpha ({alpha}) olduğundan,")
    print("Pearson korelasyon katsayısı istatistiksel olarak **anlamlıdır**.")
    print("Yani, reklam harcaması ile satışlar arasında anlamlı bir doğrusal ilişki vardır.")
else:
    print(f"\nSonuç: p-değeri ({p_degeri:.4f}) >= alpha ({alpha}) olduğundan,")
    print("Pearson korelasyon katsayısı istatistiksel olarak **anlamlı değildir**.")
    print("Yani, reklam harcaması ile satışlar arasında anlamlı bir doğrusal ilişki olduğuna dair yeterli kanıt yoktur.")

# Not: Pingouin kütüphanesi bu işi çok daha basit yapar:
import pingouin as pg
corr_test = pg.corr(reklam_harcamasi, satislar)
print("\nPingouin ile Korelasyon Testi Sonucu:")
print(corr_test)
````

**Örnek:**

Bir araştırmacı, öğrencilerin sınav notları ile çalışma saatleri arasında bir ilişki olup olmadığını incelemek istiyor. 20 öğrencinin verilerini topluyor ve Pearson korelasyon katsayısını r = 0.7 olarak hesaplıyor. Anlamlılık düzeyi α = 0.05 olsun.

1. **Hipotezler:** H₀: ρ = 0, H₁: ρ > 0 (Tek yönlü, pozitif korelasyon testi)
2. **Test İstatistiği:** t = 0.7 / √[(1 - 0.7²) / (20 - 2)] ≈ 4.37
3. **Kritik Değer:** Serbestlik derecesi 18 ve α = 0.05 için tek yönlü t tablosundan kritik değer yaklaşık 1.73'tür.
4. **Karar:** Hesaplanan t değeri (4.37), kritik değerden (1.73) büyük olduğu için sıfır hipotezi reddedilir.

**Sonuç:** Öğrencilerin sınav notları ile çalışma saatleri arasında istatistiksel olarak anlamlı bir pozitif ilişki vardır.

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Örnek veri oluşturma (sizin durumunuzda gerçek verilerinizi kullanacaksınız)
np.random.seed(42)
calisma_saatleri = np.random.normal(10, 2, 20)  # Ortalama 10 saat, standart sapma 2
sinav_notlari = 50 + 2*calisma_saatleri + np.random.normal(0, 5, 20)  # Pozitif ilişki

# Pearson korelasyon katsayısı ve p-değeri hesaplama
r, p_value = stats.pearsonr(calisma_saatleri, sinav_notlari)

# Sonuçları yazdırma
print(f"Pearson korelasyon katsayısı (r): {r:.3f}")
print(f"P-değeri: {p_value:.4f}")
````

````python
# Hipotez testi sonucu
alpha = 0.05
if p_value < alpha:
    print(f"p-değeri {alpha}'dan küçük olduğu için H₀ reddedilir (anlamlı korelasyon var).")
else:
    print(f"p-değeri {alpha}'dan büyük olduğu için H₀ reddedilemez (anlamlı korelasyon yok).")
````

````python
# Serpilme diyagramı çizme
plt.figure(figsize=(8, 6))
plt.scatter(calisma_saatleri, sinav_notlari, color='blue')
plt.title('Çalışma Saatleri ve Sınav Notları Arasındaki İlişki')
plt.xlabel('Çalışma Saatleri')
plt.ylabel('Sınav Notları')
plt.grid(True)
````

### <span style ="color:yellow">**Pearson Korelasyon Katsayısı İçin Hipotezler**</span>

  

1. **Null Hypothesis (H0):** İki değişken arasında korelasyon yoktur (r = 0)

  

- $ H_0: \rho = 0 $


Pearson Korelasyon Katsayısı Nasıl Hesaplanır?

Pearson korelasyon katsayısının hesaplanması için birkaç formül bulunmaktadır. En yaygın kullanılan formüllerden biri şöyledir:

r = Σ[(xᵢ - x̄)(yᵢ - ȳ)] / √[Σ(xᵢ - x̄)² * Σ(yᵢ - ȳ)²]
Burada:

xᵢ ve yᵢ: Her bir veri noktasındaki değişkenlerin değerleri
x̄ ve ȳ: x ve y değişkenlerinin ortalamaları
Σ: Toplam sembolü
Bu formül, her bir veri noktasının ortalamadan sapmalarının çarpımının, bu sapmaların karelerinin çarpımının kareköküne bölünmesiyle elde edilir.

Pearson Korelasyon Katsayısının Yorumlanması:

Korelasyon katsayısının büyüklüğü, ilişkinin gücünü gösterir:

0.70 veya daha yüksek: Güçlü korelasyon
0.50 ile 0.69 arası: Orta düzeyde korelasyon
0.30 ile 0.49 arası: Zayıf korelasyon
0.00 ile 0.29 arası: Çok zayıf veya korelasyon yok
Ancak, bu aralıklar genel bir kılavuzdur ve yorumlama, incelenen alana ve bağlama göre değişebilir.


  

2. **Alternative Hypothesis (H1):** İki değişken arasında korelasyon vardır (r ≠ 0).

- $ H_1: \rho \neq 0 $

  

Bu durumda, \(\rho\) Pearson korelasyon katsayısıdır.


Eğer varsayım sağlanmıyorsa, Non-Parametrik Korelasyon ANalizi Hipotez Testi için  Spearman Korelasyon Katsayısını kullanırız.



### <span style ="color:yellow">**Spearman Sıra Korelasyon Katsayısı İçin Hipotezler**</span>

  

1. **Null Hypothesis (H0):** İki değişken arasında sıra korelasyonu yoktur (ρ = 0).

- $ H_0: \rho = 0 $

  

2. **Alternative Hypothesis (H1):** İki değişken arasında sıra korelasyonu vardır (ρ ≠ 0).

- $ H_1: \rho \neq 0 $

  

Bu durumda, $\rho$ Spearman sıra korelasyon katsayısıdır.

Bu hipotezler, korelasyonun varlığı veya yokluğunu test etmek amacıyla kullanılır. Eğer null hypothesis (H0) reddedilirse, bu, iki değişken arasında anlamlı bir korelasyon olduğu anlamına gelir. Bu hipotez testleri genellikle belirli bir anlamlılık düzeyi (alfa, genellikle 0.05) kullanılarak yapılır.

Korelasyon analizi, iki değişken arasındaki ilişkiyi değerlendiren bir istatistiksel yöntemdir. Korelasyon varsayımları, korelasyon katsayılarının güvenilir bir şekilde yorumlanabilmesi ve kullanılabilmesi için dikkate alınması gereken temel önermelerdir.


### **Korelasyon varsayımları:**

  

1. **Normallik Varsayımı:**

- Korelasyon katsayıları genellikle değişkenlerin normal dağılıma sahip olduğunda daha güvenilir olarak kabul edilir.

- Bu varsayım, korelasyon analizi sonuçları üzerindeki güvenilirliği artırmak için önemlidir.

  

2. **Doğrusallık Varsayımı:**

- Korelasyon analizi, iki değişken arasındaki ilişkinin doğrusal bir ilişki olduğunu varsayar. Eğer ilişki doğrusal değilse, korelasyon katsayıları güvenilir olmayabilir.

  

3. **Aykırı Gözlem ve Uç Değerlerin Etkisi:**

- Aykırı gözlemler ve uç değerler, korelasyon analizini etkileyebilir. Bu değerlerin analizden önce incelenmesi ve gerekirse düzeltilmesi önemlidir.

  

4. **Homoscedasticity (Homojen Varyans) Varsayımı:**

- Homoscedasticity, değişkenler arasındaki değişkenliğin eşit olduğu anlamına gelir. Korelasyon analizi için bu varsayım önemlidir.

  

5. **Bağımsızlık Varsayımı:**

- Gözlemler arasında bağımsızlık varsayımı önemlidir. Örneğin, zamanla ilişkili gözlemler yapıldığında, bu bağımsızlık varsayımı ihlal edilebilir.

  

Bu varsayımlar, korelasyon analizi sonuçlarının güvenilirliğini artırmaya yöneliktir. Ancak, bazı durumlarda bu varsayımlar tam olarak karşılanmazsa, alternatif yöntemler ve dikkatli yorumlar kullanılarak analiz yapılabilir. Örneğin, Spearman's rho veya Kendall's tau gibi parametrik olmayan(nonparametrik) korelasyon katsayıları, doğrusallık varsayımının karşılanmadığı durumlarda kullanılabilir.




**Korelasyon ve Nedensellik:**

Korelasyon, nedensellik anlamına gelmez. İki değişken arasında bir korelasyon olması, birinin diğerine neden olduğu anlamına gelmez. Örneğin, dondurma satışları ile boğulma vakaları arasında pozitif bir korelasyon olabilir, ancak dondurma yemenin boğulmaya neden olduğu anlamına gelmez. Bu durumda, her iki değişkeni de etkileyen üçüncü bir faktör (örneğin, sıcak hava) olabilir.


````python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Veri Oluşturma (Örnek veri - kendi verilerinizle değiştirebilirsiniz)
np.random.seed(42)
n = 50  # Örneklem büyüklüğü
calisma_saatleri = np.random.normal(15, 5, n)  # Ortalama 15 saat, std=5
sinav_notlari = 50 + 2*calisma_saatleri + np.random.normal(0, 10, n)  # Pozitif ilişki

# Veriyi DataFrame'e dönüştürme
df = pd.DataFrame({
    'Calisma_Saatleri': calisma_saatleri,
    'Sinav_Notlari': sinav_notlari
})
````

````python
# 2. Veri Ön İnceleme
print("\nVerinin İlk 5 Satırı:")
print(df.head())

print("\nTemel İstatistikler:")
print(df.describe())
````

````python
# 3. Normallik Testi (Shapiro-Wilk)
print("\nNormallik Testi Sonuçları:")
for col in df.columns:
    stat, p = stats.shapiro(df[col])
    print(f"{col}: W={stat:.3f}, p={p:.4f} - {'Normal' if p>0.05 else 'Normal değil'}")
````

````python
# 4. Görselleştirme
plt.figure(figsize=(12, 5))

# Histogramlar
plt.subplot(1, 2, 1)
sns.histplot(df['Calisma_Saatleri'], kde=True, color='blue', label='Çalışma Saatleri')
sns.histplot(df['Sinav_Notlari'], kde=True, color='red', label='Sınav Notları')
plt.title('Değişkenlerin Dağılımı')
plt.legend()

# Scatter plot
plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='Calisma_Saatleri', y='Sinav_Notlari', color='green')
plt.title('Çalışma Saatleri vs Sınav Notları')
plt.tight_layout()
plt.show()
````

````python
# 5. Korelasyon Analizi
# Pearson Korelasyonu (Parametrik)
pearson_r, pearson_p = stats.pearsonr(df['Calisma_Saatleri'], df['Sinav_Notlari'])

# Spearman Korelasyonu (Non-parametrik)
spearman_r, spearman_p = stats.spearmanr(df['Calisma_Saatleri'], df['Sinav_Notlari'])
````

````python
# 6. Sonuçların Yazdırılması
alpha = 0.05

print("\nKORELASYON ANALİZİ SONUÇLARI:")
print(f"\nPearson Korelasyon Katsayısı (r): {pearson_r:.3f}")
print(f"Pearson p-değeri: {pearson_p:.4f}")
if pearson_p < alpha:
    print(f"Pearson: p < {alpha} → H0 reddedilir (Anlamlı korelasyon var)")
else:
    print(f"Pearson: p >= {alpha} → H0 reddedilemez (Anlamlı korelasyon yok)")

print(f"\nSpearman Korelasyon Katsayısı (ρ): {spearman_r:.3f}")
print(f"Spearman p-değeri: {spearman_p:.4f}")
if spearman_p < alpha:
    print(f"Spearman: p < {alpha} → H0 reddedilir (Anlamlı korelasyon var)")
else:
    print(f"Spearman: p >= {alpha} → H0 reddedilemez (Anlamlı korelasyon yok)")
````

````python
# 7. Korelasyon Matrisi ve Heatmap
corr_matrix = df.corr(method='pearson')
print("\nKorelasyon Matrisi (Pearson):")
print(corr_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Korelasyon Matrisi (Pearson)')
plt.show()

# 8. Ekstra: Regresyon Doğrusu ile Scatter Plot
plt.figure(figsize=(8, 6))
sns.regplot(data=df, x='Calisma_Saatleri', y='Sinav_Notlari', 
            scatter_kws={'color': 'blue'}, 
            line_kws={'color': 'red', 'linewidth': 2})
plt.title('Çalışma Saatleri ve Sınav Notları Arasındaki İlişki\n(Regresyon Doğrusu ile)')
plt.xlabel('Çalışma Saatleri')
plt.ylabel('Sınav Notları')
plt.grid(True)
plt.show()
````

### Elimizdeki Katsayının testini yapalım. Korelasyonun Anlamlılığının Testi


Kolerasyon anlamlılığı testi, iki değişken arasındaki ilişkinin istatistiksel olarak anlamlı olup olmadığını değerlendirmek için kullanılan bir istatistiksel testtir. İki değişken arasındaki ilişkinin gücü ve yönü hakkında bilgi sağlar. Kolerasyon, genellikle Pearson korelasyon katsayısı ile ölçülür, ancak diğer korelasyon katsayıları da kullanılabilir.

Pearson korelasyon katsayısı, iki değişken arasındaki lineer ilişkiyi ölçen bir istatistiksel değerdir. Değer, -1 ile 1 arasında bir aralıkta olabilir. -1, tam ters orantılı bir ilişkiyi, 1 ise tam orantılı bir ilişkiyi gösterir. 0 ise iki değişken arasında lineer bir ilişki olmadığını gösterir.

Kolerasyon anlamlılığı testi genellikle şu hipotezleri test etmek için kullanılır:

- **H0 (Null Hypothesis):** İki değişken arasında bir ilişki yoktur (Pearson korelasyon katsayısı 0'dır).
- **H1 (Alternative Hypothesis):** İki değişken arasında bir ilişki vardır (Pearson korelasyon katsayısı 0 değildir).

Bu test, genellikle örneklemler arasındaki kolerasyonun anlamlı olup olmadığını belirlemek için kullanılır. Test sonuçları, belirli bir güven düzeyinde (genellikle %95 veya %99) istatistiksel olarak anlamlı olup olmadığını belirtir.

Kolerasyon anlamlılığı testleri genellikle istatistiksel yazılımlar veya istatistiksel paketler kullanılarak gerçekleştirilir. Python, R ve SPSS gibi araçlar bu tür analizler için sıklıkla kullanılan araçlardır.

````python
from scipy.stats import pearsonr #pearson korelasyon katsayısını hesaplamak için gerekli modülü import edelim
````

````python
test_istatistigi, pvalue = pearsonr(df["tip"], df["total_bill"]) #pearson korelasyon katsayısını hesaplıyoruz, bu fonksiyonun içerisine x ve y değişkenlerini giriyoruz. sol taraftaki değer test istatistiğini, sağ taraftaki değer ise p-value değerini verir

print('Korelasyon Katsayısı: %.10f, P-değeri: %.20f' % (test_istatistigi, pvalue))
````

Değişkenler arasında anlamlı bir ilişki yoktur diyen H0 hipotezini reddediyoruz. Yani değişkenler arasında anlamlı bir ilişki, korelasyon vardır diyoruz.

## Nonparametrik Korelasyon Hipotez Testi



Nonparametrik korelasyon testleri, değişkenler arasındaki ilişkiyi değerlendirmek için kullanılan istatistiksel yöntemlerdir. Bu testler, verilerin belirli bir dağılıma uymadığı durumlar veya sıralı verilerle çalışıldığında tercih edilebilir. Nonparametrik korelasyon testleri, parametrik korelasyon testlerine alternatif olarak kullanılır ve bazen veriler normal dağılıma uymadığında veya sıralı verilerle çalışıldığında daha güvenilir sonuçlar verir.

İşte bazı yaygın nonparametrik korelasyon testleri:

1. **Spearman's Rho (Spearman Rank Korelasyon Katsayısı):** Spearman korelasyon katsayısı, değişkenler arasındaki monotonic bir ilişkiyi ölçer. Monotonik bir ilişki, değişkenlerin birbirine doğrusal olmayan bir şekilde sıralanabilir olduğu anlamına gelir.

2. **Kendall's Tau (Kendall Tau Rank Korelasyon Katsayısı):** Kendall korelasyon katsayısı da sıralı veriler arasındaki ilişkiyi değerlendiren bir nonparametrik korelasyon testidir. Spearman gibi, Kendall da monotonic bir ilişkiyi ölçer.

Bu nonparametrik korelasyon testleri, özellikle değişkenler arasındaki ilişkinin doğrusal olmadığı durumlarda veya verilerin normal dağılıma uymadığı durumlarda tercih edilir. Parametrik testlerin varsayımlarına uymayan durumlarda daha güçlü ve güvenilir sonuçlar sağlayabilirler. Bu testler genellikle sıralı verilerle çalışırken, parametrik testler genellikle sürekli veriler üzerinde çalışır.

````python
#varsayalım sağlanmadığından yapabileceğimiz iki tane nonparametrik yöntem var: spearman ve kendall
import pandas as pd
from scipy.stats import spearmanr # Import spearmanr directly

spearmanr(df["tip"], df["total_bill"]) #spearman korelasyon katsayısını hesaplıyoruz, bu fonksiyonun içerisine x ve y değişkenlerini giriyoruz
````

````python
# Spearman korelasyon katsayısını hesaplama
# stats.spearmanr yerine doğrudan spearmanr kullanıyoruz
rho, p_value = spearmanr(df["tip"], df["total_bill"])

print(f"Spearman Korelasyon Katsayısı (rho): {rho:.4f}")
print(f"p-değeri: {p_value:.4f}")

# Yorumlama
alpha = 0.05
if p_value < alpha:
    print(f"\nSonuç: p-değeri ({p_value:.4f}) < alpha ({alpha}) olduğundan,")
    print("Bahşiş miktarı ile toplam hesap arasında istatistiksel olarak **anlamlı bir Spearman korelasyonu** vardır.")
else:
    print(f"\nSonuç: p-değeri ({p_value:.4f}) >= alpha ({alpha}) olduğundan,")
    print("Bahşiş miktarı ile toplam hesap arasında istatistiksel olarak **anlamlı bir Spearman korelasyonu yoktur**.")
````

````python
from scipy.stats import spearmanr,kendalltau
test_istatistigi,pvalue=spearmanr(df['tip'],df['total_bill'])  
print('test istatistiği/Korelasyon Katsayısı =%.4f , p-value =%.24f' % (test_istatistigi,pvalue)) 

test_istatistigi,pvalue=kendalltau(df['tip'],df['total_bill'])
print('test istatistiği/Korelasyon Katsayısı =%.4f , p-value =%.24f' % (test_istatistigi,pvalue)) 
````

Değişkenlerin arasında orta şiddette pozitif  bir ilişki vardır ve anlamlıdır.  p < 0.05 ---> H0 reddedilir.

Kendall testinde korelasyon daha temkinli, biraz daha düşük çıkmıştır.

Bizim için varsayımlar sağlanmadığından dolayı en güvenilir test spearman korelasyon testidir.
