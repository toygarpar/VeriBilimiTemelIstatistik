## <span style ="color:red">**Kolerasyon Analizi**</span>

Korelasyon analizi, iki deÄŸiÅŸken arasÄ±ndaki doÄŸrusal iliÅŸkinin yÃ¶nÃ¼nÃ¼ ve gÃ¼cÃ¼nÃ¼ Ã¶lÃ§mek iÃ§in kullanÄ±lan bir istatistiksel yÃ¶ntemdir. - Korelasyon analizinde genellikle test edilen temel hipotez, iki deÄŸiÅŸken arasÄ±nda bir iliÅŸki olup olmadÄ±ÄŸÄ±dÄ±r. Korelasyon analizi genellikle Pearson Korelasyon KatsayÄ±sÄ± veya Spearman SÄ±ra Korelasyon KatsayÄ±sÄ± kullanÄ±larak yapÄ±lÄ±r. Ä°ÅŸte korelasyon analizi hipotezleri:

**Korelasyonun Temel KavramlarÄ±:**

- **DeÄŸiÅŸken:** Ã–lÃ§Ã¼lebilen ve farklÄ± deÄŸerler alabilen Ã¶zelliktir. Ã–rneÄŸin, boy, kilo, sÄ±caklÄ±k, gelir birer deÄŸiÅŸkendir.
- **DoÄŸrusal Ä°liÅŸki:** Ä°ki deÄŸiÅŸken arasÄ±ndaki iliÅŸkinin bir doÄŸru ile temsil edilebilmesidir.
- **Korelasyon KatsayÄ±sÄ± (r):** Ä°ki deÄŸiÅŸken arasÄ±ndaki doÄŸrusal iliÅŸkinin gÃ¼cÃ¼nÃ¼ ve yÃ¶nÃ¼nÃ¼ gÃ¶steren bir sayÄ±dÄ±r. -1 ile +1 arasÄ±nda deÄŸer alÄ±r.


**Korelasyon KatsayÄ±sÄ±nÄ±n YorumlanmasÄ±:**

- **r = +1:** MÃ¼kemmel pozitif korelasyon. Bir deÄŸiÅŸken arttÄ±kÃ§a, diÄŸer deÄŸiÅŸken de aynÄ± oranda artar.
- **0 < r < +1:** Pozitif korelasyon. Bir deÄŸiÅŸken arttÄ±kÃ§a, diÄŸer deÄŸiÅŸken de artma eÄŸilimindedir.
- **r = 0:** Korelasyon yok. Ä°ki deÄŸiÅŸken arasÄ±nda doÄŸrusal bir iliÅŸki yoktur.
- **-1 < r < 0:** Negatif korelasyon. Bir deÄŸiÅŸken arttÄ±kÃ§a, diÄŸer deÄŸiÅŸken azalma eÄŸilimindedir.
- **r = -1:** MÃ¼kemmel negatif korelasyon. Bir deÄŸiÅŸken arttÄ±kÃ§a, diÄŸer deÄŸiÅŸken aynÄ± oranda azalÄ±r.


**Korelasyon TÃ¼rleri:**

- **Pearson Korelasyon KatsayÄ±sÄ± (r):** En yaygÄ±n kullanÄ±lan korelasyon tÃ¼rÃ¼dÃ¼r. Ä°ki sÃ¼rekli deÄŸiÅŸken arasÄ±ndaki doÄŸrusal iliÅŸkiyi Ã¶lÃ§er. Verilerin normal daÄŸÄ±lÄ±ma sahip olmasÄ± varsayÄ±lÄ±r.
- **Spearman SÄ±ra Korelasyon KatsayÄ±sÄ± (Ï veya rs):** SÃ¼rekli veya sÄ±ralÄ± veriler iÃ§in kullanÄ±lÄ±r. Verilerin normal daÄŸÄ±lÄ±m gÃ¶stermesi gerekmez. Ä°ki deÄŸiÅŸken arasÄ±ndaki sÄ±ralamalarÄ±n benzerliÄŸini Ã¶lÃ§er.
- **Kendall Tau Korelasyon KatsayÄ±sÄ± (Ï„):** Spearman korelasyonuna benzerdir, ancak farklÄ± bir hesaplama yÃ¶ntemi kullanÄ±r. Ã–zellikle kÃ¼Ã§Ã¼k Ã¶rneklemlerde ve sÄ±ralÄ± verilerde daha uygundur.

**Korelasyon Analizi NasÄ±l YapÄ±lÄ±r?**

Korelasyon analizi genellikle istatistiksel yazÄ±lÄ±mlar (SPSS, R, Excel vb.) kullanÄ±larak yapÄ±lÄ±r. Temel adÄ±mlar ÅŸunlardÄ±r:

1. **Verileri HazÄ±rlama:** Ä°ki deÄŸiÅŸken iÃ§in veri setini oluÅŸturun.
2. **Korelasyon KatsayÄ±sÄ±nÄ± Hesaplama:** Uygun korelasyon katsayÄ±sÄ±nÄ± (Pearson, Spearman veya Kendall) seÃ§erek hesaplayÄ±n.
3. **AnlamlÄ±lÄ±k Testi Yapma:** Korelasyon katsayÄ±sÄ±nÄ±n istatistiksel olarak anlamlÄ± olup olmadÄ±ÄŸÄ±nÄ± test edin. Bu genellikle bir hipotez testi (t-testi) ile yapÄ±lÄ±r.


Korelasyon analizi hipotezleri, iki deÄŸiÅŸken arasÄ±nda doÄŸrusal bir iliÅŸki olup olmadÄ±ÄŸÄ±nÄ± test etmek iÃ§in kullanÄ±lÄ±r. Temel olarak iki tÃ¼r hipotez vardÄ±r: sÄ±fÄ±r hipotezi ve alternatif hipotez.

**1. SÄ±fÄ±r Hipotezi (Hâ‚€):**

SÄ±fÄ±r hipotezi, iki deÄŸiÅŸken arasÄ±nda _doÄŸrusal bir iliÅŸki olmadÄ±ÄŸÄ±nÄ±_ varsayar. BaÅŸka bir deyiÅŸle, korelasyon katsayÄ±sÄ±nÄ±n (r) sÄ±fÄ±ra eÅŸit olduÄŸunu iddia eder.

- **Hâ‚€: Ï = 0** (PopÃ¼lasyon korelasyon katsayÄ±sÄ± sÄ±fÄ±rdÄ±r)
- **Hâ‚€: r = 0** (Ã–rneklem korelasyon katsayÄ±sÄ± sÄ±fÄ±rdÄ±r)

Burada Ï (rho) popÃ¼lasyon korelasyon katsayÄ±sÄ±nÄ±, r ise Ã¶rneklem korelasyon katsayÄ±sÄ±nÄ± temsil eder.

**2. Alternatif Hipotez (Hâ‚ veya Hâ‚):**

Alternatif hipotez, iki deÄŸiÅŸken arasÄ±nda _doÄŸrusal bir iliÅŸki olduÄŸunu_ varsayar. ÃœÃ§ farklÄ± ÅŸekilde formÃ¼le edilebilir:

- **Ã‡ift YÃ¶nlÃ¼ Alternatif Hipotez:** Ä°ki deÄŸiÅŸken arasÄ±nda bir iliÅŸki _olduÄŸunu_ ancak bu iliÅŸkinin yÃ¶nÃ¼nÃ¼n (pozitif veya negatif) belirtilmediÄŸini varsayar.
    
    - **Hâ‚: Ï â‰  0** (PopÃ¼lasyon korelasyon katsayÄ±sÄ± sÄ±fÄ±ra eÅŸit deÄŸildir)
    - **Hâ‚: r â‰  0** (Ã–rneklem korelasyon katsayÄ±sÄ± sÄ±fÄ±ra eÅŸit deÄŸildir)
- **Tek YÃ¶nlÃ¼ Alternatif Hipotez (Pozitif Korelasyon):** Ä°ki deÄŸiÅŸken arasÄ±nda _pozitif bir iliÅŸki olduÄŸunu_ varsayar. Bir deÄŸiÅŸken arttÄ±kÃ§a diÄŸerinin de artacaÄŸÄ±nÄ± iddia eder.
    
    - **Hâ‚: Ï > 0** (PopÃ¼lasyon korelasyon katsayÄ±sÄ± sÄ±fÄ±rdan bÃ¼yÃ¼ktÃ¼r)
    - **Hâ‚: r > 0** (Ã–rneklem korelasyon katsayÄ±sÄ± sÄ±fÄ±rdan bÃ¼yÃ¼ktÃ¼r)
- **Tek YÃ¶nlÃ¼ Alternatif Hipotez (Negatif Korelasyon):** Ä°ki deÄŸiÅŸken arasÄ±nda _negatif bir iliÅŸki olduÄŸunu_ varsayar. Bir deÄŸiÅŸken arttÄ±kÃ§a diÄŸerinin azalacaÄŸÄ±nÄ± iddia eder.
    
    - **Hâ‚: Ï < 0** (PopÃ¼lasyon korelasyon katsayÄ±sÄ± sÄ±fÄ±rdan kÃ¼Ã§Ã¼ktÃ¼r)
    - **Hâ‚: r < 0** (Ã–rneklem korelasyon katsayÄ±sÄ± sÄ±fÄ±rdan kÃ¼Ã§Ã¼ktÃ¼r)



**Korelasyon Analizi Hipotez Testi NasÄ±l YapÄ±lÄ±r?**

Korelasyon analizinde hipotez testi genellikle bir t-testi kullanÄ±larak yapÄ±lÄ±r. Test istatistiÄŸi aÅŸaÄŸÄ±daki formÃ¼lle hesaplanÄ±r:

```
t = r / âˆš[(1 - rÂ²) / (n - 2)]
```

Burada:

- r: Ã–rneklem korelasyon katsayÄ±sÄ±
- n: Ã–rneklem bÃ¼yÃ¼klÃ¼ÄŸÃ¼

Hesaplanan t deÄŸeri, belirli bir serbestlik derecesi (n-2) ve anlamlÄ±lÄ±k dÃ¼zeyi (Î±) iÃ§in t daÄŸÄ±lÄ±m tablosundaki kritik deÄŸerle karÅŸÄ±laÅŸtÄ±rÄ±lÄ±r.

- **Ã‡ift yÃ¶nlÃ¼ testte:** Hesaplanan |t| deÄŸeri, kritik deÄŸerden bÃ¼yÃ¼kse, sÄ±fÄ±r hipotezi reddedilir ve iki deÄŸiÅŸken arasÄ±nda istatistiksel olarak anlamlÄ± bir iliÅŸki olduÄŸu sonucuna varÄ±lÄ±r.
- **Tek yÃ¶nlÃ¼ testte:** Hesaplanan t deÄŸeri, uygun yÃ¶ndeki kritik deÄŸerden bÃ¼yÃ¼kse, sÄ±fÄ±r hipotezi reddedilir ve iki deÄŸiÅŸken arasÄ±nda belirtilen yÃ¶nde istatistiksel olarak anlamlÄ± bir iliÅŸki olduÄŸu sonucuna varÄ±lÄ±r.

AyrÄ±ca, p-deÄŸeri de kullanÄ±labilir. p-deÄŸeri anlamlÄ±lÄ±k dÃ¼zeyinden (Î±) kÃ¼Ã§Ã¼kse, sÄ±fÄ±r hipotezi reddedilir.


## ğŸ“ Korelasyon TÃ¼rleri

### 1. **Pearson Korelasyon KatsayÄ±sÄ±**
SÃ¼rekli ve normal daÄŸÄ±lan veriler iÃ§in uygun.

````python
import numpy as np
from scipy.stats import pearsonr

# SimÃ¼lasyon: reklam harcamasÄ± ve satÄ±ÅŸ adedi
np.random.seed(42)
reklam = np.random.normal(loc=50, scale=10, size=100)
satis = reklam * 1.3 + np.random.normal(0, 5, size=100)  # Pozitif iliÅŸki

r, p_degeri = pearsonr(reklam, satis)
print(f"Pearson r: {r:.2f}, p-deÄŸeri: {p_degeri:.4f}")
````

ğŸ“Œ Yorum:
- GÃ¼Ã§lÃ¼ pozitif iliÅŸki
- p-deÄŸeri < 0.05 â†’ anlamlÄ±

### 2. **Spearman SÄ±ra Korelasyonu**
Veriler sÄ±ralÄ± ya da normal daÄŸÄ±lmÄ±yorsa kullanÄ±lÄ±r.

````python
from scipy.stats import spearmanr

rho, p_val_rho = spearmanr(reklam, satis)
print(f"Spearman Ï: {rho:.2f}, p-deÄŸeri: {p_val_rho:.4f}")
````

ğŸ“Œ Spearman, ranklar Ã¼zerinden hesaplandÄ±ÄŸÄ± iÃ§in aykÄ±rÄ± deÄŸerlere daha dayanÄ±klÄ±dÄ±r.

### 3. **Kendallâ€™s Tau Korelasyonu**
Ã–zellikle kÃ¼Ã§Ã¼k Ã¶rneklemlerde tercih edilir.

````python
from scipy.stats import kendalltau

tau, p_val_tau = kendalltau(reklam, satis)
print(f"Kendall Ï„: {tau:.2f}, p-deÄŸeri: {p_val_tau:.4f}")
````

ğŸ“Œ Kendallâ€™s Tau, Pearsonâ€™a gÃ¶re daha az duyarlÄ±dÄ±r ama daha gÃ¼venilirdir.

### AdÄ±m 5: Korelasyon Testi Yap

````python
from scipy.stats import ttest_1samp

# Pearson r'den t-testi yap
n = len(reklam)
t_skoru = r / np.sqrt((1 - r**2) / (n - 2))
print(f"T Skoru: {t_skoru:.2f}")

# Serbestlik derecesi ve test
_, p_degeri = ttest_1samp(t_skoru, popmean=0)
print(f"p-deÄŸeri: {p_degeri:.4f}")






#  T-skorunu hesapla (bu formÃ¼l doÄŸru)
# RuntimeWarning'dan kaÃ§Ä±nmak iÃ§in (1 - r**2) nin 0'a yakÄ±n olmamasÄ±nÄ± kontrol edin,
# ki bu da korelasyonun 1 veya -1 olmasÄ± durumudur.
# EÄŸer r tam olarak 1 veya -1 ise, (1-r**2) sÄ±fÄ±r olur ve t_skoru tanÄ±msÄ±z olur.
# GerÃ§ek verilerde bu nadirdir.
if np.isclose(r**2, 1.0) and n == 2: # n=2 ise r=1 veya -1 olacaktÄ±r
    t_skoru = np.inf * np.sign(r) # r'nin iÅŸaretine gÃ¶re inf veya -inf
elif np.isclose(r**2, 1.0): # n > 2 ise r'nin tam 1 veya -1 olmasÄ± imkansÄ±zdÄ±r
    # Ã‡ok nadir durumda r^2 tam olarak 1 ise, buraya dÃ¼ÅŸmez
    # ama yine de t_skoru formÃ¼lÃ¼ tanÄ±msÄ±z hale gelir.
    t_skoru = np.inf * np.sign(r) # Veya duruma gÃ¶re hata fÄ±rlatÄ±labilir
else:
    t_skoru = r * np.sqrt((n - 2) / (1 - r**2))

print(f"Hesaplanan T Skoru: {t_skoru:.4f}")
````

### AdÄ±m 6: Karar Ver

````python
# if p_degeri < alpha:
#     print("âœ… Hâ‚€ reddedildi â†’ Korelasyon anlamlÄ±")
# else:
#     print("âŒ Hâ‚€ reddedilemedi â†’ Korelasyon tesadÃ¼fi olabilir")


# --- SonuÃ§ Yorumlama ---
if not np.isnan(p_degeri):
    if p_degeri < alpha:
        print(f"\nSonuÃ§: p-deÄŸeri ({p_degeri:.4f}) < alpha ({alpha}) olduÄŸundan,")
        print("Pearson korelasyon katsayÄ±sÄ± istatistiksel olarak **anlamlÄ±dÄ±r**.")
        print("Yani, reklam ve satÄ±ÅŸlar arasÄ±nda anlamlÄ± bir doÄŸrusal iliÅŸki vardÄ±r.")
    else:
        print(f"\nSonuÃ§: p-deÄŸeri ({p_degeri:.4f}) >= alpha ({alpha}) olduÄŸundan,")
        print("Pearson korelasyon katsayÄ±sÄ± istatistiksel olarak **anlamlÄ± deÄŸildir**.")
        print("Yani, reklam ve satÄ±ÅŸlar arasÄ±nda anlamlÄ± bir doÄŸrusal iliÅŸki olduÄŸuna dair yeterli kanÄ±t yoktur.")
else:
    print("\nSonuÃ§: p-deÄŸeri hesaplanamadÄ±ÄŸÄ± iÃ§in anlamlÄ±lÄ±k yorumu yapÄ±lamÄ±yor.")
````

### Ã–rnek:

````python
import numpy as np
from scipy import stats

# --- SimÃ¼lasyon Verisi OluÅŸturma (Ã–rnek AmaÃ§lÄ±) ---
# GerÃ§ek senaryoda bu verileriniz olurdu
np.random.seed(42)
reklam_harcamasi = np.random.normal(loc=100, scale=20, size=50) # Ã–rn. reklam harcamasÄ±
satislar = 0.8 * reklam_harcamasi + np.random.normal(loc=0, scale=10, size=50) # Reklamla iliÅŸkili satÄ±ÅŸlar

# --- Pearson Korelasyon KatsayÄ±sÄ±nÄ± Hesaplama ---
r = np.corrcoef(reklam_harcamasi, satislar)[0, 1]
print(f"Hesaplanan Pearson Korelasyon KatsayÄ±sÄ± (r): {r:.4f}")
````

````python
# --- Pearson r'nin AnlamlÄ±lÄ±ÄŸÄ±nÄ± Test Etme ---
# AdÄ±m 1: T-skorunu hesapla
n = len(reklam_harcamasi) # Ã–rneklem bÃ¼yÃ¼klÃ¼ÄŸÃ¼
# Korelasyon katsayÄ±sÄ± iÃ§in t-skoru formÃ¼lÃ¼
t_skoru = r * np.sqrt((n - 2) / (1 - r**2))
print(f"Hesaplanan T Skoru: {t_skoru:.4f}")
````

````python
# AdÄ±m 2: Serbestlik derecesini belirle
df = n - 2 # Pearson korelasyonu iÃ§in serbestlik derecesi
````

````python
# AdÄ±m 3: T-daÄŸÄ±lÄ±mÄ±nÄ± kullanarak p-deÄŸerini bul
# stats.t.sf(abs(t_skoru), df) tek kuyruk p-deÄŸerini verir.
# Ã‡ift kuyruk (iki taraflÄ±) test iÃ§in 2 ile Ã§arpÄ±lÄ±r.
p_degeri = 2 * stats.t.sf(abs(t_skoru), df)
print(f"p-deÄŸeri (iki taraflÄ±): {p_degeri:.4f}")
````

````python
# --- SonuÃ§ Yorumlama ---
alpha = 0.05 # AnlamlÄ±lÄ±k dÃ¼zeyi

if p_degeri < alpha:
    print(f"\nSonuÃ§: p-deÄŸeri ({p_degeri:.4f}) < alpha ({alpha}) olduÄŸundan,")
    print("Pearson korelasyon katsayÄ±sÄ± istatistiksel olarak **anlamlÄ±dÄ±r**.")
    print("Yani, reklam harcamasÄ± ile satÄ±ÅŸlar arasÄ±nda anlamlÄ± bir doÄŸrusal iliÅŸki vardÄ±r.")
else:
    print(f"\nSonuÃ§: p-deÄŸeri ({p_degeri:.4f}) >= alpha ({alpha}) olduÄŸundan,")
    print("Pearson korelasyon katsayÄ±sÄ± istatistiksel olarak **anlamlÄ± deÄŸildir**.")
    print("Yani, reklam harcamasÄ± ile satÄ±ÅŸlar arasÄ±nda anlamlÄ± bir doÄŸrusal iliÅŸki olduÄŸuna dair yeterli kanÄ±t yoktur.")

# Not: Pingouin kÃ¼tÃ¼phanesi bu iÅŸi Ã§ok daha basit yapar:
import pingouin as pg
corr_test = pg.corr(reklam_harcamasi, satislar)
print("\nPingouin ile Korelasyon Testi Sonucu:")
print(corr_test)
````

### <span style ="color:yellow">**Pearson Korelasyon KatsayÄ±sÄ± Ä°Ã§in Hipotezler**</span>

  

1. **Null Hypothesis (H0):** Ä°ki deÄŸiÅŸken arasÄ±nda korelasyon yoktur (r = 0)

  

- $ H_0: \rho = 0 $


Pearson Korelasyon KatsayÄ±sÄ± NasÄ±l HesaplanÄ±r?

Pearson korelasyon katsayÄ±sÄ±nÄ±n hesaplanmasÄ± iÃ§in birkaÃ§ formÃ¼l bulunmaktadÄ±r. En yaygÄ±n kullanÄ±lan formÃ¼llerden biri ÅŸÃ¶yledir:

r = Î£[(xáµ¢ - xÌ„)(yáµ¢ - È³)] / âˆš[Î£(xáµ¢ - xÌ„)Â² * Î£(yáµ¢ - È³)Â²]
Burada:

xáµ¢ ve yáµ¢: Her bir veri noktasÄ±ndaki deÄŸiÅŸkenlerin deÄŸerleri
xÌ„ ve È³: x ve y deÄŸiÅŸkenlerinin ortalamalarÄ±
Î£: Toplam sembolÃ¼
Bu formÃ¼l, her bir veri noktasÄ±nÄ±n ortalamadan sapmalarÄ±nÄ±n Ã§arpÄ±mÄ±nÄ±n, bu sapmalarÄ±n karelerinin Ã§arpÄ±mÄ±nÄ±n karekÃ¶kÃ¼ne bÃ¶lÃ¼nmesiyle elde edilir.

Pearson Korelasyon KatsayÄ±sÄ±nÄ±n YorumlanmasÄ±:

Korelasyon katsayÄ±sÄ±nÄ±n bÃ¼yÃ¼klÃ¼ÄŸÃ¼, iliÅŸkinin gÃ¼cÃ¼nÃ¼ gÃ¶sterir:

0.70 veya daha yÃ¼ksek: GÃ¼Ã§lÃ¼ korelasyon
0.50 ile 0.69 arasÄ±: Orta dÃ¼zeyde korelasyon
0.30 ile 0.49 arasÄ±: ZayÄ±f korelasyon
0.00 ile 0.29 arasÄ±: Ã‡ok zayÄ±f veya korelasyon yok
Ancak, bu aralÄ±klar genel bir kÄ±lavuzdur ve yorumlama, incelenen alana ve baÄŸlama gÃ¶re deÄŸiÅŸebilir.


  

2. **Alternative Hypothesis (H1):** Ä°ki deÄŸiÅŸken arasÄ±nda korelasyon vardÄ±r (r â‰  0).

- $ H_1: \rho \neq 0 $

  

Bu durumda, \(\rho\) Pearson korelasyon katsayÄ±sÄ±dÄ±r.


EÄŸer varsayÄ±m saÄŸlanmÄ±yorsa, Non-Parametrik Korelasyon ANalizi Hipotez Testi iÃ§in  Spearman Korelasyon KatsayÄ±sÄ±nÄ± kullanÄ±rÄ±z.



### <span style ="color:yellow">**Spearman SÄ±ra Korelasyon KatsayÄ±sÄ± Ä°Ã§in Hipotezler**</span>

  

1. **Null Hypothesis (H0):** Ä°ki deÄŸiÅŸken arasÄ±nda sÄ±ra korelasyonu yoktur (Ï = 0).

- $ H_0: \rho = 0 $

  

2. **Alternative Hypothesis (H1):** Ä°ki deÄŸiÅŸken arasÄ±nda sÄ±ra korelasyonu vardÄ±r (Ï â‰  0).

- $ H_1: \rho \neq 0 $

  

Bu durumda, $\rho$ Spearman sÄ±ra korelasyon katsayÄ±sÄ±dÄ±r.

Bu hipotezler, korelasyonun varlÄ±ÄŸÄ± veya yokluÄŸunu test etmek amacÄ±yla kullanÄ±lÄ±r. EÄŸer null hypothesis (H0) reddedilirse, bu, iki deÄŸiÅŸken arasÄ±nda anlamlÄ± bir korelasyon olduÄŸu anlamÄ±na gelir. Bu hipotez testleri genellikle belirli bir anlamlÄ±lÄ±k dÃ¼zeyi (alfa, genellikle 0.05) kullanÄ±larak yapÄ±lÄ±r.

Korelasyon analizi, iki deÄŸiÅŸken arasÄ±ndaki iliÅŸkiyi deÄŸerlendiren bir istatistiksel yÃ¶ntemdir. Korelasyon varsayÄ±mlarÄ±, korelasyon katsayÄ±larÄ±nÄ±n gÃ¼venilir bir ÅŸekilde yorumlanabilmesi ve kullanÄ±labilmesi iÃ§in dikkate alÄ±nmasÄ± gereken temel Ã¶nermelerdir.

- Korelasyon katsayÄ±larÄ± genellikle deÄŸiÅŸkenlerin normal daÄŸÄ±lÄ±ma sahip olduÄŸunda daha gÃ¼venilir olarak kabul edilir.

- Bu varsayÄ±m, korelasyon analizi sonuÃ§larÄ± Ã¼zerindeki gÃ¼venilirliÄŸi artÄ±rmak iÃ§in Ã¶nemlidir.

- AykÄ±rÄ± gÃ¶zlemler ve uÃ§ deÄŸerler, korelasyon analizini etkileyebilir. Bu deÄŸerlerin analizden Ã¶nce incelenmesi ve gerekirse dÃ¼zeltilmesi Ã¶nemlidir.

- Homoscedasticity, deÄŸiÅŸkenler arasÄ±ndaki deÄŸiÅŸkenliÄŸin eÅŸit olduÄŸu anlamÄ±na gelir. Korelasyon analizi iÃ§in bu varsayÄ±m Ã¶nemlidir.

Bu varsayÄ±mlar, korelasyon analizi sonuÃ§larÄ±nÄ±n gÃ¼venilirliÄŸini artÄ±rmaya yÃ¶neliktir. **Korelasyon ve Nedensellik:**

Korelasyon, nedensellik anlamÄ±na gelmez. Ä°ki deÄŸiÅŸken arasÄ±nda bir korelasyon olmasÄ±, birinin diÄŸerine neden olduÄŸu anlamÄ±na gelmez. Ã–rneÄŸin, dondurma satÄ±ÅŸlarÄ± ile boÄŸulma vakalarÄ± arasÄ±nda pozitif bir korelasyon olabilir, ancak dondurma yemenin boÄŸulmaya neden olduÄŸu anlamÄ±na gelmez. Bu durumda, her iki deÄŸiÅŸkeni de etkileyen Ã¼Ã§Ã¼ncÃ¼ bir faktÃ¶r (Ã¶rneÄŸin, sÄ±cak hava) olabilir.


````python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Veri OluÅŸturma (Ã–rnek veri - kendi verilerinizle deÄŸiÅŸtirebilirsiniz)
np.random.seed(42)
n = 50  # Ã–rneklem bÃ¼yÃ¼klÃ¼ÄŸÃ¼
calisma_saatleri = np.random.normal(15, 5, n)  # Ortalama 15 saat, std=5
sinav_notlari = 50 + 2*calisma_saatleri + np.random.normal(0, 10, n)  # Pozitif iliÅŸki

# Veriyi DataFrame'e dÃ¶nÃ¼ÅŸtÃ¼rme
df = pd.DataFrame({
    'Calisma_Saatleri': calisma_saatleri,
    'Sinav_Notlari': sinav_notlari
})
````

````python
# 2. Veri Ã–n Ä°nceleme
print("\nVerinin Ä°lk 5 SatÄ±rÄ±:")
print(df.head())

print("\nTemel Ä°statistikler:")
print(df.describe())
````

````python
# 3. Normallik Testi (Shapiro-Wilk)
print("\nNormallik Testi SonuÃ§larÄ±:")
for col in df.columns:
    stat, p = stats.shapiro(df[col])
    print(f"{col}: W={stat:.3f}, p={p:.4f} - {'Normal' if p>0.05 else 'Normal deÄŸil'}")
````

````python
# 4. GÃ¶rselleÅŸtirme
plt.figure(figsize=(12, 5))

# Histogramlar
plt.subplot(1, 2, 1)
sns.histplot(df['Calisma_Saatleri'], kde=True, color='blue', label='Ã‡alÄ±ÅŸma Saatleri')
sns.histplot(df['Sinav_Notlari'], kde=True, color='red', label='SÄ±nav NotlarÄ±')
plt.title('DeÄŸiÅŸkenlerin DaÄŸÄ±lÄ±mÄ±')
plt.legend()

# Scatter plot
plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='Calisma_Saatleri', y='Sinav_Notlari', color='green')
plt.title('Ã‡alÄ±ÅŸma Saatleri vs SÄ±nav NotlarÄ±')
plt.tight_layout()
plt.show()
````

````python
# 5. Korelasyon Analizi
# Pearson Korelasyonu (Parametrik)
pearson_r, pearson_p = stats.pearsonr(df['Calisma_Saatleri'], df['Sinav_Notlari'])

# Spearman Korelasyonu (Non-parametrik)
spearman_r, spearman_p = stats.spearmanr(df['Calisma_Saatleri'], df['Sinav_Notlari'])
````

````python
# 6. SonuÃ§larÄ±n YazdÄ±rÄ±lmasÄ±
alpha = 0.05

print("\nKORELASYON ANALÄ°ZÄ° SONUÃ‡LARI:")
print(f"\nPearson Korelasyon KatsayÄ±sÄ± (r): {pearson_r:.3f}")
print(f"Pearson p-deÄŸeri: {pearson_p:.4f}")
if pearson_p < alpha:
    print(f"Pearson: p < {alpha} â†’ H0 reddedilir (AnlamlÄ± korelasyon var)")
else:
    print(f"Pearson: p >= {alpha} â†’ H0 reddedilemez (AnlamlÄ± korelasyon yok)")

print(f"\nSpearman Korelasyon KatsayÄ±sÄ± (Ï): {spearman_r:.3f}")
print(f"Spearman p-deÄŸeri: {spearman_p:.4f}")
if spearman_p < alpha:
    print(f"Spearman: p < {alpha} â†’ H0 reddedilir (AnlamlÄ± korelasyon var)")
else:
    print(f"Spearman: p >= {alpha} â†’ H0 reddedilemez (AnlamlÄ± korelasyon yok)")
````

````python
# 7. Korelasyon Matrisi ve Heatmap
corr_matrix = df.corr(method='pearson')
print("\nKorelasyon Matrisi (Pearson):")
print(corr_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Korelasyon Matrisi (Pearson)')
plt.show()

# 8. Ekstra: Regresyon DoÄŸrusu ile Scatter Plot
plt.figure(figsize=(8, 6))
sns.regplot(data=df, x='Calisma_Saatleri', y='Sinav_Notlari', 
            scatter_kws={'color': 'blue'}, 
            line_kws={'color': 'red', 'linewidth': 2})
plt.title('Ã‡alÄ±ÅŸma Saatleri ve SÄ±nav NotlarÄ± ArasÄ±ndaki Ä°liÅŸki\n(Regresyon DoÄŸrusu ile)')
plt.xlabel('Ã‡alÄ±ÅŸma Saatleri')
plt.ylabel('SÄ±nav NotlarÄ±')
plt.grid(True)
plt.show()
````

### Elimizdeki KatsayÄ±nÄ±n testini yapalÄ±m. Korelasyonun AnlamlÄ±lÄ±ÄŸÄ±nÄ±n Testi


Kolerasyon anlamlÄ±lÄ±ÄŸÄ± testi, iki deÄŸiÅŸken arasÄ±ndaki iliÅŸkinin istatistiksel olarak anlamlÄ± olup olmadÄ±ÄŸÄ±nÄ± deÄŸerlendirmek iÃ§in kullanÄ±lan bir istatistiksel testtir. Ä°ki deÄŸiÅŸken arasÄ±ndaki iliÅŸkinin gÃ¼cÃ¼ ve yÃ¶nÃ¼ hakkÄ±nda bilgi saÄŸlar. Kolerasyon, genellikle Pearson korelasyon katsayÄ±sÄ± ile Ã¶lÃ§Ã¼lÃ¼r, ancak diÄŸer korelasyon katsayÄ±larÄ± da kullanÄ±labilir.

Pearson korelasyon katsayÄ±sÄ±, iki deÄŸiÅŸken arasÄ±ndaki lineer iliÅŸkiyi Ã¶lÃ§en bir istatistiksel deÄŸerdir. DeÄŸer, -1 ile 1 arasÄ±nda bir aralÄ±kta olabilir. -1, tam ters orantÄ±lÄ± bir iliÅŸkiyi, 1 ise tam orantÄ±lÄ± bir iliÅŸkiyi gÃ¶sterir. 0 ise iki deÄŸiÅŸken arasÄ±nda lineer bir iliÅŸki olmadÄ±ÄŸÄ±nÄ± gÃ¶sterir.

Kolerasyon anlamlÄ±lÄ±ÄŸÄ± testi genellikle ÅŸu hipotezleri test etmek iÃ§in kullanÄ±lÄ±r:

- **H0 (Null Hypothesis):** Ä°ki deÄŸiÅŸken arasÄ±nda bir iliÅŸki yoktur (Pearson korelasyon katsayÄ±sÄ± 0'dÄ±r).
- **H1 (Alternative Hypothesis):** Ä°ki deÄŸiÅŸken arasÄ±nda bir iliÅŸki vardÄ±r (Pearson korelasyon katsayÄ±sÄ± 0 deÄŸildir).

Bu test, genellikle Ã¶rneklemler arasÄ±ndaki kolerasyonun anlamlÄ± olup olmadÄ±ÄŸÄ±nÄ± belirlemek iÃ§in kullanÄ±lÄ±r. Test sonuÃ§larÄ±, belirli bir gÃ¼ven dÃ¼zeyinde (genellikle %95 veya %99) istatistiksel olarak anlamlÄ± olup olmadÄ±ÄŸÄ±nÄ± belirtir.

Kolerasyon anlamlÄ±lÄ±ÄŸÄ± testleri genellikle istatistiksel yazÄ±lÄ±mlar veya istatistiksel paketler kullanÄ±larak gerÃ§ekleÅŸtirilir. Python, R ve SPSS gibi araÃ§lar bu tÃ¼r analizler iÃ§in sÄ±klÄ±kla kullanÄ±lan araÃ§lardÄ±r.

````python
from scipy.stats import pearsonr #pearson korelasyon katsayÄ±sÄ±nÄ± hesaplamak iÃ§in gerekli modÃ¼lÃ¼ import edelim
````

````python
test_istatistigi, pvalue = pearsonr(df["tip"], df["total_bill"]) #pearson korelasyon katsayÄ±sÄ±nÄ± hesaplÄ±yoruz, bu fonksiyonun iÃ§erisine x ve y deÄŸiÅŸkenlerini giriyoruz. sol taraftaki deÄŸer test istatistiÄŸini, saÄŸ taraftaki deÄŸer ise p-value deÄŸerini verir

print('Korelasyon KatsayÄ±sÄ±: %.10f, P-deÄŸeri: %.20f' % (test_istatistigi, pvalue))
````

DeÄŸiÅŸkenler arasÄ±nda anlamlÄ± bir iliÅŸki yoktur diyen H0 hipotezini reddediyoruz. Yani deÄŸiÅŸkenler arasÄ±nda anlamlÄ± bir iliÅŸki, korelasyon vardÄ±r diyoruz.

## Nonparametrik Korelasyon Hipotez Testi



Nonparametrik korelasyon testleri, deÄŸiÅŸkenler arasÄ±ndaki iliÅŸkiyi deÄŸerlendirmek iÃ§in kullanÄ±lan istatistiksel yÃ¶ntemlerdir. Bu testler, verilerin belirli bir daÄŸÄ±lÄ±ma uymadÄ±ÄŸÄ± durumlar veya sÄ±ralÄ± verilerle Ã§alÄ±ÅŸÄ±ldÄ±ÄŸÄ±nda tercih edilebilir. Nonparametrik korelasyon testleri, parametrik korelasyon testlerine alternatif olarak kullanÄ±lÄ±r ve bazen veriler normal daÄŸÄ±lÄ±ma uymadÄ±ÄŸÄ±nda veya sÄ±ralÄ± verilerle Ã§alÄ±ÅŸÄ±ldÄ±ÄŸÄ±nda daha gÃ¼venilir sonuÃ§lar verir.

Ä°ÅŸte bazÄ± yaygÄ±n nonparametrik korelasyon testleri:

1. **Spearman's Rho (Spearman Rank Korelasyon KatsayÄ±sÄ±):** Spearman korelasyon katsayÄ±sÄ±, deÄŸiÅŸkenler arasÄ±ndaki monotonic bir iliÅŸkiyi Ã¶lÃ§er. Monotonik bir iliÅŸki, deÄŸiÅŸkenlerin birbirine doÄŸrusal olmayan bir ÅŸekilde sÄ±ralanabilir olduÄŸu anlamÄ±na gelir.

2. **Kendall's Tau (Kendall Tau Rank Korelasyon KatsayÄ±sÄ±):** Kendall korelasyon katsayÄ±sÄ± da sÄ±ralÄ± veriler arasÄ±ndaki iliÅŸkiyi deÄŸerlendiren bir nonparametrik korelasyon testidir. Spearman gibi, Kendall da monotonic bir iliÅŸkiyi Ã¶lÃ§er.

Bu nonparametrik korelasyon testleri, Ã¶zellikle deÄŸiÅŸkenler arasÄ±ndaki iliÅŸkinin doÄŸrusal olmadÄ±ÄŸÄ± durumlarda veya verilerin normal daÄŸÄ±lÄ±ma uymadÄ±ÄŸÄ± durumlarda tercih edilir. Parametrik testlerin varsayÄ±mlarÄ±na uymayan durumlarda daha gÃ¼Ã§lÃ¼ ve gÃ¼venilir sonuÃ§lar saÄŸlayabilirler. Bu testler genellikle sÄ±ralÄ± verilerle Ã§alÄ±ÅŸÄ±rken, parametrik testler genellikle sÃ¼rekli veriler Ã¼zerinde Ã§alÄ±ÅŸÄ±r.

````python
#varsayalÄ±m saÄŸlanmadÄ±ÄŸÄ±ndan yapabileceÄŸimiz iki tane nonparametrik yÃ¶ntem var: spearman ve kendall
import pandas as pd
from scipy.stats import spearmanr # Import spearmanr directly

spearmanr(df["tip"], df["total_bill"]) #spearman korelasyon katsayÄ±sÄ±nÄ± hesaplÄ±yoruz, bu fonksiyonun iÃ§erisine x ve y deÄŸiÅŸkenlerini giriyoruz
````

````python
# Spearman korelasyon katsayÄ±sÄ±nÄ± hesaplama
# stats.spearmanr yerine doÄŸrudan spearmanr kullanÄ±yoruz
rho, p_value = spearmanr(df["tip"], df["total_bill"])

print(f"Spearman Korelasyon KatsayÄ±sÄ± (rho): {rho:.4f}")
print(f"p-deÄŸeri: {p_value:.4f}")

# Yorumlama
alpha = 0.05
if p_value < alpha:
    print(f"\nSonuÃ§: p-deÄŸeri ({p_value:.4f}) < alpha ({alpha}) olduÄŸundan,")
    print("BahÅŸiÅŸ miktarÄ± ile toplam hesap arasÄ±nda istatistiksel olarak **anlamlÄ± bir Spearman korelasyonu** vardÄ±r.")
else:
    print(f"\nSonuÃ§: p-deÄŸeri ({p_value:.4f}) >= alpha ({alpha}) olduÄŸundan,")
    print("BahÅŸiÅŸ miktarÄ± ile toplam hesap arasÄ±nda istatistiksel olarak **anlamlÄ± bir Spearman korelasyonu yoktur**.")
````

````python
from scipy.stats import spearmanr,kendalltau
test_istatistigi,pvalue=spearmanr(df['tip'],df['total_bill'])  
print('test istatistiÄŸi/Korelasyon KatsayÄ±sÄ± =%.4f , p-value =%.24f' % (test_istatistigi,pvalue)) 

test_istatistigi,pvalue=kendalltau(df['tip'],df['total_bill'])
print('test istatistiÄŸi/Korelasyon KatsayÄ±sÄ± =%.4f , p-value =%.24f' % (test_istatistigi,pvalue)) 
````
