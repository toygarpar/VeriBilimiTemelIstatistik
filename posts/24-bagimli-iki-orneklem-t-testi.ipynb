##  **Testin hipotezleri ÅŸu ÅŸekildedir:**

  
  

## $t = \frac{\bar{d} - \mu_d}{\frac{s_d}{\sqrt{n}}}$

  

Burada:

- $t $: Test istatistiÄŸi

- $ \bar{d} $: Ä°ki Ã¶rnek arasÄ±ndaki farkÄ±n ortalamasÄ±

- $ \mu_d $: PopÃ¼lasyondaki farkÄ±n teorik ortalamasÄ± (genellikle sÄ±fÄ±r olarak alÄ±nÄ±r, Ã§Ã¼nkÃ¼ H0 hipotezi genellikle "iki durum arasÄ±nda bir fark yoktur" ÅŸeklindedir.)

- $s_d $: Ä°ki Ã¶rnek arasÄ±ndaki farklarÄ±n standart sapmasÄ±

- $n$: Ã–rneklemin bÃ¼yÃ¼klÃ¼ÄŸÃ¼

  

Bu formÃ¼lde, $t $ test istatistiÄŸi, iki Ã¶rnek arasÄ±ndaki farkÄ±n Ã¶rnek ortalamasÄ± $ \bar{d} $ ile farkÄ±n popÃ¼lasyon ortalamasÄ± $ \mu_d $ arasÄ±ndaki farkÄ±n standart hata ${\frac{s_d}{\sqrt{n}}}$ oranÄ±dÄ±r.

  

Bu test istatistiÄŸi daha sonra belirli bir anlamlÄ±lÄ±k dÃ¼zeyinde (genellikle 0.05) karar vermek iÃ§in t-daÄŸÄ±lÄ±m tablosu kullanÄ±larak karÅŸÄ±laÅŸtÄ±rÄ±lÄ±r. EÄŸer test istatistiÄŸi, t-daÄŸÄ±lÄ±m tablosundaki kritik deÄŸerden bÃ¼yÃ¼kse, H0 hipotezi reddedilir ve iki durum arasÄ±nda anlamlÄ± bir fark olduÄŸu sonucuna varÄ±lÄ±r.

  
  

- **Null Hypothesis (H0):** Ä°ki koÅŸul arasÄ±nda herhangi bir fark yoktur.

- **Alternative Hypothesis (H1):** Ä°ki koÅŸul arasÄ±nda istatistiksel olarak anlamlÄ± bir fark vardÄ±r.

Bu testin sonucunda elde edilen t istatistiÄŸi ve p deÄŸeri, null hipotezin reddedilip edilemeyeceÄŸini belirler. DÃ¼ÅŸÃ¼k bir p deÄŸeri (genellikle 0.05'ten kÃ¼Ã§Ã¼k) null hipotezin reddedilmesi iÃ§in yeterli kanÄ±t olarak kabul edilir, bu da iki koÅŸul arasÄ±nda bir fark olduÄŸunu gÃ¶sterir.

  

Python'da, bu testi gerÃ§ekleÅŸtirmek iÃ§in `scipy` kÃ¼tÃ¼phanesindeki `stats.ttest_rel` fonksiyonunu kullanabilirsiniz. Ã–rnek bir kullanÄ±m ÅŸu ÅŸekildedir:















**BaÄŸÄ±mlÄ± Ä°ki Ã–rneklem T Testi NasÄ±l YapÄ±lÄ±r?**

    - Hâ‚€: Ä°ki Ã¶lÃ§Ã¼m arasÄ±nda ortalama fark yoktur (Î¼d = 0). Burada Î¼d, fark puanlarÄ±nÄ±n ortalamasÄ±dÄ±r.
    - Hâ‚: Ä°ki Ã¶lÃ§Ã¼m arasÄ±nda ortalama fark vardÄ±r (Î¼d â‰  0) (Ä°ki yÃ¶nlÃ¼ test). Veya, bir Ã¶lÃ§Ã¼m diÄŸerinden bÃ¼yÃ¼ktÃ¼r (Î¼d > 0 veya Î¼d < 0) (Tek yÃ¶nlÃ¼ test).

3. **Fark PuanlarÄ±nÄ± HesaplayÄ±n:** Her bir denek iÃ§in iki Ã¶lÃ§Ã¼m arasÄ±ndaki farkÄ± hesaplayÄ±n (d = Xâ‚ - Xâ‚‚).
    
4. **Fark PuanlarÄ±nÄ±n OrtalamasÄ±nÄ± (dÌ„) ve Standart SapmasÄ±nÄ± (sd) HesaplayÄ±n:**
    
5. **T Ä°statistiÄŸini HesaplayÄ±n:**
    
    ```
    t = dÌ„ / (sd / âˆšn)
    ```

    Burada:

8. **Karar Verin:**
    
    - EÄŸer hesaplanan t istatistiÄŸinin mutlak deÄŸeri, kritik t deÄŸerinden bÃ¼yÃ¼kse, sÄ±fÄ±r hipotezi reddedilir. Bu, iki Ã¶lÃ§Ã¼m arasÄ±nda istatistiksel olarak anlamlÄ± bir fark olduÄŸu anlamÄ±na gelir.
    - EÄŸer hesaplanan t istatistiÄŸinin mutlak deÄŸeri, kritik t deÄŸerinden kÃ¼Ã§Ã¼kse, sÄ±fÄ±r hipotezi reddedilmez.

````python
# AdÄ±m 2: AnlamlÄ±lÄ±k DÃ¼zeyini Belirle

alpha = 0.05
print(f"AnlamlÄ±lÄ±k dÃ¼zeyi (Î±): {alpha}")

# %5 yanÄ±lma payÄ± kabul ettik.
````

````python
# AdÄ±m 3:sample veri Ã¼ret

import numpy as np

np.random.seed(42)
n = 20

# Ã–ncesi ve sonrasÄ± veriler
kan_basinci_oncesi = np.random.normal(loc=140, scale=10, size=n)
kan_basinci_sonrasi = kan_basinci_oncesi - np.random.normal(loc=10, scale=5, size=n)

print("Ortalama Kan BasÄ±ncÄ± Ã–ncesi:", kan_basinci_oncesi.mean().round(2))
print("Ortalama Kan BasÄ±ncÄ± SonrasÄ±:", kan_basinci_sonrasi.mean().round(2))
````

````python
# AdÄ±m 4: FarklarÄ± Hesapla

farklar = kan_basinci_oncesi - kan_basinci_sonrasi
print("FarklarÄ±n OrtalamasÄ±:", farklar.mean().round(2))
print("FarklarÄ±n Standart SapmasÄ±:", farklar.std().round(2))

#ArtÄ±k farklar Ã¼zerinden test yapabiliriz.
````

````python
# AdÄ±m 5: Paired T-Testi Yap

from scipy.stats import ttest_rel

t_skoru, p_degeri = ttest_rel(kan_basinci_oncesi, kan_basinci_sonrasi)
print(f"T Skoru: {t_skoru:.2f}, p-deÄŸeri: {p_degeri:.4f}")
````

````python
# AdÄ±m 6: Karar Ver


if p_degeri < alpha:
    print("âœ… Hâ‚€ reddedildi â†’ Ä°lacÄ±n etkisi var")
else:
    print("âŒ Hâ‚€ reddedilemedi â†’ Etki tesadÃ¼fi olabilir")


# Yani ilacÄ±n kan basÄ±ncÄ±nÄ± dÃ¼ÅŸÃ¼rdÃ¼ÄŸÃ¼nÃ¼ sayÄ±larla kanÄ±tlanÄ±yor.
````

## ğŸ“Œ BaÄŸÄ±mlÄ± T-Testi ve Modelleme SÃ¼reci BaÄŸlantÄ±sÄ±

ğŸ“Œ Ã–zellikle zamana baÄŸlÄ± verilerde gÃ¼Ã§lÃ¼ bir araÃ§tÄ±r.

**Ã–rnek:**

Bir diyet programÄ±nÄ±n kilo verme Ã¼zerindeki etkisini Ã¶lÃ§mek iÃ§in bir araÅŸtÄ±rma yapÄ±lÄ±yor. 10 kiÅŸi programa baÅŸlamadan Ã¶nce ve program bittikten sonra tartÄ±lÄ±yor. Kilo deÄŸiÅŸimleri aÅŸaÄŸÄ±daki gibi olsun:


|KiÅŸi|BaÅŸlangÄ±Ã§ Kilosu (Xâ‚)|BitiÅŸ Kilosu (Xâ‚‚)|Fark (d = Xâ‚ - Xâ‚‚)|
|---|---|---|---|
|1|80|75|5|
|2|90|82|8|
|3|75|70|5|
|4|85|80|5|
|5|70|68|2|
|6|95|88|7|
|7|65|62|3|
|8|100|92|8|
|9|78|72|6|
|10|82|77|5|



Bu verilerle BaÄŸÄ±mlÄ± Ä°ki Ã–rneklem T Testi yapalÄ±m (Î± = 0.05):

1. **Hipotezler:**
    
    - Hâ‚€: Î¼d = 0 (Diyetin kilo verme Ã¼zerinde etkisi yoktur)
    - Hâ‚: Î¼d > 0 (Diyet kilo vermeye yardÄ±mcÄ± olur - Tek yÃ¶nlÃ¼ test)
2. **Fark PuanlarÄ±nÄ±n OrtalamasÄ± ve Standart SapmasÄ±:**
    
    dÌ„ = 5.4 sd = 2.07
    
3. **T Ä°statistiÄŸi:**
    
    ```
    t = 5.4 / (2.07 / âˆš10) â‰ˆ 8.24
    ```
    
4. **Serbestlik Derecesi:**
    
    sd = 10 - 1 = 9
    
5. **Kritik T DeÄŸeri:**
    
    Î± = 0.05 ve sd = 9 iÃ§in tek yÃ¶nlÃ¼ t tablosundan kritik t deÄŸeri yaklaÅŸÄ±k olarak 1.833 olarak bulunur.
    
6. **Karar:**
    
    Hesaplanan t istatistiÄŸi (8.24), kritik t deÄŸerinden (1.833) bÃ¼yÃ¼ktÃ¼r. Bu nedenle, sÄ±fÄ±r hipotezi reddedilir. Bu, diyet programÄ±nÄ±n kilo verme Ã¼zerinde istatistiksel olarak anlamlÄ± bir etkisi olduÄŸunu gÃ¶sterir.


````python
import numpy as np
from scipy import stats

# Verilerin tanÄ±mlanmasÄ±
baslangic = np.array([80, 90, 75, 85, 70, 95, 65, 100, 78, 82])
bitis = np.array([75, 82, 70, 80, 68, 88, 62, 92, 72, 77])

# BaÄŸÄ±mlÄ± iki Ã¶rneklem t testi (paired t-test)
t_statistic, p_value = stats.ttest_rel(baslangic, bitis)

# Tek yÃ¶nlÃ¼ test iÃ§in p deÄŸerinin yarÄ±sÄ±nÄ± alÄ±yoruz (H1: Î¼d > 0)
p_value_one_tailed = p_value / 2
````

````python
# SonuÃ§larÄ±n yazdÄ±rÄ±lmasÄ±
print("BaÄŸÄ±mlÄ± Ä°ki Ã–rneklem T Testi SonuÃ§larÄ±:")
print(f"T istatistiÄŸi: {t_statistic:.4f}")
print(f"P deÄŸeri (tek yÃ¶nlÃ¼): {p_value_one_tailed:.6f}")
````

````python
# Kritik t deÄŸerinin hesaplanmasÄ±
alpha = 0.05
degrees_of_freedom = len(baslangic) - 1
critical_t = stats.t.ppf(1 - alpha, degrees_of_freedom)
print(f"Kritik t deÄŸeri (Î±=0.05, tek yÃ¶nlÃ¼): {critical_t:.4f}")
````

````python
# Karar kuralÄ±
if t_statistic > critical_t:
    print("SonuÃ§: H0 reddedilir - Diyetin kilo verme Ã¼zerinde anlamlÄ± bir etkisi vardÄ±r.")
else:
    print("SonuÃ§: H0 reddedilemez - Diyetin kilo verme Ã¼zerinde anlamlÄ± bir etkisi yoktur.")
````

````python
# Ek bilgiler
farklar = baslangic - bitis
print("\nEk Bilgiler:")
print(f"FarklarÄ±n ortalamasÄ±: {np.mean(farklar):.2f}")
print(f"FarklarÄ±n standart sapmasÄ±: {np.std(farklar, ddof=1):.2f}")
print(f"Ã–rneklem bÃ¼yÃ¼klÃ¼ÄŸÃ¼: {len(baslangic)}")
````

# AdÄ±m adÄ±m gÃ¶rmek istersek:

````python
# baÄŸÄ±mlÄ± iki Ã¶rneklem t-testi (paired samples t-test) iÃ§in adÄ±m adÄ±m python kod:
import numpy as np
from scipy import stats

# Veriler
baslangic = np.array([80, 90, 75, 85, 70, 95, 65, 100, 78, 82])
bitis = np.array([75, 82, 70, 80, 68, 88, 62, 92, 72, 77])

# 1. FarklarÄ±n hesaplanmasÄ±
farklar = baslangic - bitis

# 2. TanÄ±mlayÄ±cÄ± istatistikler
ortalama_fark = np.mean(farklar)
std_fark = np.std(farklar, ddof=1)  # ddof=1 ile Ã¶rneklem standart sapmasÄ±
n = len(farklar)

print("--- TanÄ±mlayÄ±cÄ± Ä°statistikler ---")
print(f"Ortalama Fark (BaÅŸlangÄ±Ã§ - BitiÅŸ): {ortalama_fark:.2f} kg")
print(f"Standart Sapma: {std_fark:.2f} kg")
print(f"Ã–rneklem BÃ¼yÃ¼klÃ¼ÄŸÃ¼ (n): {n}")
````

````python
# 3. BaÄŸÄ±mlÄ± t-testi (paired t-test)
# NOT: scipy varsayÄ±lan olarak iki yÃ¶nlÃ¼ test yapar, tek yÃ¶nlÃ¼ iÃ§in p-deÄŸerini 2'ye bÃ¶leriz
t_statistic, p_value = stats.ttest_rel(baslangic, bitis)
p_value_tek_yonlu = p_value / 2  # Ã‡Ã¼nkÃ¼ H1: Î¼d > 0 (tek yÃ¶nlÃ¼)

print("\n--- Hipotez Testi SonuÃ§larÄ± ---")
print(f"t-istatistiÄŸi: {t_statistic:.4f}")
print(f"p-deÄŸeri (tek yÃ¶nlÃ¼): {p_value_tek_yonlu:.6f}")
````

````python
# 4. Kritik t deÄŸeri (Î± = 0.05, tek yÃ¶nlÃ¼)
alpha = 0.05
serbestlik_derecesi = n - 1
kritik_t = stats.t.ppf(1 - alpha, serbestlik_derecesi)
print(f"\nKritik t deÄŸeri (Î±={alpha}, tek yÃ¶nlÃ¼): {kritik_t:.4f}")
````

````python
# 5. Karar
if t_statistic > kritik_t:
    print("\nSonuÃ§: H0 REDDEDÄ°LÄ°R (Diyetin kilo verme Ã¼zerinde anlamlÄ± bir etkisi vardÄ±r.)")
else:
    print("\nSonuÃ§: H0 REDDEDÄ°LEMEZ (Diyetin etkisi istatistiksel olarak anlamlÄ± deÄŸildir.)")
````

````python
# 6. Ek: GÃ¼ven aralÄ±ÄŸÄ± (isteÄŸe baÄŸlÄ±)
guven_araligi = stats.t.interval(confidence=0.95, df=serbestlik_derecesi, 
                                loc=ortalama_fark, scale=std_fark/np.sqrt(n))
print(f"\nOrtalama fark iÃ§in %95 gÃ¼ven aralÄ±ÄŸÄ±: ({guven_araligi[0]:.2f}, {guven_araligi[1]:.2f}) kg")
````

````python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Ã–rnek veri oluÅŸturma (eÄŸitim Ã¶ncesi ve sonrasÄ± sÄ±nav notlarÄ±)
np.random.seed(42)  # Tekrarlanabilirlik iÃ§in
data = {
    'KatÄ±lÄ±mcÄ±': [f'K{i+1}' for i in range(20)],
    'Ã–n_Test': np.random.normal(60, 10, 20).round(1),
    'Son_Test': np.random.normal(70, 8, 20).round(1)  # EÄŸitim sonrasÄ± notlar daha yÃ¼ksek
}

# Veriyi DataFrame'e dÃ¶nÃ¼ÅŸtÃ¼rme
df = pd.DataFrame(data)
df['Fark'] = df['Son_Test'] - df['Ã–n_Test']
````

- Ã–lÃ§Ã¼m: Ä°lk eÄŸitim yÃ¶ntemi ve ikinci eÄŸitim yÃ¶ntemi Ã¶ncesi ve sonrasÄ± performans puanlarÄ±.

- Hipotez: Ä°ki eÄŸitim yÃ¶ntemi arasÄ±nda beceride istatistiksel olarak anlamlÄ± bir fark var mÄ±?

````python
# EÄŸitim YÃ¶ntemlerinin KarÅŸÄ±laÅŸtÄ±rÄ±lmasÄ± iÃ§in BaÄŸÄ±mlÄ± ve BaÄŸÄ±msÄ±z Ã–rneklem Analizi

import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.power import TTestIndPower

# 1. Veri OluÅŸturma ve HazÄ±rlÄ±k
np.random.seed(51)
katilimci_sayisi = 30  # Her grup iÃ§in

# Grup 1 (Geleneksel YÃ¶ntem)
grup1_on = np.random.normal(65, 8, katilimci_sayisi).round(1)
grup1_son = np.random.normal(72, 7, katilimci_sayisi).round(1)

# Grup 2 (Yeni YÃ¶ntem)
grup2_on = np.random.normal(64, 9, katilimci_sayisi).round(1)
grup2_son = np.random.normal(78, 6, katilimci_sayisi).round(1)

# DataFrame oluÅŸturma
df = pd.DataFrame({
    'Grup': ['Geleneksel']*katilimci_sayisi + ['Yeni']*katilimci_sayisi,
    'KatÄ±lÄ±mcÄ±': [f'K{i+1}' for i in range(katilimci_sayisi)]*2,
    'Ã–n_Test': np.concatenate([grup1_on, grup2_on]),
    'Son_Test': np.concatenate([grup1_son, grup2_son])
})

# FarklarÄ± hesaplama
df['Fark'] = df['Son_Test'] - df['Ã–n_Test']
df['YÃ¼zde_ArtÄ±ÅŸ'] = (df['Fark'] / df['Ã–n_Test']) * 100
````

````python
# 2. TanÄ±mlayÄ±cÄ± Ä°statistikler
print("="*60)
print("ğŸ“š EÄÄ°TÄ°M YÃ–NTEMLERÄ° KARÅILAÅTIRMA ANALÄ°ZÄ°")
print("="*60)

print("\nğŸ“Š Gruplara GÃ¶re Temel Ä°statistikler:")
group_stats = df.groupby('Grup')[['Ã–n_Test', 'Son_Test', 'Fark', 'YÃ¼zde_ArtÄ±ÅŸ']].agg(['mean', 'std', 'count'])
print(group_stats.round(2))
````

````python
# 3. GÃ¶rselleÅŸtirme
plt.figure(figsize=(18, 6))

# Kutu grafikleri
plt.subplot(1, 3, 1)
sns.boxplot(x='Grup', y='Fark', data=df, palette="Set2")
plt.title('Gruplara GÃ¶re Performans ArtÄ±ÅŸÄ± DaÄŸÄ±lÄ±mÄ±')
plt.ylabel('Performans ArtÄ±ÅŸÄ± (Puan)')

# Ã–n-Son Test KarÅŸÄ±laÅŸtÄ±rmasÄ±
plt.subplot(1, 3, 2)
sns.barplot(x='Grup', y='value', hue='variable', 
            data=pd.melt(df, id_vars=['Grup'], value_vars=['Ã–n_Test', 'Son_Test']),
            palette="coolwarm", ci='sd')
plt.title('GruplarÄ±n Ã–n ve Son Test OrtalamalarÄ±')
plt.ylabel('Ortalama Puan')

# Performans ArtÄ±ÅŸÄ± DaÄŸÄ±lÄ±mÄ±
plt.subplot(1, 3, 3)
sns.histplot(data=df, x='Fark', hue='Grup', kde=True, 
             palette="viridis", bins=10, element="step")
plt.title('Performans ArtÄ±ÅŸÄ± DaÄŸÄ±lÄ±mlarÄ±')
plt.xlabel('Performans ArtÄ±ÅŸÄ± (Puan)')

plt.tight_layout()
plt.show()
````

````python
# 4. Normallik Testleri
print("\nğŸ“ Normallik Testleri (Shapiro-Wilk):")
for grup in df['Grup'].unique():
    stat, p = stats.shapiro(df[df['Grup']==grup]['Fark'])
    print(f"{grup} YÃ¶ntemi FarklarÄ±: W = {stat:.3f}, p = {p:.4f}")
````

````python
# 5. Varyans HomojenliÄŸi Testi (Levene)
print("\nğŸ” Varyans HomojenliÄŸi Testi (Levene):")
lev_stat, lev_p = stats.levene(df[df['Grup']=='Geleneksel']['Fark'],
                             df[df['Grup']=='Yeni']['Fark'])
print(f"Levene Testi: F = {lev_stat:.3f}, p = {lev_p:.4f}")
````

````python
# 6. BaÄŸÄ±msÄ±z Ä°ki Ã–rneklem T Testi (YÃ¶ntemler ArasÄ± KarÅŸÄ±laÅŸtÄ±rma)
print("\nğŸ”¬ BaÄŸÄ±msÄ±z Ä°ki Ã–rneklem T Testi SonuÃ§larÄ±:")

# Hipotezler
print("\nHipotezler:")
print("Hâ‚€: Ä°ki eÄŸitim yÃ¶ntemi arasÄ±nda performans artÄ±ÅŸÄ± farkÄ± yoktur (Î¼1 = Î¼2)")
print("Hâ‚: Ä°ki eÄŸitim yÃ¶ntemi arasÄ±nda performans artÄ±ÅŸÄ± farkÄ± vardÄ±r (Î¼1 â‰  Î¼2)")

# Test (varyanslarÄ±n homojenliÄŸine gÃ¶re)
if lev_p > 0.05:
    t_stat, p_value = stats.ttest_ind(df[df['Grup']=='Geleneksel']['Fark'],
                                    df[df['Grup']=='Yeni']['Fark'],
                                    equal_var=True)
    print("\nVaryanslar homojen (p > 0.05) - EÅŸit varyanslÄ± t-testi kullanÄ±ldÄ±")
else:
    t_stat, p_value = stats.ttest_ind(df[df['Grup']=='Geleneksel']['Fark'],
                                    df[df['Grup']=='Yeni']['Fark'],
                                    equal_var=False)
    print("\nVaryanslar homojen deÄŸil (p â‰¤ 0.05) - Welch'in t-testi kullanÄ±ldÄ±")

print(f"\nt-istatistiÄŸi: {t_stat:.4f}")
print(f"p-deÄŸeri: {p_value:.4f}")

# Karar kuralÄ± (Î± = 0.05)
alpha = 0.05
if p_value < alpha:
    print(f"\nâœ… SONUÃ‡: p < {alpha} â†’ Hâ‚€ REDDEDÄ°LÄ°R")
    print("Ä°ki eÄŸitim yÃ¶ntemi arasÄ±nda istatistiksel olarak anlamlÄ± bir fark vardÄ±r.")
    
    # Hangi yÃ¶ntem daha iyi?
    if df[df['Grup']=='Geleneksel']['Fark'].mean() > df[df['Grup']=='Yeni']['Fark'].mean():
        print("â†’ Geleneksel yÃ¶ntem daha fazla performans artÄ±ÅŸÄ± saÄŸlamÄ±ÅŸtÄ±r")
    else:
        print("â†’ Yeni yÃ¶ntem daha fazla performans artÄ±ÅŸÄ± saÄŸlamÄ±ÅŸtÄ±r")
else:
    print(f"\nâŒ SONUÃ‡: p â‰¥ {alpha} â†’ Hâ‚€ REDDEDÄ°LEMEZ")
    print("Ä°ki eÄŸitim yÃ¶ntemi arasÄ±nda istatistiksel olarak anlamlÄ± bir fark yoktur.")
````

````python
# 7. KapsamlÄ± Ek Analizler
print("\nğŸ“ˆ DetaylÄ± Analizler:")

# Etki bÃ¼yÃ¼klÃ¼ÄŸÃ¼ (Cohen's d)
def cohens_d(group1, group2):
    diff = group1.mean() - group2.mean()
    n1, n2 = len(group1), len(group2)
    var1, var2 = group1.var(ddof=1), group2.var(ddof=1)
    pooled_std = np.sqrt(((n1-1)*var1 + (n2-1)*var2) / (n1+n2-2))
    return diff / pooled_std

d = cohens_d(df[df['Grup']=='Yeni']['Fark'], 
             df[df['Grup']=='Geleneksel']['Fark'])
print(f"\nCohen's d (Etki BÃ¼yÃ¼klÃ¼ÄŸÃ¼): {abs(d):.2f}")

# Etki yorumu
if abs(d) < 0.2:
    etki_boyutu = "Ã‡ok kÃ¼Ã§Ã¼k etki"
elif 0.2 <= abs(d) < 0.5:
    etki_boyutu = "KÃ¼Ã§Ã¼k etki"
elif 0.5 <= abs(d) < 0.8:
    etki_boyutu = "Orta etki"
else:
    etki_boyutu = "BÃ¼yÃ¼k etki"
print(f"â†’ Yorum: {etki_boyutu}")

# GÃ¼Ã§ analizi
effect_size = abs(d)
alpha = 0.05
power = 0.8
analysis = TTestIndPower()
required_n = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power)
print(f"\nâ³ GÃ¼Ã§ analizi (0.8 gÃ¼Ã§, Î±=0.05 iÃ§in):")
print(f"â†’ Tespit edilebilir minimum etki bÃ¼yÃ¼klÃ¼ÄŸÃ¼: {effect_size:.2f}")
print(f"â†’ Ã–nerilen minimum Ã¶rneklem bÃ¼yÃ¼klÃ¼ÄŸÃ¼ (grup baÅŸÄ±na): {int(np.ceil(required_n))}")

# Grup iÃ§i deÄŸiÅŸimler (paired t-test)
print("\nğŸ” Grup Ä°Ã§i DeÄŸiÅŸim Analizleri:")
for grup in df['Grup'].unique():
    t_stat, p_value = stats.ttest_rel(df[df['Grup']==grup]['Son_Test'],
                                      df[df['Grup']==grup]['Ã–n_Test'])
    print(f"\n{grup} YÃ¶ntemi:")
    print(f"â†’ Ã–n Test Ort: {df[df['Grup']==grup]['Ã–n_Test'].mean():.1f}")
    print(f"â†’ Son Test Ort: {df[df['Grup']==grup]['Son_Test'].mean():.1f}")
    print(f"â†’ t({len(df[df['Grup']==grup])-1}) = {t_stat:.2f}, p = {p_value:.4f}")
    if p_value < 0.05:
        print("  â†’ Ä°statistiksel olarak anlamlÄ± geliÅŸme (p < 0.05)")
    else:
        print("  â†’ Ä°statistiksel olarak anlamlÄ± geliÅŸme yok")
````

5. **YazÄ±lÄ±m GÃ¼ncellemesi Etkisi:**

- Problem: Bir yazÄ±lÄ±m gÃ¼ncellemesi yapÄ±ldÄ±ktan sonra, kullanÄ±cÄ±larÄ±n belirli bir iÅŸlem sÃ¼resindeki deÄŸiÅŸimi deÄŸerlendirmek isteniyor.

- Ã–lÃ§Ã¼m: YazÄ±lÄ±m Ã¶ncesi ve yazÄ±lÄ±m sonrasÄ± iÅŸlem sÃ¼releri.

- Hipotez: YazÄ±lÄ±m gÃ¼ncellemesi Ã¶ncesi ve sonrasÄ±nda iÅŸlem sÃ¼relerinde istatistiksel olarak anlamlÄ± bir fark var mÄ±?


````python
# YazÄ±lÄ±m GÃ¼ncellemesinin Performans Etkisi Analizi# - Test: Ä°ki baÄŸÄ±mlÄ± Ã¶rneklem t-testi (Paired Sample T-Test)

import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.power import TTestPower

# 1. Veri HazÄ±rlama ve Ã–n Ä°ÅŸleme
np.random.seed(42)
kullanici_sayisi = 50

# GerÃ§ekÃ§i veri simÃ¼lasyonu (gÃ¼ncelleme sonrasÄ± iyileÅŸme olduÄŸunu varsayalÄ±m)
oncesi = np.random.exponential(scale=2.5, size=kullanici_sayisi).round(2)
sonrasi = np.random.exponential(scale=2.0, size=kullanici_sayisi).round(2)

# AykÄ±rÄ± deÄŸerleri kÄ±rpma (realist veri iÃ§in)
oncesi = np.clip(oncesi, 0.5, 8.0)
sonrasi = np.clip(sonrasi, 0.4, 6.5)

df = pd.DataFrame({
    'KullanÄ±cÄ±_ID': [f'K{i+1:03d}' for i in range(kullanici_sayisi)],
    'GÃ¼ncelleme_Ã–ncesi': oncesi,
    'GÃ¼ncelleme_SonrasÄ±': sonrasi
})

# Performans farkÄ±nÄ± hesaplama
df['SÃ¼re_FarkÄ±'] = df['GÃ¼ncelleme_Ã–ncesi'] - df['GÃ¼ncelleme_SonrasÄ±']
df['YÃ¼zde_Ä°yileÅŸme'] = (df['SÃ¼re_FarkÄ±'] / df['GÃ¼ncelleme_Ã–ncesi']) * 100
````

````python
# 2. TanÄ±mlayÄ±cÄ± Ä°statistikler
print("="*65)
print("ğŸ–¥ï¸ YAZILIM GÃœNCELLEMESÄ° PERFORMANS ANALÄ°ZÄ°")
print("="*65)

print("\nğŸ“Š Temel Performans Ä°statistikleri:")
desc_stats = df[['GÃ¼ncelleme_Ã–ncesi', 'GÃ¼ncelleme_SonrasÄ±', 'SÃ¼re_FarkÄ±', 'YÃ¼zde_Ä°yileÅŸme']].describe().T
desc_stats['IQR'] = desc_stats['75%'] - desc_stats['25%']
print(desc_stats[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 'IQR']].round(2))
````

````python
# 3. Veri GÃ¶rselleÅŸtirme
plt.figure(figsize=(16, 5))

# SÃ¼re DaÄŸÄ±lÄ±mlarÄ±
plt.subplot(1, 3, 1)
sns.violinplot(data=df[['GÃ¼ncelleme_Ã–ncesi', 'GÃ¼ncelleme_SonrasÄ±']], palette="coolwarm")
plt.title('Ä°ÅŸlem SÃ¼resi DaÄŸÄ±lÄ±mlarÄ±\n(GÃ¼ncelleme Ã–ncesi ve SonrasÄ±)')
plt.ylabel('SÃ¼re (saniye)')

# Performans Ä°yileÅŸmesi
plt.subplot(1, 3, 2)
sns.histplot(df['SÃ¼re_FarkÄ±'], kde=True, color='green', bins=15)
plt.axvline(df['SÃ¼re_FarkÄ±'].mean(), color='red', linestyle='--', label=f'Ortalama: {df["SÃ¼re_FarkÄ±"].mean():.2f}s')
plt.title('Performans Ä°yileÅŸme DaÄŸÄ±lÄ±mÄ±')
plt.xlabel('SÃ¼re FarkÄ± (saniye)')
plt.legend()

# Korelasyon Analizi
plt.subplot(1, 3, 3)
sns.regplot(x='GÃ¼ncelleme_Ã–ncesi', y='SÃ¼re_FarkÄ±', data=df, 
            scatter_kws={'alpha':0.6}, line_kws={'color':'orange'})
plt.axhline(0, color='gray', linestyle='--')
plt.title('BaÅŸlangÄ±Ã§ SÃ¼resi-Ä°yileÅŸme Ä°liÅŸkisi')
plt.xlabel('GÃ¼ncelleme Ã–ncesi SÃ¼re (s)')
plt.ylabel('Ä°yileÅŸme MiktarÄ± (s)')

plt.tight_layout()
plt.show()
````

````python
# 4. Normallik Testi
print("\nğŸ“ Normallik Testi SonuÃ§larÄ± (Shapiro-Wilk):")
shapiro_stat, shapiro_p = stats.shapiro(df['SÃ¼re_FarkÄ±'])
print(f"Farklar iÃ§in: W = {shapiro_stat:.3f}, p = {shapiro_p:.4f}")
if shapiro_p > 0.05:
    print("â†’ Farklar normal daÄŸÄ±lÄ±yor (p > 0.05)")
else:
    print("â†’ Farklar normal daÄŸÄ±lmÄ±yor (p â‰¤ 0.05)")
````

````python
# 5. BaÄŸÄ±mlÄ± T-Testi (Paired Samples)
print("\nğŸ” BaÄŸÄ±mlÄ± Ä°ki Ã–rneklem T Testi SonuÃ§larÄ±:")

# Hipotezler
print("\nHipotezler:")
print("Hâ‚€: GÃ¼ncelleme Ã¶ncesi ve sonrasÄ± iÅŸlem sÃ¼releri arasÄ±nda fark yoktur (Î¼d = 0)")
print("Hâ‚: GÃ¼ncelleme sonrasÄ± iÅŸlem sÃ¼releri daha kÄ±sadÄ±r (Î¼d > 0)")

# Tek yÃ¶nlÃ¼ test iÃ§in
t_stat, p_value = stats.ttest_rel(df['GÃ¼ncelleme_Ã–ncesi'], df['GÃ¼ncelleme_SonrasÄ±'])
p_value_tek_yonlu = p_value / 2

print(f"\nt-istatistiÄŸi: {t_stat:.4f}")
print(f"p-deÄŸeri (tek yÃ¶nlÃ¼): {p_value_tek_yonlu:.6f}")

# Karar kuralÄ± (Î± = 0.05)
alpha = 0.05
if p_value_tek_yonlu < alpha:
    print(f"\nâœ… SONUÃ‡: p < {alpha} â†’ Hâ‚€ REDDEDÄ°LÄ°R")
    print("YazÄ±lÄ±m gÃ¼ncellemesi iÅŸlem sÃ¼relerinde istatistiksel olarak anlamlÄ± bir iyileÅŸme saÄŸlamÄ±ÅŸtÄ±r.")
else:
    print(f"\nâŒ SONUÃ‡: p â‰¥ {alpha} â†’ Hâ‚€ REDDEDÄ°LEMEZ")
    print("YazÄ±lÄ±m gÃ¼ncellemesinin iÅŸlem sÃ¼releri Ã¼zerinde istatistiksel olarak anlamlÄ± bir etkisi yoktur.")
````

````python
# 6. KapsamlÄ± Ek Analizler
print("\nğŸ“ˆ DetaylÄ± Performans Analizleri:")

# GÃ¼ven aralÄ±ÄŸÄ±
n = len(df)
dof = n - 1
ci = stats.t.interval(confidence=0.95, df=dof,
                     loc=df['SÃ¼re_FarkÄ±'].mean(),
                     scale=df['SÃ¼re_FarkÄ±'].std(ddof=1)/np.sqrt(n))
print(f"\nOrtalama iyileÅŸme iÃ§in %95 GÃ¼ven AralÄ±ÄŸÄ±: ({ci[0]:.3f}s, {ci[1]:.3f}s)")

# Etki bÃ¼yÃ¼klÃ¼ÄŸÃ¼ (Cohen's d)
cohen_d = df['SÃ¼re_FarkÄ±'].mean() / df['SÃ¼re_FarkÄ±'].std(ddof=1)
print(f"Cohen's d (Etki BÃ¼yÃ¼klÃ¼ÄŸÃ¼): {cohen_d:.3f}")

# Etki yorumu
if cohen_d < 0.2:
    etki_boyutu = "Ã‡ok kÃ¼Ã§Ã¼k etki"
elif 0.2 <= cohen_d < 0.5:
    etki_boyutu = "KÃ¼Ã§Ã¼k etki"
elif 0.5 <= cohen_d < 0.8:
    etki_boyutu = "Orta etki"
else:
    etki_boyutu = "BÃ¼yÃ¼k etki"
print(f"â†’ Yorum: {etki_boyutu}")

# Ä°yileÅŸen kullanÄ±cÄ± oranÄ±
iyilesen = (df['SÃ¼re_FarkÄ±'] > 0).sum()
iyilesme_orani = (iyilesen / n) * 100
print(f"\nğŸ‘¥ KullanÄ±cÄ±larÄ±n %{iyilesme_orani:.1f}'si ({iyilesen}/{n}) gÃ¼ncelleme sonrasÄ±nda daha hÄ±zlÄ± iÅŸlem deneyimi yaÅŸamÄ±ÅŸtÄ±r")

# GÃ¼Ã§ analizi
effect_size = cohen_d
power_analysis = TTestPower()
power = power_analysis.solve_power(effect_size=effect_size, nobs=n, alpha=alpha)
print(f"\nğŸ“Š GÃ¼Ã§ Analizi (Mevcut Ã§alÄ±ÅŸma iÃ§in):")
print(f"â†’ Tespit edilen etki bÃ¼yÃ¼klÃ¼ÄŸÃ¼: {effect_size:.3f}")
print(f"â†’ Ã‡alÄ±ÅŸmanÄ±n istatistiksel gÃ¼cÃ¼: {power:.3f}")

# SÃ¼re iyileÅŸmesi ile baÅŸlangÄ±Ã§ sÃ¼resi korelasyonu
corr, p_corr = stats.pearsonr(df['GÃ¼ncelleme_Ã–ncesi'], df['SÃ¼re_FarkÄ±'])
print(f"\nğŸ“‰ BaÅŸlangÄ±Ã§ SÃ¼resi-Ä°yileÅŸme Korelasyonu: r = {corr:.3f}, p = {p_corr:.4f}")
if p_corr < 0.05:
    if corr > 0:
        print("â†’ Daha yavaÅŸ sistemlerde daha fazla iyileÅŸme gÃ¶zlenmiÅŸtir")
    else:
        print("â†’ Daha hÄ±zlÄ± sistemlerde daha fazla iyileÅŸme gÃ¶zlenmiÅŸtir")
else:
    print("â†’ BaÅŸlangÄ±Ã§ sÃ¼resi ile iyileÅŸme miktarÄ± arasÄ±nda anlamlÄ± iliÅŸki yok")
````

## <span style ="color:red">**Non Parametrik BaÄŸÄ±mlÄ± Ä°ki Ã–rneklem T-testi**</span>

Nonparametrik baÄŸÄ±mlÄ± iki Ã¶rneklem testi, aynÄ± birey veya Ã¶ÄŸe Ã¼zerinde iki farklÄ± koÅŸul arasÄ±ndaki farklÄ±lÄ±klarÄ± deÄŸerlendirmek iÃ§in kullanÄ±lÄ±r. Bu tÃ¼r bir test, verilerin normal daÄŸÄ±lÄ±ma uymadÄ±ÄŸÄ± veya normal daÄŸÄ±lÄ±m varsayÄ±mlarÄ±nÄ±n saÄŸlanamadÄ±ÄŸÄ± durumlarda tercih edilebilir. Wilcoxon iÅŸaretli sÄ±ralar testi (Wilcoxon Signed-Rank Test) bu tÃ¼r bir nonparametrik baÄŸÄ±mlÄ± iki Ã¶rneklem testi Ã¶rneÄŸidir.

Ä°ÅŸte bir Ã¶rnek Wilcoxon iÅŸaretli sÄ±ralar testi kullanÄ±mÄ±:


VarsayalÄ±m Ã¶nceki iÅŸlemlerde normallik ve varyans homojenliÄŸi varsayÄ±m kontrolleri saÄŸlanamadÄ±.

````python
from scipy.stats import wilcoxon #non-parametrik baÄŸÄ±mlÄ± iki Ã¶rneklem t testi uygulamak iÃ§in gerekli modÃ¼lÃ¼ import edelim
````

````python
wilcoxon(AYRIK.ONCESI, AYRIK.SONRASI) #non-parametrik baÄŸÄ±mlÄ± iki Ã¶rneklem t testi uyguluyoruz, bu fonsiyonun iÃ§erisine testini yapmak istediÄŸimiz iki deÄŸiÅŸkeni giriyoruz. sol taraftaki deÄŸer test istatistiÄŸini, saÄŸ taraftaki deÄŸer ise p-value deÄŸerini verir
````

````python
#daha yapÄ±sal bir formda gÃ¶rmek iÃ§in:
# Wilcoxon iÅŸaretli sÄ±ralar testi
statistic, p_value = wilcoxon(AYRIK.ONCESI, AYRIK.SONRASI)

# SonuÃ§larÄ± yazdÄ±rma
print("Test Ä°statistiÄŸi:", statistic)
print("P-Value:", p_value)

# AnlamlÄ±lÄ±k dÃ¼zeyi kontrolÃ¼ (Ã¶rneÄŸin, %5 anlamlÄ±lÄ±k dÃ¼zeyi)
alpha = 0.05
if p_value < alpha:
    print("Ä°ki durum arasÄ±nda anlamlÄ± bir fark vardÄ±r.")
else:
    print("Ä°ki durum arasÄ±nda anlamlÄ± bir fark yoktur.")
````
