Ä°ÅŸte p deÄŸerinin bazÄ± temel Ã¶zellikleri:

- **Null hipotezi genellikle ÅŸu ÅŸekildedir: "Ä°ki grup arasÄ±nda bir fark yoktur" veya "Bir etki yoktur."

- KÃ¼Ã§Ã¼k bir p deÄŸeri, null hipotezinin reddedilmesine ve iki grup arasÄ±nda anlamlÄ± bir farkÄ±n veya bir etkinin olduÄŸuna iÅŸaret edebilir.

  

2. **P DeÄŸeri ve AnlamlÄ±lÄ±k DÃ¼zeyi (Î±):**

- P deÄŸeri, anlamlÄ±lÄ±k dÃ¼zeyi (Î±) ile karÅŸÄ±laÅŸtÄ±rÄ±lÄ±r.

- AnlamlÄ±lÄ±k dÃ¼zeyi, bir testin yanlÄ±ÅŸ pozitif sonuÃ§ verme olasÄ±lÄ±ÄŸÄ±nÄ± kontrol eder. Tip I hatasÄ±nÄ±n kabul edilebilir seviyede tutulmasÄ± Ã¶nemlidir.

**AnlamlÄ±lÄ±k DÃ¼zeyi (Î±):**

AnlamlÄ±lÄ±k dÃ¼zeyi (alfa), hipotez testinde Ã¶nceden belirlenmiÅŸ bir eÅŸik deÄŸerdir. Genellikle 0.05 olarak seÃ§ilir, bu da %5 yanÄ±lma riskini (I. tip hata yapma riskini) kabul ettiÄŸimiz anlamÄ±na gelir.

**P-deÄŸeri ve AnlamlÄ±lÄ±k DÃ¼zeyi KarÅŸÄ±laÅŸtÄ±rmasÄ±:**

Bir hipotez testinde karar vermek iÃ§in p-deÄŸeri anlamlÄ±lÄ±k dÃ¼zeyi (Î±) ile karÅŸÄ±laÅŸtÄ±rÄ±lÄ±r:

- **p-deÄŸeri â‰¤ Î±:** SÄ±fÄ±r hipotezi reddedilir. SonuÃ§lar istatistiksel olarak anlamlÄ±dÄ±r.
- **p-deÄŸeri > Î±:** SÄ±fÄ±r hipotezi reddedilmez. SonuÃ§lar istatistiksel olarak anlamlÄ± deÄŸildir.


**Ã–rnek:**

Bir ilaÃ§ firmasÄ±, yeni bir ilacÄ±n kan basÄ±ncÄ±nÄ± dÃ¼ÅŸÃ¼rdÃ¼ÄŸÃ¼nÃ¼ iddia etmektedir. Bu iddiayÄ± test etmek iÃ§in bir deney yaparlar ve p-deÄŸerini 0.03 olarak bulurlar. AnlamlÄ±lÄ±k dÃ¼zeyini 0.05 olarak belirlediysek:

- p-deÄŸeri (0.03) < Î± (0.05) olduÄŸundan, sÄ±fÄ±r hipotezi (ilacÄ±n kan basÄ±ncÄ±nÄ± dÃ¼ÅŸÃ¼rmediÄŸi) reddedilir. Bu, ilacÄ±n kan basÄ±ncÄ±nÄ± dÃ¼ÅŸÃ¼rdÃ¼ÄŸÃ¼ne dair istatistiksel olarak anlamlÄ± kanÄ±t olduÄŸunu gÃ¶sterir.

BaÅŸka bir Ã¶rnekte, aynÄ± deneyde p-deÄŸeri 0.12 olarak bulunmuÅŸ olsaydÄ±:

- p-deÄŸeri (0.12) > Î± (0.05) olduÄŸundan, sÄ±fÄ±r hipotezi reddedilmez. Bu, ilacÄ±n kan basÄ±ncÄ±nÄ± dÃ¼ÅŸÃ¼rmediÄŸi anlamÄ±na gelmez, sadece elimizdeki verilerin bu iddiayÄ± desteklemek iÃ§in yeterli kanÄ±t saÄŸlamadÄ±ÄŸÄ± anlamÄ±na gelir.

## ğŸ“ˆ p-deÄŸeri NasÄ±l HesaplanÄ±r?

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns


# Ã–rnek veri oluÅŸturma (boy Ã¶lÃ§Ã¼mleri)
np.random.seed(42)
orneklem = np.random.normal(loc=174, scale=5, size=100)  # cm cinsinden

# Test parametreleri
mu_0 = 172  # H0 hipotezi: Ortalama boy = 172 cm
alpha = 0.05  # AnlamlÄ±lÄ±k dÃ¼zeyi
```

```python
# Betimsel istatistikler
orneklem_ort = np.mean(orneklem)
orneklem_std = np.std(orneklem, ddof=1)  # Ã–rneklem standart sapmasÄ±
n = len(orneklem)

# 1. Z-testi ile p-deÄŸeri hesaplama (n > 30 olduÄŸu iÃ§in)
z_skor = (orneklem_ort - mu_0) / (orneklem_std / np.sqrt(n))
p_deger_z = 2 * (1 - stats.norm.cdf(abs(z_skor)))  # Ä°ki yÃ¶nlÃ¼ test

# 2. t-testi ile p-deÄŸeri hesaplama (daha doÄŸru)
t_skor = (orneklem_ort - mu_0) / (orneklem_std / np.sqrt(n))
p_deger_t = 2 * (1 - stats.t.cdf(abs(t_skor), df=n-1))  # Ä°ki yÃ¶nlÃ¼ test

# 3. Scipy'nin hazÄ±r fonksiyonu ile
t_test_sonuc = stats.ttest_1samp(orneklem, popmean=mu_0)

# SonuÃ§larÄ± yazdÄ±rma
print(f"Ã–rneklem OrtalamasÄ±: {orneklem_ort:.2f} cm")
print(f"Standart Sapma: {orneklem_std:.2f} cm")
print(f"\nZ-testi:")
print(f"  Z-skoru: {z_skor:.4f}")
print(f"  p-deÄŸeri: {p_deger_z:.6f}")
print(f"\nt-testi:")
print(f"  t-skoru: {t_skor:.4f}")
print(f"  p-deÄŸeri: {p_deger_t:.6f}")
print(f"\nScipy t-test fonksiyonu:")
print(f"  t-skoru: {t_test_sonuc.statistic:.4f}")
print(f"  p-deÄŸeri: {t_test_sonuc.pvalue:.6f}")
```

```python
# Karar verme
if p_deger_t < alpha:
    print(f"\nSonuÃ§: p < {alpha} â†’ H0 reddedilir (ortalama {mu_0} cm deÄŸil)")
else:
    print(f"\nSonuÃ§: p â‰¥ {alpha} â†’ H0 reddedilemez (ortalama {mu_0} cm olabilir)")
```

```python
# GÃ¶rselleÅŸtirme
plt.figure(figsize=(12, 6))

# 1. Ã–rneklem daÄŸÄ±lÄ±mÄ±
plt.subplot(1, 2, 1)
sns.histplot(orneklem, kde=True, color='skyblue')
plt.axvline(orneklem_ort, color='red', linestyle='--', label=f'Ã–rneklem ort. ({orneklem_ort:.1f} cm)')
plt.axvline(mu_0, color='green', linestyle='-', label=f'H0 deÄŸeri ({mu_0} cm)')
plt.title('Ã–rneklem DaÄŸÄ±lÄ±mÄ±')
plt.xlabel('Boy (cm)')
plt.legend()

# 2. t-daÄŸÄ±lÄ±mÄ± ve p-deÄŸeri
plt.subplot(1, 2, 2)
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df=n-1)
plt.plot(x, y, 'b-', label=f't-daÄŸÄ±lÄ±mÄ± (df={n-1})')

# Kritik bÃ¶lge ve p-deÄŸeri
t_kritik = stats.t.ppf(1-alpha/2, df=n-1)
plt.axvline(t_kritik, color='red', linestyle=':', label=f'Kritik deÄŸer (Â±{t_kritik:.2f})')
plt.axvline(-t_kritik, color='red', linestyle=':')
plt.axvline(t_skor, color='purple', linestyle='-', label=f't-skoru ({t_skor:.2f})')

# p-deÄŸerini gÃ¶lgelendirme
if t_skor > 0:
    p_x = np.linspace(t_skor, 4, 100)
else:
    p_x = np.linspace(-4, t_skor, 100)
plt.fill_between(p_x, stats.t.pdf(p_x, df=n-1), color='purple', alpha=0.3, label='p-deÄŸeri alanÄ±')

plt.title('t-DaÄŸÄ±lÄ±mÄ± ve p-deÄŸeri')
plt.xlabel('t DeÄŸeri')
plt.ylabel('YoÄŸunluk')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

ğŸ“ˆ GÃ¶rselleÅŸtirme: p-deÄŸeri ve Red BÃ¶lgesi

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Set parameters
alpha = 0.05  # Significance level
z_skoru = 2.5  # Example z-score (replace with your calculated value)

# Generate data for standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

# Create figure with improved styling
plt.figure(figsize=(12, 6))

# Plot the standard normal curve
plt.plot(x, y, label="Standard Normal Distribution", linewidth=2, color='navy')

# Calculate critical values
kritik_deger = norm.ppf(1 - alpha/2)

# Add critical value lines
plt.axvline(kritik_deger, color='red', linestyle='--', 
           linewidth=1.5, label=f'Critical Value (Â±{kritik_deger:.2f})')
plt.axvline(-kritik_deger, color='red', linestyle='--', linewidth=1.5)

# Add z-score line
plt.axvline(z_skoru, color='green', linestyle='-', 
           linewidth=2, label=f'Test Statistic (z = {z_skoru:.2f})')

# Shade rejection regions
plt.fill_between(x, y, where=(x >= kritik_deger), 
                color='red', alpha=0.2, label='Rejection Region')
plt.fill_between(x, y, where=(x <= -kritik_deger), 
                color='red', alpha=0.2)


# p-deÄŸeri (iki yÃ¶nlÃ¼ test)
p_degeri = 2 * (1 - norm.cdf(abs(z_skoru)))
print(f"p-deÄŸeri: {p_degeri:.6f}")


# Add p-value area
if z_skoru > 0:
    p_x = np.linspace(z_skoru, 4, 100)
else:
    p_x = np.linspace(-4, z_skoru, 100)
plt.fill_between(p_x, norm.pdf(p_x), color='green', alpha=0.1, label='p-value area')

# Add decision text
decision = "Reject Hâ‚€" if abs(z_skoru) > kritik_deger else "Fail to reject Hâ‚€"
plt.text(0, 0.1, f'Decision: {decision}', ha='center', va='center',
         bbox=dict(facecolor='white', alpha=0.8), fontsize=12)

# Formatting
plt.title("Hypothesis Testing Visualization (Î± = 0.05)", pad=20, fontsize=14)
plt.xlabel("Standardized Z Score", labelpad=10)
plt.ylabel("Probability Density", labelpad=10)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout
plt.tight_layout()
plt.show()
```
