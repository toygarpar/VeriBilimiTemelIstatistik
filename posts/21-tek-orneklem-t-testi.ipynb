## 🔍 Tek Örneklem T Testi Nedir?

Tek örneklem t-testi, bir örneklemin ortalamasının,  
bilinen veya varsayılan bir popülasyon ortalamasıyla  
istatistiksel olarak farklı olup olmadığını test eder.

📌 Genellikle:
- n < 30
- σ bilinmiyor
- Dağılım normal ya da yaklaşık normal


# Tek Örneklem T Testi - One Sample T-Test






**Popülasyon ortalaması ile varsayımsal bir değer arasında istatistiksel olarak anlamlı bir farklılık olup olmadığını test etmek için kullanılan bir parametrik testtir.

Örnek ortalamasına ilişkin test yapmak için kullanılır.

Tek Örneklem T Testinde hipotezler, örneklemden elde edilen ortalama ile evren ortalaması arasında fark olup olmamasına göre oluşturulur.



## $ t = \frac{{\bar{x} - \mu}}{{s/\sqrt{n}}} $

  

## $ z = \frac{{\bar{x} - \mu}}{{σ/\sqrt{n}}} $




örnek: 10K bir ilçeden 100 kişilik bir örnek çekmiştik. Buna bir örnek oluşturur, tek örnek denir.  Ana Kitleye ilişkin bir değerlenmede bulunmak istiyorsak ki burada popülasyon ortalamasına karşılık gelir. Ana Kütle ortalamasına ilişkin teorik bir değer ile bir test yapmak istersek Tek Örneklem T Testi yapmış oluruz.



**Tek Örneklem T Testi (One-Sample t-test), istatistikte bir örneklem grubunun ortalamasının, bilinen veya hipotez edilen bir popülasyon ortalamasından anlamlı bir şekilde farklı olup olmadığını belirlemek için kullanılan 1 bir parametrik testtir. Başka bir deyişle, bir grup verinin ortalamasının belirli bir sabit değerden istatistiksel olarak farklı olup olmadığını test etmek için kullanılır.





**Tek Örneklem T Testi Ne Zaman Kullanılır?**

- Bir grup verinin ortalamasını bilinen bir değerle karşılaştırmak istediğinizde.
- Örneklem büyüklüğü küçük olduğunda (genellikle n < 30) ve popülasyon standart sapması bilinmediğinde.
- Verilerin normal dağılıma yakın olduğu varsayıldığında.





**Tek Örneklem T Testinin Varsayımları:**

- **Bağımsızlık:** Her veri noktası farklı bir bireye ait olmalıdır. Yani, aynı bireyden birden fazla ölçüm alınmamış olmalıdır.
- **Sürekli Veri:** Ölçülen değişken sürekli veri tipinde olmalıdır (örneğin, boy, kilo, sıcaklık). Nominal veya ordinal veri olamaz.
- **Normallik:** Ölçülen değişkenin popülasyonda normal dağılıma sahip olması varsayılır. Ancak, örneklem büyüklüğü yeterince büyükse (genellikle n > 30), bu varsayımın ihlali testin sonuçlarını önemli ölçüde etkilemeyebilir (Merkezi Limit Teoremi).
- **Uç Değer Yokluğu (Outlier):** Verilerde aşırı uç değerler olmamalıdır. Uç değerler testin sonuçlarını etkileyebilir.





## **Tek Örneklem T Testi Nasıl Yapılır?**

1. **Sıfır Hipotezi (H₀) ve Alternatif Hipotezi (H₁) Belirleyin:**
    
    - H₀: Örneklem ortalaması, bilinen popülasyon ortalamasına eşittir (μ = μ₀).
    - H₁: Örneklem ortalaması, bilinen popülasyon ortalamasından farklıdır (μ ≠ μ₀) (İki yönlü test). Veya, örneklem ortalaması bilinen ortalamadan büyüktür (μ > μ₀) veya küçüktür (μ < μ₀) (Tek yönlü test).
2. **Anlamlılık Düzeyini (α) Belirleyin:** Genellikle α = 0.05 kullanılır. Bu, %5 yanılma riskini kabul ettiğimiz anlamına gelir.
    
3. **T İstatistiğini Hesaplayın:**
    
    ```
    t = (x̄ - μ₀) / (s / √n)
    
	z= (x̄ - μ₀) / (σ / √n)
    ```


Bu formül, örneklem ortalaması ile popülasyon ortalaması arasındaki farkın örneklem standart sapması ve örneklem büyüklüğü ile normalize edilmiş halini ifade eder.

  

Tek örneklem t testi için bazı varsayımlar şunlardır:

  

1. **Normalite Varsayımı:** Veri seti normal dağılıma uymalıdır. Ancak, örneklemin büyüklüğü arttıkça, bu varsayımın daha az önemli hale gelebilir.

  

2. **Bağımsızlık Varsayımı:** Gözlemler birbirinden bağımsız olmalıdır. Bu, bir gözlemin diğerine olan sonuçları üzerinde herhangi bir etkisi olmaması anlamına gelir.

  

3. **Örneklemin Rassal Seçimi:** Veri seti, popülasyondan rassal olarak seçilmelidir. Bu, örneklemin popülasyonu temsil etmesini sağlar.

  

* 1. Popülasyon standard sapması biliniyorsa z istatistiği kullanılır.

* 2. Popülasyon standard sapması bilinmiyorsa ve n>30 ise z istatistiği kullanılır.

* 3. Popülasyon standard sapması biliniyor fakat n<30 ise t istatistiği kullanılır.

  
  

Eğer normalite varsayımı sağlanmıyorsa, ancak örneklem büyüklüğü büyükse, merkezi limit teoremi sayesinde t testi kullanılabilir. Ayrıca, normalite varsayımını kontrol etmek için görsel analiz veya alternatif testler de kullanılabilir.

  
  

- **Tek Örneklem T-Testi, bir örneklem grubunun ortalamasının, bir popülasyonun bilinen bir değerinden (veya tahmin edilen bir değerinden) istatistiksel olarak farklı olup olmadığını değerlendirmek için kullanılır. Bu test, özellikle bir popülasyonun genel eğilimini belirlemek için bir referans değeriyle karşılaştırmak istediğiniz durumlar için uygundur.**

- **Tek Örneklem T-Testi'nin kullanılmasının temel amacı, örneklem grubunun ortalamasının popülasyon ortalamasından istatistiksel olarak anlamlı bir şekilde farklı olup olmadığını belirlemektir. Bu, bir örneklemle elde edilen verilerin genel popülasyon için temsil edip etmediğini değerlendirmenize yardımcı olabilir.**

  

**Hipotezler:**

  

**Null Hipotezi (H0):** Sınıfın sınav notları, genel popülasyonun sınav notlarıyla aynıdır.

  

**Alternatif Hipotez (H1):** Sınıfın sınav notları, genel popülasyonun sınav notlarından istatistiksel olarak farklıdır.

  

Test sonuçları, null hipotezin reddedilip edilmediğini belirleyerek öğretmene, sınıfın genel popülasyondan istatistiksel olarak farklı bir performans sergileyip sergilemediği konusunda bir bilgi sağlar.

t = (x̄ - μ₀) / (s / √n)

z= (x̄ - μ₀) / (σ / √n)


    Burada:
    
    - x̄: Örneklem ortalaması
    - μ₀: Bilinen popülasyon ortalaması
    - s: Örneklem standart sapması
    - n: Örneklem büyüklüğü
    
4. **Serbestlik Derecesini (sd) Hesaplayın:**
    
    sd = n - 1
    
5. **T Dağılım Tablosundan veya İstatistiksel Yazılımdan Kritik T Değerini Bulun:** Belirlenen anlamlılık düzeyi (α) ve serbestlik derecesi (sd) ile tabloya bakarak kritik t değerini bulunur.
    
6. **Karar Verin:**
    
    - Eğer hesaplanan t istatistiğinin mutlak değeri, kritik t değerinden büyükse, sıfır hipotezi reddedilir. Bu, örneklem ortalamasının bilinen popülasyon ortalamasından istatistiksel olarak anlamlı bir şekilde farklı olduğu anlamına gelir.
    - Eğer hesaplanan t istatistiğinin mutlak değeri, kritik t değerinden küçükse, sıfır hipotezi reddedilmez. Bu, örneklem ortalamasının bilinen popülasyon ortalamasından farklı olmadığı anlamına gelmez, sadece elimizdeki verilerin bu farkı göstermek için yeterli kanıt sağlamadığı anlamına gelir.

## 📐 Formülü:

$$
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
$$

| Sembol | Açıklama |
|--------|----------|
| $\bar{x}$ | Örneklem ortalaması |
| $\mu_0$ | Popülasyon ortalaması (varsayımsal değer) |
| $s$ | Örneklem standart sapması |
| $n$ | Örneklem büyüklüğü |

📌 Serbestlik derecesi: `df = n - 1`

## 🧠 Neden Önemlidir?

| Sebep | Açıklama |
|--------|----------|
| ✅ **Küçük veri setleri için uygun** | n < 30 olsa bile test yapabilirsin |
| ✅ **Popülasyon standart sapması bilinmeyebilir** | O zaman t-testi tercih edilir |
| ✅ **Modelleme sürecinde kullanılır** | Özelliklerin anlamlılığını test etmede |
| ✅ **Karar verme aracıdır** | Fark gerçekten var mı? Yoksa tesadüf mü? |

📌 Özellikle klinik deneyler, küçük ölçekli anketler, ürün testleri gibi alanlarda hayati önem taşır.

## 🛠️ Adım Adım Uygulama: Gerçek Hayat Senaryosu

### Örnek:
Bir okulda öğrencilerin matematik notları inceleniyor.  
Okul genelinde ortalama not: **75**  
Senin sınıfındaki 25 öğrencinin ortalaması: **80**, std: **10**

#### Soru:
Senin sınıfının ortalaması, okul ortalamasından farklı mı?

### Adım 1: Hipotez Kurulumu

```python
print("H₀: Ortalama = 75")
print("H₁: Ortalama ≠ 75")
```

✅ Null hipotezi → fark yok  
Alternatif → fark var

Adım 2: Anlamlılık Düzeyini Belirle

````python
alpha = 0.05
print(f"Anlamlılık düzeyi (α): {alpha}")
````

📌 %5 yanılma payı kabul ettik

````python
#sample veri üret
import numpy as np

np.random.seed(42)
populasyon_ortalamasi = 75
orneklem_buyuklugu = 25
orneklem = np.random.normal(loc=80, scale=10, size=orneklem_buyuklugu)
print(f"Örneklem Ortalaması: {orneklem.mean():.2f}, Standart Sapma: {orneklem.std():.2f}")
````

Adım 4: T-Testi Yap

````python
from scipy.stats import ttest_1samp

t_skoru, p_degeri = ttest_1samp(orneklem, popmean=populasyon_ortalamasi)
print(f"T Skoru: {t_skoru:.2f}, p-değeri: {p_degeri:.4f}")
````

Adım 5: Karar Ver

````python
if p_degeri < alpha:
    print("✅ H₀ reddedildi → Sınıfınız okuldan farklı")
else:
    print("❌ H₀ reddedilemedi → Fark tesadüfi olabilir")
````

## 📐 T-Testi ile Regresyon Bağlantısı

Regresyon analizinde her katsayı bir t-testine tabidir.

📌 Modeldeki her X değişkeni için:
- H₀: β = 0 → değişkenin etkisi yok
- H₁: β ≠ 0 → değişkenin etkisi var

📌 Bu yüzden t-testi, makine öğrenmesinde de temeldir.

## **Örnek:**

Bir araştırmacı, bir şehrin ortalama IQ puanının 100 olduğunu iddia etmektedir. Rastgele seçilen 25 kişiden oluşan bir örneklemde ortalama IQ puanı 105 ve standart sapma 15 olarak bulunmuştur. Bu iddiayı test etmek için Tek Örneklem T Testi kullanalım (α = 0.05):

1. **Hipotezler:**
    
    - H₀: μ = 100
    - H₁: μ ≠ 100
2. **T İstatistiği:**
    
    ```
    t = (105 - 100) / (15 / √25) = 5 / 3 = 1.67
    ```
    
3. **Serbestlik Derecesi:**
    
    sd = 25 - 1 = 24
    
4. **Kritik T Değeri:**
    
    α = 0.05 ve sd = 24 için iki yönlü t tablosundan kritik t değeri yaklaşık olarak ±2.064 olarak bulunur.
    
5. **Karar:**
    
    Hesaplanan t istatistiğinin mutlak değeri (1.67), kritik t değerinden (2.064) küçüktür. Bu nedenle, sıfır hipotezi reddedilmez. Bu, örneklemdeki ortalama IQ puanının 105 olmasının, şehrin genel ortalama IQ puanının 100 olduğu iddiasını çürütmek için yeterli kanıt sağlamadığı anlamına gelir.

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Stil ayarları
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (14, 6)

# Veri parametreleri
n = 25               # Örneklem büyüklüğü
sample_mean = 105    # Örneklem ortalaması (IQ puanı)
sample_std = 15      # Örneklem standart sapması
pop_mean = 100       # H0 hipotezindeki popülasyon ortalaması
alpha = 0.05         # Anlamlılık düzeyi

# Simüle edilmiş veri oluşturma (normal dağılım varsayımıyla)
np.random.seed(42)
data = np.random.normal(loc=sample_mean, scale=sample_std, size=n)
````

````python
# Normallik testi (Shapiro-Wilk)
shapiro_test = stats.shapiro(data)
print("📊 NORMALLİK TESTİ SONUCU")
print("------------------------")
print(f"Shapiro-Wilk Test İstatistiği: {shapiro_test.statistic:.4f}")
print(f"p-değeri: {shapiro_test.pvalue:.4f}")

if shapiro_test.pvalue > alpha:
    print("✅ Normallik varsayımı sağlanıyor (p > 0.05)")
else:
    print("❌ Normallik varsayımı sağlanmıyor! Non-parametrik test düşünülmeli")
````

````python
# Tek örneklem t-testi
t_stat, p_value = stats.ttest_1samp(data, popmean=pop_mean)
effect_size = (sample_mean - pop_mean) / sample_std  # Cohen's d

# Kritik değerler
df = n - 1  # Serbestlik derecesi
t_critical = stats.t.ppf(1 - alpha/2, df=df)
ci = stats.t.interval(0.95, df=df, loc=sample_mean, scale=sample_std/np.sqrt(n))

# Sonuçları yazdırma
print("\n🔍 TEK ÖRNEKLEM t-TEST SONUÇLARI")
print("--------------------------------")
print(f"Örneklem Büyüklüğü (n): {n}")
print(f"Örneklem Ortalaması: {sample_mean:.2f} IQ puanı")
print(f"Örneklem Standart Sapması: {sample_std:.2f} IQ puanı")
print(f"Serbestlik Derecesi (df): {df}")
print(f"t-istatistiği: {t_stat:.4f}")
print(f"Kritik t-değeri: ±{t_critical:.4f}")
print(f"p-değeri: {p_value:.6f}")
print(f"Etki Büyüklüğü (Cohen's d): {effect_size:.3f}")
print(f"%95 Güven Aralığı: ({ci[0]:.2f}, {ci[1]:.2f})")
````

````python
# Karar ve yorum
if p_value < alpha:
    print("\n✅ SONUÇ: p < 0.05 → H₀ REDDEDİLİR")
    if sample_mean > pop_mean:
        print(f"   Örneklem ortalaması ({sample_mean}) popülasyon ortalamasından ({pop_mean}) ANLAMLI ŞEKİLDE YÜKSEK")
    else:
        print(f"   Örneklem ortalaması ({sample_mean}) popülasyon ortalamasından ({pop_mean}) ANLAMLI ŞEKİLDE DÜŞÜK")
else:
    print("\n❌ SONUÇ: p ≥ 0.05 → H₀ REDDEDİLEMEZ")
    print(f"   Örneklem ortalamasının ({sample_mean}) popülasyon ortalamasından ({pop_mean}) farklı olduğuna dair YETERLİ KANIT YOK")
````

````python
# Görselleştirme
plt.figure(figsize=(14, 6))

# 1. Veri dağılımı
plt.subplot(1, 2, 1)
sns.histplot(data, kde=True, bins=8, color='skyblue')
plt.axvline(pop_mean, color='red', linestyle='--', label=f'H₀ değeri ({pop_mean} IQ)')
plt.axvline(sample_mean, color='green', linestyle='-', label=f'Örneklem ort. ({sample_mean} IQ)')
plt.title('IQ Puanları Dağılımı (n=25)')
plt.xlabel('IQ Puanı')
plt.ylabel('Frekans')
plt.legend()

# 2. t-dağılımı ve test sonucu
plt.subplot(1, 2, 2)
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df=df)
plt.plot(x, y, label=f't-dağılımı (df={df})')

# Kritik bölgeler
plt.axvline(t_critical, color='red', linestyle=':', label=f'Kritik değer (±{t_critical:.2f})')
plt.axvline(-t_critical, color='red', linestyle=':')

# Test istatistiği
plt.axvline(t_stat, color='green', linestyle='-', label=f't-istatistiği ({t_stat:.2f})')

# p-değeri alanı
if t_stat > 0:
    p_x = np.linspace(t_stat, 4, 100)
else:
    p_x = np.linspace(-4, t_stat, 100)
plt.fill_between(p_x, stats.t.pdf(p_x, df=df), color='green', alpha=0.2, label='p-değeri alanı')

plt.title('t-Testi Sonuçları (α=0.05, iki yönlü)')
plt.xlabel('t Değeri')
plt.ylabel('Yoğunluk')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

##  İş Uygulaması: Ürün Satın Alma Optimizasyonu 

Problem:

Sepete ürün ekleme işlemi sonrasında ödeme ekranında 5 adım vardır ve bu adımların birisi sorgulanmaktadır.

Sepete ürün ekleme, fatura ve adres bilgisini girme, vb. bilgilerin girildiği 5 adımlık bir süreç var.

Detaylar:

Her adımın 20'şer saniye olması hedefi var. 4. adım sorgulanıyor. Bu durumu test etmek için 100 örnek alınıyor.  
Örnek standart sapması 5 saniyedir. Örnek ortalaması ise 19 saniyedir.


Ortaya çıkan bir etkinin şans eseri olup olmadığını, istatistiki olarak şansa yer bırakmayacak şekilde ispat etmeye çalışıyoruz.

19 saniyenin istatistiki olarak nalamlı olup olmadığını değerlendireceğiz.


**Adım 1 :**

Hipotezlerin kurulması ve yönlerinin belirlenmesi

H0 : μ = 20

H1 : μ != 20

H0: Bilinmeyen ana kütle parametresi μ yani kişilerin 4. adımda geçirdikleri süre ortalaması 20'dir.

H1 : Bilinmeyen ana kitle ortalaması μ yani kişilerin 4. adımda geçirdikleri süre ortalaması 20 saniyeden farklıdır.



**Adım 2 :** 

Anlamlılık düzeyinin ve tablo değerinin belirlenmesi


α = 0.05  α/2 = 0.025 -> hipotezimiz iki yanlı/yönlü olduğundan 2'ye bölüyoruz

Ztablo tablo olasılık değeri : 0.5 - 0.025 = 0.475  -> 0.5 ise 1 olasılık alanının dağılımının yarısı, mean ile z'nin arasında kalan bölüm, grafiğin üst tarafının toplamı 1 olacaktı , yarısını aldık

Ztablo kritik değer = -/+ 1.96


**Adım 3 :**  

Test istatistiğinin belirlenmesi ve test istatistiğinin hesaplanması   n = 100, standart sapma = 5 , örnek ortalaması 19 sn

zhesap = (19-20)/(5/sqrt(100)) = -2.00

şimdi elde ettiğimiz zhesap değerle ztablo değerini karşılaştırıp karar vereceğiz.

**Adım 4 :**

Ztablo ve Zhesap karşılaştırması Zh > Zt ya da -Zh < -Zt ise H0 Red

Zhesap = -2.00 < Ztablo = -1.96 olduğu için **H0 reddedilir.**


Eğer hipotez red bölgelerine düşmeseydi, bu durumda H0 hipotezini red etmek için yeterli kanıt bulunamadı veya istatistiki olarak anlamlı bir sonuç elde edelememiştir yorumu yapılabilirdi.

Yorum

4.adımda geçirilen sürenin 20 sn olduğunu iddia eden H0 hipotezi reddedilmiştir Buna göre kullanıcılar istatiksel olarak %95 güvenilirlik ile 4. adımda 20 saniyeden farklı zaman geçirmektedir.


Eğer H0'ı reddedemeseydik bu durumda kullanıcılarımız web sitemizde 4.adımda yaklaşık 20 sn zaman geçirdiğini kabul edecektik. 

Kullanıcı deneyimini hipotez testleri üzerine oturtturursak buradan elde edilecek kararlar tamamen bilimsel olur. Yani insan faktörü, yanlılıklar ortadan kaldırılmış olacaktır.

Normallik varsayımını bu örnekte incelemedik. Elimizde örneğin normallik varsayımını sağlayığ sağlamadığı test edilmelidir. Bu sağlanmıyorsa non parametric t testi yapılmalıdır.

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Stil ayarları
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)

# Veri parametreleri
n = 100               # Örneklem büyüklüğü
sample_mean = 19      # Örneklem ortalaması (saniye)
sample_std = 5        # Örneklem standart sapması
pop_mean = 20         # H0 hipotezindeki popülasyon ortalaması
alpha = 0.05          # Anlamlılık düzeyi

# Simüle edilmiş veri oluşturma (normal dağılım varsayımıyla)
np.random.seed(42)
data = np.random.normal(loc=sample_mean, scale=sample_std, size=n)
````

````python
# Normallik testi (Shapiro-Wilk)
shapiro_test = stats.shapiro(data)
print("📊 NORMALLİK TESTİ SONUCU")
print("------------------------")
print(f"Shapiro-Wilk Test İstatistiği: {shapiro_test.statistic:.4f}")
print(f"p-değeri: {shapiro_test.pvalue:.4f}")

if shapiro_test.pvalue > alpha:
    print("✅ Normallik varsayımı sağlanıyor (p > 0.05)")
else:
    print("❌ Normallik varsayımı sağlanmıyor! Non-parametrik test düşünülmeli")
````

````python
# Tek örneklem t-testi
t_stat, p_value = stats.ttest_1samp(data, popmean=pop_mean)
effect_size = (sample_mean - pop_mean) / sample_std  # Cohen's d

# Kritik değerler
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
ci = stats.t.interval(0.95, df=n-1, loc=sample_mean, scale=sample_std/np.sqrt(n))

# Sonuçları yazdırma
print("\n🔍 TEK ÖRNEKLEM t-TEST SONUÇLARI")
print("--------------------------------")
print(f"Örneklem Ortalaması: {sample_mean:.2f} saniye")
print(f"Örneklem Standart Sapması: {sample_std:.2f} saniye")
print(f"t-istatistiği: {t_stat:.4f}")
print(f"Kritik t-değeri: ±{t_critical:.4f}")
print(f"p-değeri: {p_value:.6f}")
print(f"Etki Büyüklüğü (Cohen's d): {effect_size:.3f}")
print(f"%95 Güven Aralığı: ({ci[0]:.2f}, {ci[1]:.2f})")
````

````python
# Karar ve yorum
if p_value < alpha:
    print("\n✅ SONUÇ: p < 0.05 → H0 REDDEDİLİR")
    if sample_mean > pop_mean:
        print("   Kullanıcılar 4. adımda ortalamadan DAHA UZUN süre geçiriyor")
    else:
        print("   Kullanıcılar 4. adımda ortalamadan DAHA KISA süre geçiriyor")
else:
    print("\n❌ SONUÇ: p ≥ 0.05 → H0 REDDEDİLEMEZ")
    print("   4. adım süresinin 20 saniyeden farklı olduğuna dair yeterli kanıt yok")
````

````python
# Görselleştirme
plt.figure(figsize=(14, 6))

# 1. Veri dağılımı
plt.subplot(1, 2, 1)
sns.histplot(data, kde=True, bins=15, color='skyblue')
plt.axvline(pop_mean, color='red', linestyle='--', label='Hipotez değeri (20 sn)')
plt.axvline(sample_mean, color='green', linestyle='-', label=f'Örneklem ort. ({sample_mean:.1f} sn)')
plt.title('Adım 4 Süre Dağılımı')
plt.xlabel('Süre (saniye)')
plt.ylabel('Frekans')
plt.legend()

# 2. t-dağılımı ve test sonucu
plt.subplot(1, 2, 2)
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df=n-1)
plt.plot(x, y, label=f't-dağılımı (df={n-1})')

# Kritik bölgeler
plt.axvline(t_critical, color='red', linestyle=':', label=f'Kritik değer (±{t_critical:.2f})')
plt.axvline(-t_critical, color='red', linestyle=':')

# Test istatistiği
plt.axvline(t_stat, color='green', linestyle='-', label=f't-istatistiği ({t_stat:.2f})')

# p-değeri alanı
if t_stat > 0:
    p_x = np.linspace(t_stat, 4, 100)
else:
    p_x = np.linspace(-4, t_stat, 100)
plt.fill_between(p_x, stats.t.pdf(p_x, df=n-1), color='green', alpha=0.2, label='p-değeri alanı')

plt.title('t-Testi Sonuçları')
plt.xlabel('t Değeri')
plt.ylabel('Yoğunluk')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

## Nonparametrik Tek Örneklem Testi

Diyelim ki varsayım sağlanmıyor.

Tek örneklem testleri, bir örneklem ile popülasyon arasındaki bir parametrenin eşitliğini test etmek için kullanılır. Eğer veri normal dağılıma uymuyorsa veya örneklem büyüklüğü küçükse, nonparametrik yöntemler tercih edilebilir. Nonparametrik yöntemlerde tek örneklem testi için yaygın olarak kullanılan bir test, Wilcoxon işaret testidir.

### Wilcoxon İşaret Testi:

Wilcoxon işaret testi, bir örneklemle popülasyon ortalaması arasında bir fark olup olmadığını kontrol etmek için kullanılan nonparametrik bir testtir. Bu test, özellikle veri normal dağılıma uymadığında veya sıralı veri türlerinde etkilidir.

Wilcoxon işaret testinin hipotezleri şu şekildedir:

- **Null Hypothesis (H0):** Popülasyon medyanı, örneklem medyanıyla eşittir.
- **Alternative Hypothesis (H1):** Popülasyon medyanı, örneklem medyanından farklıdır.

Wilcoxon işaret testi, gözlemleri sıralayarak işlem yapar ve gözlemlerin işaretli sıralarının toplamını kullanarak test istatistiğini oluşturur.

## eğitici ve gerçek hayattan bir örnek:

Akıllı şehir projesi ve bisiklet kullanım analizi

1. **Gerçekçi Senaryo**: Akıllı şehir projesi ve bisiklet kullanım analizi
    
2. **Simüle Edilmiş Veri**:

    - Gerçekçi aykırı değer içeren veri seti
        
3. **Kapsamlı Analiz**:
    
    - Betimsel istatistikler
        
    - t-testi sonuçları
        
    - Etki büyüklüğü (Cohen's d)
        
4. **Profesyonel Görseller**:
    
    - Kutu grafiği ile veri dağılımı
        
    - t-dağılımı ve test istatistiği
        
    - Güven aralığı görselleştirmesi
        
5. **Yorumlanabilir Çıktılar**:
    
    - İstatistiksel sonuçların pratik yorumu
        
    - Otomatik karar mekanizması (H0 reddedilir/reddedilemez)



````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------
# ÖZEL EĞİTİCİ ÖRNEK: 
# "Akıllı Şehir" Bisiklet Kullanım Analizi
# --------------------------

# Şehir planlamacıları, bir akıllı şehir projesi kapsamında
# vatandaşların haftalık bisiklet kullanım sürelerini araştırıyor.
# Önceki verilere göre ortalama 150 dakika olduğu biliniyor.
# Yeni altyapı yatırımları sonrası bu süre değişti mi?

np.random.seed(42)  # Tekrarlanabilirlik için

# Veri oluşturma (Simüle edilmiş gerçekçi veri)
n = 30  # 30 katılımcı
eski_ortalama = 150  # H0: μ = 150 dakika
yeni_veri = np.round(np.concatenate([
    np.random.normal(loc=180, scale=30, size=n//2),  # Düzenli bisikletçiler
    np.random.normal(loc=120, scale=40, size=n//2)   # Yeni başlayanlar
]))

# Aykırı değer ekleme (gerçek hayat verisi gibi)
yeni_veri[0] = 300  # Bir bisiklet tutkunu
````

````python
# --------------------------
# BETİMSEL İSTATİSTİKLER
# --------------------------
print("📊 BETİMSEL İSTATİSTİKLER")
print("-------------------------")
print(f"Örneklem Büyüklüğü: {n}")
print(f"Ortalama: {np.mean(yeni_veri):.1f} dakika")
print(f"Standart Sapma: {np.std(yeni_veri, ddof=1):.1f} dakika")
print(f"Medyan: {np.median(yeni_veri):.1f} dakika")
print(f"Minimum-Maksimum: {np.min(yeni_veri)}-{np.max(yeni_veri)} dakika")
````

````python
# --------------------------
# TEK ÖRNEKLEM t-TESTİ
# --------------------------
t_stat, p_value = stats.ttest_1samp(yeni_veri, popmean=eski_ortalama)
effect_size = (np.mean(yeni_veri) - eski_ortalama) / np.std(yeni_veri, ddof=1)

print("\n🔍 TEK ÖRNEKLEM t-TEST SONUÇLARI")
print("--------------------------------")
print(f"t-istatistiği: {t_stat:.3f}")
print(f"p-değeri: {p_value:.5f}")
print(f"Etki Büyüklüğü (Cohen's d): {effect_size:.3f}")

if p_value < 0.05:
    print("\n✅ SONUÇ: p < 0.05 → H0 REDDEDİLİR")
    if np.mean(yeni_veri) > eski_ortalama:
        print("   Yeni altyapı sonrası bisiklet kullanım süresi ARTMIŞTIR")
    else:
        print("   Yeni altyapı sonrası bisiklet kullanım süresi AZALMIŞTIR")
else:
    print("\n❌ SONUÇ: p ≥ 0.05 → H0 REDDEDİLEMEZ")
    print("   Bisiklet kullanım süresinde anlamlı değişiklik YOKTUR")
````

````python
# --------------------------
# GÖRSELLEŞTİRMELER
# --------------------------
plt.figure(figsize=(15, 5))

# 1. Veri Dağılımı
plt.subplot(1, 3, 1)
sns.boxplot(x=yeni_veri, color='skyblue')
plt.axvline(eski_ortalama, color='red', linestyle='--', label='Eski ortalama (150 dk)')
plt.title('Haftalık Bisiklet Kullanım Süreleri')
plt.xlabel('Dakika')
plt.legend()

# 2. t-Dağılımı ve Test İstatistiği
plt.subplot(1, 3, 2)
df = n - 1
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df)
plt.plot(x, y, label=f't-dağılımı (df={df})')
plt.axvline(t_stat, color='green', linestyle='-', label=f't-istatistiği ({t_stat:.2f})')

# Kritik bölge ve p-değeri
t_critical = stats.t.ppf(1-0.025, df)
plt.axvline(t_critical, color='red', linestyle=':', label=f'Kritik değer (±{t_critical:.2f})')
plt.axvline(-t_critical, color='red', linestyle=':')

# p-değeri alanını gölgelendirme
if t_stat > 0:
    p_x = np.linspace(t_stat, 4, 100)
else:
    p_x = np.linspace(-4, t_stat, 100)
plt.fill_between(p_x, stats.t.pdf(p_x, df), color='green', alpha=0.2, label='p-değeri alanı')

plt.title('t-Testi Sonuçları')
plt.xlabel('t Değeri')
plt.ylabel('Yoğunluk')
plt.legend()

# 3. Güven Aralığı
plt.subplot(1, 3, 3)
ci = stats.t.interval(0.95, df, loc=np.mean(yeni_veri), scale=stats.sem(yeni_veri))
plt.errorbar(x=0, y=np.mean(yeni_veri), 
             yerr=[[np.mean(yeni_veri)-ci[0]], [ci[1]-np.mean(yeni_veri)]], 
             fmt='o', color='green', markersize=10, capsize=5)
plt.axhline(eski_ortalama, color='red', linestyle='--')
plt.xlim(-1, 1)
plt.title('%95 Güven Aralığı')
plt.ylabel('Ortalama Süre (dakika)')
plt.xticks([])
plt.grid(axis='y')

plt.tight_layout()
plt.show()
````
