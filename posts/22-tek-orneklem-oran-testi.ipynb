### **Ã–rnek:**

Ã–rneÄŸin, bir Ã¼niversite mezunlarÄ±nÄ±n iÅŸ bulma oranÄ± hakkÄ±nda bir popÃ¼lasyon tahmini yapmak istediÄŸinizi dÃ¼ÅŸÃ¼nelim. VarsayalÄ±m ki, genel popÃ¼lasyonun iÅŸ bulma oranÄ±nÄ±n %75 olduÄŸunu dÃ¼ÅŸÃ¼nÃ¼yorsunuz. Ancak, sadece belirli bir Ã¶rnekleme dayalÄ± olarak elde ettiÄŸiniz iÅŸ bulma oranÄ±nÄ±n genel popÃ¼lasyondan istatistiksel olarak farklÄ± olup olmadÄ±ÄŸÄ±nÄ± test etmek istiyorsunuz.

  

Bu durumu deÄŸerlendirmek iÃ§in Tek Ã–rneklem Oran Testi ÅŸu adÄ±mlarÄ± takip eder:

  

1. **Hipotez Kurma:**

- **Null Hipotezi (H0):** Ã–rneklem oranÄ±, popÃ¼lasyon oranÄ± ile aynÄ±dÄ±r.

- **Alternatif Hipotez (H1):** Ã–rneklem oranÄ±, popÃ¼lasyon oranÄ±ndan istatistiksel olarak farklÄ±dÄ±r.

  

2. **Veri Toplama:**

- Belirli bir Ã¶rneklemden ilgili kategorik Ã¶zellik (Ã¶rneÄŸin, iÅŸ bulma durumu) verilerini topla.

Oransal bir ifade test edilmek istediÄŸinde kullanÄ±lÄ±r.

  

Tek Ã–rneklem Oran Testi'nin formÃ¼lÃ¼ ÅŸu ÅŸekildedir:

  
  
  

## $z = \frac{\hat{p} - p_0}{\sqrt{\frac{\hat{p}(1-\hat{p})}{n}}}$

  

Bu formÃ¼lde:

  

- $z$: Test istatistiÄŸini temsil eder.

- $p$: Ã–rneklem oranÄ±nÄ± temsil eder.

- $p0$: PopÃ¼lasyon oranÄ±nÄ± temsil eder (null hipotezde belirtilen deÄŸer).

- $n$: Ã–rneklem bÃ¼yÃ¼klÃ¼ÄŸÃ¼nÃ¼ temsil eder.

  

4. **P DeÄŸerini Hesaplama:**

- Hesaplanan z-istatistiÄŸi Ã¼zerinden p deÄŸerini belirle.

  

5. **Karar Verme:**

- Belirlenen anlamlÄ±lÄ±k dÃ¼zeyi (Ã¶rneÄŸin, \(\alpha = 0.05\)) Ã¼zerinden p deÄŸerini karÅŸÄ±laÅŸtÄ±r ve null hipotezini reddet veya reddetmeme kararÄ± ver.

  

Tek Ã–rneklem Oran Testi, Ã¶zellikle popÃ¼lasyon oranÄ±na iliÅŸkin bir Ã¶nerme Ã¼zerinde istatistiksel Ã§Ä±karÄ±mlar yapmak istediÄŸiniz durumlarda kullanÄ±lÄ±r.

  

Bu formÃ¼l, Ã¶rneklem oranÄ±nÄ±n popÃ¼lasyon oranÄ±ndan ne kadar farklÄ± olduÄŸunu Ã¶lÃ§en bir z-skoru Ã¼retir. Bu z-skorunu kullanarak, hipotez testi yapÄ±lÄ±r ve null hipotezi reddedilip edilmeyeceÄŸi belirlenir.>

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Stil ayarlarÄ±

sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)

def tek_oran_testi(sample_success, sample_size, pop_prop, alpha=0.05, alternative='two-sided'):
    """
    Tek Ã–rneklem Oran Testi UygulamasÄ±
    
    Parametreler:
        sample_success (int): Ã–rneklemdeki baÅŸarÄ± sayÄ±sÄ±
        sample_size (int): Ã–rneklem bÃ¼yÃ¼klÃ¼ÄŸÃ¼
        pop_prop (float): PopÃ¼lasyon oranÄ± (H0 deÄŸeri)
        alpha (float): AnlamlÄ±lÄ±k dÃ¼zeyi
        alternative (str): 'two-sided', 'greater', 'less'
    """
    # Ã–rneklem oranÄ±
    sample_prop = sample_success / sample_size
    
    # Test istatistiÄŸi (z-score)
    z = (sample_prop - pop_prop) / np.sqrt((pop_prop * (1 - pop_prop)) / sample_size)
    
    # p-deÄŸeri hesaplama
    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(abs(z)))
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z)
    else:  # 'less'
        p_value = stats.norm.cdf(z)
    
    # Kritik deÄŸerler
    if alternative == 'two-sided':
        critical_value = stats.norm.ppf(1 - alpha/2)
    else:
        critical_value = stats.norm.ppf(1 - alpha)
    
    # GÃ¼ven aralÄ±ÄŸÄ± (Wald yÃ¶ntemi)
    margin_of_error = stats.norm.ppf(1 - alpha/2) * np.sqrt((sample_prop * (1 - sample_prop)) / sample_size)
    ci_lower = sample_prop - margin_of_error
    ci_upper = sample_prop + margin_of_error
    
    # SonuÃ§larÄ± yazdÄ±rma
    print("ğŸ“Š TEK Ã–RNEKLEM ORAN TESTÄ° SONUÃ‡LARI")
    print("-----------------------------------")
    print(f"Ã–rneklem BaÅŸarÄ± SayÄ±sÄ±: {sample_success}")
    print(f"Ã–rneklem BÃ¼yÃ¼klÃ¼ÄŸÃ¼ (n): {sample_size}")
    print(f"Ã–rneklem OranÄ± (pÌ‚): {sample_prop:.4f}")
    print(f"PopÃ¼lasyon OranÄ± (pâ‚€): {pop_prop:.4f}")
    print(f"Test Ä°statistiÄŸi (z): {z:.4f}")
    print(f"Kritik z-deÄŸeri: Â±{critical_value:.4f}" if alternative == 'two-sided' else f"Kritik z-deÄŸeri: {critical_value:.4f}")
    print(f"p-deÄŸeri: {p_value:.6f}")
    print(f"%{(1-alpha)*100} GÃ¼ven AralÄ±ÄŸÄ±: ({ci_lower:.4f}, {ci_upper:.4f})")
    
    # Karar ve yorum
    print("\nğŸ” YORUM:")
    if p_value < alpha:
        print(f"âœ… p < {alpha} â†’ Hâ‚€ REDDEDÄ°LÄ°R")
        if alternative == 'two-sided':
            if sample_prop > pop_prop:
                print(f"   Ã–rneklem oranÄ± ({sample_prop:.4f}), popÃ¼lasyon oranÄ±ndan ({pop_prop:.4f}) ANLAMLI ÅEKÄ°LDE YÃœKSEK")
            else:
                print(f"   Ã–rneklem oranÄ± ({sample_prop:.4f}), popÃ¼lasyon oranÄ±ndan ({pop_prop:.4f}) ANLAMLI ÅEKÄ°LDE DÃœÅÃœK")
        elif alternative == 'greater':
            print(f"   Ã–rneklem oranÄ± ({sample_prop:.4f}), popÃ¼lasyon oranÄ±ndan ({pop_prop:.4f}) ANLAMLI ÅEKÄ°LDE YÃœKSEK")
        else:
            print(f"   Ã–rneklem oranÄ± ({sample_prop:.4f}), popÃ¼lasyon oranÄ±ndan ({pop_prop:.4f}) ANLAMLI ÅEKÄ°LDE DÃœÅÃœK")
    else:
        print(f"âŒ p â‰¥ {alpha} â†’ Hâ‚€ REDDEDÄ°LEMEZ")
        print(f"   Ã–rneklem oranÄ±nÄ±n ({sample_prop:.4f}) popÃ¼lasyon oranÄ±ndan ({pop_prop:.4f}) farklÄ± olduÄŸuna dair YETERLÄ° KANIT YOK")
    
    # GÃ¶rselleÅŸtirme
    plt.figure(figsize=(12, 6))
    
    # 1. OranlarÄ±n karÅŸÄ±laÅŸtÄ±rmasÄ±
    plt.subplot(1, 2, 1)
    plt.bar(['PopÃ¼lasyon (pâ‚€)', 'Ã–rneklem (pÌ‚)'], [pop_prop, sample_prop], color=['red', 'green'], alpha=0.7)
    plt.errorbar(x=1, y=sample_prop, yerr=margin_of_error, fmt='o', color='black', 
                 capsize=5, label='GÃ¼ven AralÄ±ÄŸÄ±')
    plt.title('Oran KarÅŸÄ±laÅŸtÄ±rmasÄ±')
    plt.ylabel('Oran')
    plt.ylim(0, 1)
    plt.legend()
    
    # 2. z-daÄŸÄ±lÄ±mÄ± ve test sonucu
    plt.subplot(1, 2, 2)
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)
    plt.plot(x, y, label='Standart Normal DaÄŸÄ±lÄ±m')
    
    # Kritik bÃ¶lgeler
    if alternative == 'two-sided':
        plt.axvline(critical_value, color='red', linestyle=':', label=f'Kritik deÄŸer (Â±{critical_value:.2f})')
        plt.axvline(-critical_value, color='red', linestyle=':')
        plt.fill_between(x, y, where=(x >= critical_value), color='red', alpha=0.2, label='Red bÃ¶lgesi')
        plt.fill_between(x, y, where=(x <= -critical_value), color='red', alpha=0.2)
    elif alternative == 'greater':
        plt.axvline(critical_value, color='red', linestyle=':', label=f'Kritik deÄŸer ({critical_value:.2f})')
        plt.fill_between(x, y, where=(x >= critical_value), color='red', alpha=0.2, label='Red bÃ¶lgesi')
    else:
        plt.axvline(-critical_value, color='red', linestyle=':', label=f'Kritik deÄŸer (-{critical_value:.2f})')
        plt.fill_between(x, y, where=(x <= -critical_value), color='red', alpha=0.2, label='Red bÃ¶lgesi')
    
    # Test istatistiÄŸi
    plt.axvline(z, color='green', linestyle='-', label=f'z-istatistiÄŸi ({z:.2f})')
    
    plt.title('z-Testi SonuÃ§larÄ±')
    plt.xlabel('z DeÄŸeri')
    plt.ylabel('YoÄŸunluk')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

````

````python
# Ãœniversite mezunlarÄ± iÅŸ bulma oranÄ± Ã¶rneÄŸi
print("ğŸ“ ÃœNÄ°VERSÄ°TE MEZUNLARI Ä°Å BULMA ORANI ANALÄ°ZÄ°")
tek_oran_testi(sample_success=85, sample_size=100, pop_prop=0.75, alpha=0.05, alternative='two-sided')
````

## Ä°ÅŸ UygulamasÄ± : DÃ¶nÃ¼ÅŸÃ¼m Oran Testi

Problem: 

Bir yazÄ±lÄ±m ile bir mecrada reklam verilmiÅŸ ve bu reklama iliÅŸkin yazÄ±lÄ±m tarafÄ±ndan 0.125 dÃ¶nÃ¼ÅŸÃ¼m oranÄ± elde edildiÄŸi ifade edilmiÅŸ. Fakat bu durum kontrol edilmek isteniyor. Ã‡Ã¼nkÃ¼ bu yÃ¼ksek bir oran ve gelirler incelendiÄŸinde Ã¶rtÃ¼ÅŸmÃ¼yor.


Detaylar: 

500 kiÅŸi dÄ±ÅŸ mecrada reklamlara tÄ±klamÄ±ÅŸ, 40 tanesini sitemize gelip alÄ±ÅŸveriÅŸ yapmÄ±ÅŸ.

Ã–rnek Ã¼zerinden elde edilen dÃ¶nÃ¼ÅŸÃ¼m oranÄ±: 40/500 = 0.08




## VarsayÄ±mlar

- **n>30 olmalÄ± p^ :dÃ¶nÃ¼ÅŸÃ¼m oranÄ± P0 test edilmek istenen deÄŸer**

- **HO: p=0.125**
- **H1: p!=0.0125**

````python
from statsmodels.stats.proportion import proportions_ztest #iki oran arasÄ±nda karÅŸÄ±laÅŸtÄ±rma yapmak iÃ§in gerekli modÃ¼lÃ¼ import edelim
````

````python
count = 40 #baÅŸarÄ± sayÄ±sÄ±
nobs = 500 #gÃ¶zlem sayÄ±sÄ±
value = 0.125 #karÅŸÄ±laÅŸtÄ±rÄ±lacak deÄŸer
````

````python
proportions_ztest(count, nobs, value) #iki oran arasÄ±nda karÅŸÄ±laÅŸtÄ±rma yapmak iÃ§in bu fonksiyonu kullanÄ±yoruz, bu fonksiyonun iÃ§erisine baÅŸarÄ± sayÄ±sÄ±, gÃ¶zlem sayÄ±sÄ± ve karÅŸÄ±laÅŸtÄ±rÄ±lacak deÄŸeri giriyoruz. sol taraftaki deÄŸer test istatistiÄŸini, saÄŸ taraftaki deÄŸer ise p-value deÄŸerini verir  
````

P-value deÄŸeri < alfa 0.05 deÄŸerinden daha kÃ¼Ã§Ã¼k olduÄŸundan dolayÄ± H0 hipotezini reddediyoruz.

ÃœrÃ¼n satÄ±n alma oranÄ±na iliÅŸkin dÃ¶nÃ¼ÅŸÃ¼m oranÄ± 0.125 deÄŸerinin yanlÄ±ÅŸ olduÄŸunu istatistiksel olarak %95 gÃ¼ven ile sÃ¶yleme imkanÄ± yakalamÄ±ÅŸ oluyoruz.



- **z-test degeri**: Bu deÄŸer, oranÄ±n popÃ¼lasyon oranÄ±ndan kaÃ§ standart sapma uzaklÄ±kta olduÄŸunu ifade eder. Pozitif bir deÄŸer, Ã¶rnek oranÄ±nÄ±n popÃ¼lasyon oranÄ±ndan yÃ¼ksek olduÄŸunu gÃ¶sterir, negatif bir deÄŸer ise dÃ¼ÅŸÃ¼k olduÄŸunu gÃ¶sterir.

- **p-value degeri**: Bu deÄŸer, null hipotezinin reddedilip reddedilmeyeceÄŸini belirlemek iÃ§in kullanÄ±lÄ±r. EÄŸer p-deÄŸeri belirli bir anlamlÄ±lÄ±k dÃ¼zeyinden (genellikle 0.05) kÃ¼Ã§Ã¼kse, null hipotezi reddedilir ve oranÄ±n popÃ¼lasyon oranÄ±ndan farklÄ± olduÄŸu sÃ¶ylenir.

- Bu Ã¶rnekte, p-deÄŸeri test ettiÄŸiniz oranÄ±n belirlediÄŸiniz popÃ¼lasyon oranÄ±ndan istatistiksel olarak farklÄ± olup olmadÄ±ÄŸÄ±nÄ± gÃ¶sterir. EÄŸer p-deÄŸeri anlamlÄ±lÄ±k dÃ¼zeyinden kÃ¼Ã§Ã¼kse, bu durumu destekleyecek yeterli istatistiksel kanÄ±tlar vardÄ±r ve null hipotezi reddedilir.

````python
donusum = proportions_ztest(count,nobs,value)

print('z test degeri : '+str(donusum[0]))
print('p-value degeri : '+str(donusum[1]))
````
