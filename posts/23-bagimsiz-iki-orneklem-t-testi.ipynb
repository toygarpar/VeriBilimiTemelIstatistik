### AdÄ±m 2: AnlamlÄ±lÄ±k DÃ¼zeyini Ayarla

### AdÄ±m 3: Sample Veri Ãœret

````python
import numpy as np

np.random.seed(42)

grup_A = np.random.normal(loc=7.8, scale=1.5, size=50)  # TasarÄ±m A
grup_B = np.random.normal(loc=6.2, scale=1.5, size=50)  # TasarÄ±m B

print("Grup A OrtalamasÄ±:", grup_A.mean().round(2))
print("Grup B OrtalamasÄ±:", grup_B.mean().round(2))
````

### AdÄ±m 4: Levene Testi ile Varyans KontrolÃ¼

````python
from scipy.stats import levene

stat, p = levene(grup_A, grup_B)
print(f"Levene Testi p-deÄŸeri: {p:.4f}")

if p > alpha:
    print("âœ… EÅŸit varyans varsayÄ±mÄ± saÄŸlanÄ±yor")
else:
    print("âŒ Varyanslar eÅŸit deÄŸil â†’ Welchâ€™s t-testi yapÄ±lacak")
````

### AdÄ±m 5: T-Testi Yap

````python
from scipy.stats import ttest_ind

t_skoru, p_degeri = ttest_ind(grup_A, grup_B)
print(f"T Skoru: {t_skoru:.2f}, p-deÄŸeri: {p_degeri:.4f}")
````


### AdÄ±m 6: Karar Ver

````python
if p_degeri < alpha:
    print("âœ… Hâ‚€ reddedildi â†’ TasarÄ±mlar arasÄ±nda fark var")
else:
    print("âŒ Hâ‚€ reddedilemedi â†’ Fark tesadÃ¼fi olabilir")
````

### **Ã–rnek:**

Bu Ã¶rneÄŸin detaylÄ± hesaplamasÄ±, varyanslarÄ±n homojenliÄŸi testinin yapÄ±lmasÄ± ve doÄŸru t testinin seÃ§ilmesini gerektirir. Bu hesaplamalar genellikle istatistiksel yazÄ±lÄ±mlar (SPSS, R, Python vb.) ile yapÄ±lÄ±r.

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Stil ayarlarÄ±

sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (14, 6)

def bagimsiz_iki_orneklem_t_testi(grup1, grup2, alpha=0.05):
    """
    BaÄŸÄ±msÄ±z Ä°ki Ã–rneklem t-Testi UygulamasÄ±
    
    Parametreler:
        grup1 (array): Birinci gruba ait veriler
        grup2 (array): Ä°kinci gruba ait veriler
        alpha (float): AnlamlÄ±lÄ±k dÃ¼zeyi
    """
    # Betimsel istatistikler
    n1, n2 = len(grup1), len(grup2)
    mean1, mean2 = np.mean(grup1), np.mean(grup2)
    std1, std2 = np.std(grup1, ddof=1), np.std(grup2, ddof=1)
    
    # Varyans homojenliÄŸi testi (Levene)
    levene_test = stats.levene(grup1, grup2)
    
    # Varyanslara gÃ¶re uygun t-testi seÃ§imi
    if levene_test.pvalue > alpha:
        # Varyanslar homojen
        t_stat, p_value = stats.ttest_ind(grup1, grup2, equal_var=True)
        test_turu = "Varyanslar homojen (Student t-testi)"
    else:
        # Varyanslar heterojen
        t_stat, p_value = stats.ttest_ind(grup1, grup2, equal_var=False)
        test_turu = "Varyanslar heterojen (Welch t-testi)"
    
    # Etki bÃ¼yÃ¼klÃ¼ÄŸÃ¼ (Cohen's d)
    pooled_std = np.sqrt(((n1-1)*std1**2 + (n2-1)*std2**2) / (n1+n2-2))
    cohen_d = (mean1 - mean2) / pooled_std
    
    # GÃ¼ven aralÄ±ÄŸÄ±
    diff_mean = mean1 - mean2
    margin_of_error = stats.t.ppf(1 - alpha/2, df=n1+n2-2) * np.sqrt(std1**2/n1 + std2**2/n2)
    ci_lower = diff_mean - margin_of_error
    ci_upper = diff_mean + margin_of_error
    
    # SonuÃ§larÄ± yazdÄ±rma
    print("ğŸ“Š BAÄIMSIZ Ä°KÄ° Ã–RNEKLEM t-TESTÄ° SONUÃ‡LARI")
    print("----------------------------------------")
    print(f"Grup 1 (n={n1}): Ortalama = {mean1:.2f}, Standart Sapma = {std1:.2f}")
    print(f"Grup 2 (n={n2}): Ortalama = {mean2:.2f}, Standart Sapma = {std2:.2f}")
    print(f"\nVaryans HomojenliÄŸi Testi (Levene):")
    print(f"  F-istatistiÄŸi = {levene_test.statistic:.4f}, p-deÄŸeri = {levene_test.pvalue:.4f}")
    print(f"  SonuÃ§: {test_turu}")
    print(f"\nt-istatistiÄŸi: {t_stat:.4f}")
    print(f"p-deÄŸeri: {p_value:.6f}")
    print(f"Etki BÃ¼yÃ¼klÃ¼ÄŸÃ¼ (Cohen's d): {cohen_d:.3f}")
    print(f"%95 GÃ¼ven AralÄ±ÄŸÄ± (Ortalama Fark): ({ci_lower:.2f}, {ci_upper:.2f})")
    
    # Karar ve yorum
    print("\nğŸ” YORUM:")
    if p_value < alpha:
        print(f"âœ… p < {alpha} â†’ Hâ‚€ REDDEDÄ°LÄ°R (Gruplar arasÄ±nda anlamlÄ± fark var)")
        if mean1 > mean2:
            print(f"   Grup 1 ortalamasÄ± ({mean1:.2f}), Grup 2 ortalamasÄ±ndan ({mean2:.2f}) ANLAMLI ÅEKÄ°LDE YÃœKSEK")
        else:
            print(f"   Grup 2 ortalamasÄ± ({mean2:.2f}), Grup 1 ortalamasÄ±ndan ({mean1:.2f}) ANLAMLI ÅEKÄ°LDE YÃœKSEK")
    else:
        print(f"âŒ p â‰¥ {alpha} â†’ Hâ‚€ REDDEDÄ°LEMEZ (Gruplar arasÄ±nda anlamlÄ± fark YOK)")
    
    # GÃ¶rselleÅŸtirme
    plt.figure(figsize=(14, 6))
    
    # 1. Kutu grafiÄŸi
    plt.subplot(1, 2, 1)
    sns.boxplot(data=[grup1, grup2], palette=['skyblue', 'lightgreen'])
    plt.xticks([0, 1], ['Grup 1 (Geleneksel)', 'Grup 2 (Yeni)'])
    plt.title('GruplarÄ±n SÄ±nav Puan DaÄŸÄ±lÄ±mÄ±')
    plt.ylabel('Puan')
    
    # 2. t-daÄŸÄ±lÄ±mÄ± ve test sonucu
    plt.subplot(1, 2, 2)
    df = n1 + n2 - 2  # Serbestlik derecesi
    x = np.linspace(-4, 4, 1000)
    y = stats.t.pdf(x, df)
    plt.plot(x, y, label=f't-daÄŸÄ±lÄ±mÄ± (df={df})')
    
    # Kritik bÃ¶lgeler
    t_critical = stats.t.ppf(1 - alpha/2, df)
    plt.axvline(t_critical, color='red', linestyle=':', label=f'Kritik deÄŸer (Â±{t_critical:.2f})')
    plt.axvline(-t_critical, color='red', linestyle=':')
    plt.fill_between(x, y, where=(x >= t_critical), color='red', alpha=0.2, label='Red bÃ¶lgesi')
    plt.fill_between(x, y, where=(x <= -t_critical), color='red', alpha=0.2)
    
    # Test istatistiÄŸi
    plt.axvline(t_stat, color='green', linestyle='-', label=f't-istatistiÄŸi ({t_stat:.2f})')
    
    plt.title('t-Testi SonuÃ§larÄ±')
    plt.xlabel('t DeÄŸeri')
    plt.ylabel('YoÄŸunluk')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

````

````python
# Ã–rnek veri oluÅŸturma (verilen Ã¶rnek iÃ§in)
np.random.seed(42)
grup1 = np.random.normal(loc=75, scale=10, size=20)  # Geleneksel yÃ¶ntem
grup2 = np.random.normal(loc=80, scale=8, size=20)   # Yeni yÃ¶ntem

# Testi Ã§alÄ±ÅŸtÄ±rma
print("ğŸ“ Ã–ÄRETÄ°M YÃ–NTEMLERÄ°NÄ°N KARÅILAÅTIRILMASI")
bagimsiz_iki_orneklem_t_testi(grup1, grup2, alpha=0.05)
````

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown

# Stil ayarlarÄ±

sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (14, 6)

def reklam_stratejisi_testi(strateji_a, strateji_b, alpha=0.05):
    """
    Ä°ki reklam stratejisinin performansÄ±nÄ± karÅŸÄ±laÅŸtÄ±ran t-testi
    
    Parametreler:
        strateji_a (array): Strateji A'nÄ±n satÄ±ÅŸ verileri
        strateji_b (array): Strateji B'nin satÄ±ÅŸ verileri
        alpha (float): AnlamlÄ±lÄ±k dÃ¼zeyi
    """
    # Betimsel istatistikler
    n_a, n_b = len(strateji_a), len(strateji_b)
    mean_a, mean_b = np.mean(strateji_a), np.mean(strateji_b)
    std_a, std_b = np.std(strateji_a, ddof=1), np.std(strateji_b, ddof=1)
    
    # Varyans homojenliÄŸi testi (Levene)
    levene_test = stats.levene(strateji_a, strateji_b)
    
    # Varyanslara gÃ¶re uygun t-testi seÃ§imi
    if levene_test.pvalue > alpha:
        t_stat, p_value = stats.ttest_ind(strateji_a, strateji_b, equal_var=True)
        test_turu = "Varyanslar homojen (Student t-testi)"
    else:
        t_stat, p_value = stats.ttest_ind(strateji_a, strateji_b, equal_var=False)
        test_turu = "Varyanslar heterojen (Welch t-testi)"
    
    # Etki bÃ¼yÃ¼klÃ¼ÄŸÃ¼ (Cohen's d)
    pooled_std = np.sqrt(((n_a-1)*std_a**2 + (n_b-1)*std_b**2) / (n_a+n_b-2))
    cohen_d = (mean_a - mean_b) / pooled_std
    
    # GÃ¼ven aralÄ±ÄŸÄ±
    diff_mean = mean_a - mean_b
    margin_of_error = stats.t.ppf(1 - alpha/2, df=n_a+n_b-2) * np.sqrt(std_a**2/n_a + std_b**2/n_b)
    ci_lower = diff_mean - margin_of_error
    ci_upper = diff_mean + margin_of_error
    
    # SonuÃ§ raporu
    display(Markdown(f"""
    ### ğŸ“Š Reklam Stratejileri KarÅŸÄ±laÅŸtÄ±rma SonuÃ§larÄ±
    
    | Ã–zellik | Strateji A | Strateji B |
    |---------|------------|------------|
    | Ã–rneklem BÃ¼yÃ¼klÃ¼ÄŸÃ¼ | {n_a} | {n_b} |
    | Ortalama SatÄ±ÅŸ | {mean_a:.2f} | {mean_b:.2f} |
    | Standart Sapma | {std_a:.2f} | {std_b:.2f} |
    
    **Varyans HomojenliÄŸi Testi:**  
    Levene F = {levene_test.statistic:.3f}, p = {levene_test.pvalue:.3f} â†’ *{test_turu}*
    
    **t-Test SonuÃ§larÄ±:**  
    t({n_a+n_b-2}) = {t_stat:.3f}, p = {p_value:.4f}  
    Cohen's d = {cohen_d:.3f}  
    %95 GA = [{ci_lower:.2f}, {ci_upper:.2f}]
    """))
    
    # Karar ve yorum
    if p_value < alpha:
        if mean_a > mean_b:
            sonuc = f"âœ… Strateji A, Strateji B'den anlamlÄ± derecede BAÅARILI (p = {p_value:.4f})"
        else:
            sonuc = f"âœ… Strateji B, Strateji A'dan anlamlÄ± derecede BAÅARILI (p = {p_value:.4f})"
    else:
        sonuc = f"âŒ Ä°ki strateji arasÄ±nda anlamlÄ± fark YOK (p = {p_value:.4f})"
    
    display(Markdown(f"**SonuÃ§:** {sonuc}"))
    
    # GÃ¶rselleÅŸtirme
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    
    # 1. Kutu grafiÄŸi ve swarm plot
    sns.boxplot(data=[strateji_a, strateji_b], ax=ax1, width=0.5,
                palette=["skyblue", "lightgreen"])
    sns.swarmplot(data=[strateji_a, strateji_b], ax=ax1,
                  palette=["darkblue", "darkgreen"], size=5)
    ax1.set_xticks([0, 1], ['Strateji A', 'Strateji B'])
    ax1.set_ylabel('SatÄ±ÅŸ MiktarÄ±')
    ax1.set_title('SatÄ±ÅŸ DaÄŸÄ±lÄ±mlarÄ±')
    
    # 2. Etki bÃ¼yÃ¼klÃ¼ÄŸÃ¼ gÃ¶rselleÅŸtirme
    ax2.barh(['Ortalama Fark'], [diff_mean], color='salmon')
    ax2.errorbar(x=[diff_mean], y=[0], xerr=[margin_of_error], 
                 fmt='o', color='darkred', capsize=5)
    ax2.axvline(0, color='gray', linestyle='--')
    ax2.set_title('Ortalama Fark ve %95 GÃ¼ven AralÄ±ÄŸÄ±')
    ax2.set_xlabel('Strateji A - Strateji B')
    ax2.set_yticks([])
    
    plt.tight_layout()
    plt.show()

# SimÃ¼le edilmiÅŸ reklam verileri (Ã¶rnek)
np.random.seed(42)
strateji_a = np.random.normal(loc=120, scale=15, size=30).astype(int)  # Strateji A
strateji_b = np.random.normal(loc=130, scale=20, size=35).astype(int)  # Strateji B

# Testi Ã§alÄ±ÅŸtÄ±rma
reklam_stratejisi_testi(strateji_a, strateji_b)
````

2. **ÃœrÃ¼n FiyatlandÄ±rma Analizi:**

- Problem: Bir e-ticaret ÅŸirketi, iki farklÄ± fiyatlandÄ±rma stratejisini test etmek istiyor.

- Gruplar: Ä°lk grup, YÃ¼ksek FiyatlÄ± ÃœrÃ¼nÃ¼ gÃ¶ren kiÅŸiler, ikinci grup ise DÃ¼ÅŸÃ¼k FiyatlÄ± ÃœrÃ¼nÃ¼ gÃ¶ren kiÅŸiler.

- Hipotez: Ä°ki fiyatlandÄ±rma stratejisi arasÄ±nda mÃ¼ÅŸteri satÄ±n alma davranÄ±ÅŸlarÄ±nda anlamlÄ± bir fark var mÄ±?

````python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from statsmodels.stats.power import TTestIndPower

# Veri oluÅŸturma (simÃ¼lasyon)
np.random.seed(42)
n = 50  # Her grup iÃ§in Ã¶rneklem bÃ¼yÃ¼klÃ¼ÄŸÃ¼

# YÃ¼ksek fiyat stratejisi (499 TL) - %8 dÃ¶nÃ¼ÅŸÃ¼m
yuksek_fiyat = np.random.binomial(1, 0.08, size=n)

# DÃ¼ÅŸÃ¼k fiyat stratejisi (299 TL) - %12 dÃ¶nÃ¼ÅŸÃ¼m
dusuk_fiyat = np.random.binomial(1, 0.12, size=n)

# Betimsel istatistikler
df = pd.DataFrame({
    'Strateji': ['YÃ¼ksek Fiyat']*n + ['DÃ¼ÅŸÃ¼k Fiyat']*n,
    'SatÄ±n Alma': np.concatenate([yuksek_fiyat, dusuk_fiyat])
})
````

````python
# BaÄŸÄ±msÄ±z iki Ã¶rneklem t-testi (proportion test)
def proportion_test(group1, group2, alpha=0.05):
    n1, n2 = len(group1), len(group2)
    p1, p2 = group1.mean(), group2.mean()
    
    # Pooled proportion
    p_pooled = (group1.sum() + group2.sum()) / (n1 + n2)
    
    # Z-test istatistiÄŸi
    z = (p1 - p2) / np.sqrt(p_pooled*(1-p_pooled)*(1/n1 + 1/n2))
    
    # p-deÄŸeri (iki yÃ¶nlÃ¼)
    p_value = 2 * (1 - stats.norm.cdf(abs(z)))
    
    # Etki bÃ¼yÃ¼klÃ¼ÄŸÃ¼ (Cohen's h)
    h = 2 * (np.arcsin(np.sqrt(p1)) - np.arcsin(np.sqrt(p2)))
    
    # GÃ¼Ã§ analizi
    effect_size = abs(p1 - p2)
    power_analysis = TTestIndPower()
    power = power_analysis.solve_power(effect_size=effect_size, nobs1=n1, alpha=alpha)
    
    return z, p_value, h, power
````

````python
# Testi uygula
z_score, p_val, cohen_h, test_power = proportion_test(yuksek_fiyat, dusuk_fiyat)
````

3. **EÄŸitim ProgramlarÄ± KarÅŸÄ±laÅŸtÄ±rmasÄ±:**

- Problem: Bir okul, iki farklÄ± eÄŸitim programÄ±nÄ± deÄŸerlendirmek istiyor.

- Gruplar: Ä°lk grup, EÄŸitim ProgramÄ± A'ya katÄ±lan Ã¶ÄŸrenciler, ikinci grup ise EÄŸitim ProgramÄ± B'ye katÄ±lan Ã¶ÄŸrenciler.

- Hipotez: Ä°ki eÄŸitim programÄ± arasÄ±nda Ã¶ÄŸrenci baÅŸarÄ± dÃ¼zeylerinde anlamlÄ± bir fark var mÄ±?

````python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.power import TTestIndPower

# Veri oluÅŸturma (simÃ¼lasyon)
np.random.seed(42)
n_a, n_b = 30, 35  # Ã–rneklem bÃ¼yÃ¼klÃ¼kleri

# Program A: Ortalama 75 puan, Standart sapma 10
program_a = np.random.normal(loc=75, scale=10, size=n_a).clip(0, 100)

# Program B: Ortalama 82 puan, Standart sapma 8
program_b = np.random.normal(loc=82, scale=8, size=n_b).clip(0, 100)

# Veri Ã§erÃ§evesi oluÅŸturma
df = pd.DataFrame({
    'Program': ['A']*n_a + ['B']*n_b,
    'Puan': np.concatenate([program_a, program_b]),
    'Ã–ÄŸrenciID': np.arange(n_a + n_b)
})

# Normallik testi (Shapiro-Wilk)
normal_a = stats.shapiro(program_a).pvalue > 0.05
normal_b = stats.shapiro(program_b).pvalue > 0.05

# Varyans homojenliÄŸi testi (Levene)
levene_test = stats.levene(program_a, program_b)

# Parametrik test (t-test) veya non-parametrik test (Mann-Whitney U) seÃ§imi
if normal_a and normal_b:
    # Varyans homojenliÄŸine gÃ¶re uygun t-test seÃ§imi
    if levene_test.pvalue > 0.05:
        t_test = stats.ttest_ind(program_a, program_b, equal_var=True)
        test_adi = "BaÄŸÄ±msÄ±z Ä°ki Ã–rneklem t-Testi (Student)"
    else:
        t_test = stats.ttest_ind(program_a, program_b, equal_var=False)
        test_adi = "BaÄŸÄ±msÄ±z Ä°ki Ã–rneklem t-Testi (Welch)"
    test_istatistik, p_value = t_test.statistic, t_test.pvalue
else:
    # Non-parametrik test
    test_istatistik, p_value = stats.mannwhitneyu(program_a, program_b)
    test_adi = "Mann-Whitney U Testi"
````

````python
# Etki bÃ¼yÃ¼klÃ¼ÄŸÃ¼ hesaplama
def calculate_effect_size(group1, group2):
    n1, n2 = len(group1), len(group2)
    pooled_std = np.sqrt(((n1-1)*np.std(group1, ddof=1)**2 + (n2-1)*np.std(group2, ddof=1)**2) / (n1+n2-2))
    return (np.mean(group1) - np.mean(group2)) / pooled_std

cohen_d = calculate_effect_size(program_b, program_a)

# GÃ¼Ã§ analizi
effect_size = abs(cohen_d)
power_analysis = TTestIndPower()
power = power_analysis.solve_power(effect_size=effect_size, nobs1=n_a, alpha=0.05)
````

4. **Web Sitesi KullanÄ±labilirliÄŸi KarÅŸÄ±laÅŸtÄ±rmasÄ±:**

- Gruplar: Ä°lk grup, Web Sitesi A'yÄ± kullanan kullanÄ±cÄ±lar, ikinci grup ise Web Sitesi B'yi kullanan kullanÄ±cÄ±lar.

- Hipotez: Ä°ki web sitesi arasÄ±nda kullanÄ±labilirlik performansÄ±nda anlamlÄ± bir fark var mÄ±?

````python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.power import TTestIndPower
import pingouin as pg  # Ä°leri istatistiksel analizler iÃ§in

# Veri oluÅŸturma (simÃ¼lasyon)
np.random.seed(42)
n_a, n_b = 45, 50  # KullanÄ±cÄ± sayÄ±larÄ±

# Web Sitesi A: Ortalama gÃ¶rev tamamlama sÃ¼resi (saniye)
site_a = np.random.normal(loc=120, scale=25, size=n_a).clip(60, 180)

# Web Sitesi B: Ortalama gÃ¶rev tamamlama sÃ¼resi (saniye)
site_b = np.random.normal(loc=95, scale=20, size=n_b).clip(50, 150)

# KullanÄ±cÄ± memnuniyet puanlarÄ± (1-10 arasÄ±)
memnuniyet_a = np.random.normal(loc=6.5, scale=1.5, size=n_a).clip(1, 10)
memnuniyet_b = np.random.normal(loc=8.2, scale=1.2, size=n_b).clip(1, 10)

# Veri Ã§erÃ§evesi oluÅŸturma
df = pd.DataFrame({
    'Site': ['A']*n_a + ['B']*n_b,
    'GÃ¶rev_SÃ¼resi': np.concatenate([site_a, site_b]),
    'Memnuniyet': np.concatenate([memnuniyet_a, memnuniyet_b]),
    'KullanÄ±cÄ±ID': np.arange(n_a + n_b)
})

# Normallik testi
normal_a = stats.shapiro(site_a).pvalue > 0.05
normal_b = stats.shapiro(site_b).pvalue > 0.05

# Varyans homojenliÄŸi testi
levene_test = pg.homoscedasticity(df, dv='GÃ¶rev_SÃ¼resi', group='Site')

# Uygun test seÃ§imi ve uygulanmasÄ±
if normal_a and normal_b:
    if levene_test['pval'].iloc[0] > 0.05:
        t_test = pg.ttest(df[df['Site']=='A']['GÃ¶rev_SÃ¼resi'],
                         df[df['Site']=='B']['GÃ¶rev_SÃ¼resi'],
                         correction=False)
        test_adi = "BaÄŸÄ±msÄ±z t-Testi (Student)"
    else:
        t_test = pg.ttest(df[df['Site']=='A']['GÃ¶rev_SÃ¼resi'],
                         df[df['Site']=='B']['GÃ¶rev_SÃ¼resi'],
                         correction=True)
        test_adi = "BaÄŸÄ±msÄ±z t-Testi (Welch)"
else:
    t_test = pg.mwu(df[df['Site']=='A']['GÃ¶rev_SÃ¼resi'],
                   df[df['Site']=='B']['GÃ¶rev_SÃ¼resi'])
    test_adi = "Mann-Whitney U Testi"

# Etki bÃ¼yÃ¼klÃ¼ÄŸÃ¼ hesaplama
cohen_d = pg.compute_effsize(df[df['Site']=='A']['GÃ¶rev_SÃ¼resi'],
                            df[df['Site']=='B']['GÃ¶rev_SÃ¼resi'],
                            eftype='cohen')

# Memnuniyet puanlarÄ± iÃ§in test
memnuniyet_test = pg.mwu(df[df['Site']=='A']['Memnuniyet'],
                        df[df['Site']=='B']['Memnuniyet'])

# GÃ¼Ã§ analizi
power_analysis = TTestIndPower()
# Etki bÃ¼yÃ¼klÃ¼ÄŸÃ¼nÃ¼ doÄŸrudan Cohen's d deÄŸerinden alÄ±yoruz
power = power_analysis.solve_power(effect_size=abs(cohen_d), # Etki bÃ¼yÃ¼klÃ¼ÄŸÃ¼ pozitif olmalÄ±
                                  nobs1=n_a,
                                  alpha=0.05)

# GÃ¶rselleÅŸtirme
plt.figure(figsize=(16, 6))

# 1. GÃ¶rev tamamlama sÃ¼releri
plt.subplot(1, 3, 1)
sns.violinplot(x='Site', y='GÃ¶rev_SÃ¼resi', data=df,hue='Site', palette=["#1f77b4", "#ff7f0e"])
sns.swarmplot(x='Site', y='GÃ¶rev_SÃ¼resi', data=df, color='.25', size=3, alpha=0.7)
plt.title('GÃ¶rev Tamamlama SÃ¼releri\n')
plt.ylabel('SÃ¼re (saniye)')

# 2. Memnuniyet puanlarÄ±
plt.subplot(1, 3, 2)
sns.boxplot(x='Site', y='Memnuniyet', data=df, hue='Site', palette=["#1f77b4", "#ff7f0e"])
plt.title('KullanÄ±cÄ± Memnuniyet PuanlarÄ±\n')
plt.ylabel('Memnuniyet (1-10)')
plt.ylim(0, 10)

# 3. SÃ¼re-Memnuniyet iliÅŸkisi
plt.subplot(1, 3, 3)
sns.scatterplot(x='GÃ¶rev_SÃ¼resi', y='Memnuniyet', hue='Site', data=df,
                palette=["#1f77b4", "#ff7f0e"], alpha=0.7)
plt.title('SÃ¼re-Memnuniyet Ä°liÅŸkisi\n')
plt.xlabel('GÃ¶rev SÃ¼resi (saniye)')
plt.ylabel('Memnuniyet PuanÄ±')

plt.tight_layout()
plt.show()

# SonuÃ§ raporu
print(f"""
ğŸ“Š WEB SÄ°TESÄ° KULLANILABÄ°LÄ°RLÄ°K ANALÄ°Z SONUÃ‡LARI
--------------------------------------------

â±ï¸ GÃ–REV TAMAMLAMA SÃœRELERÄ°:
---------------------------------
Web Sitesi A (n={n_a}):
- Ortalama: {np.mean(site_a):.1f} saniye
- Standart Sapma: {np.std(site_a, ddof=1):.1f} saniye
- Normallik: {'Normal' if normal_a else 'Normal DeÄŸil'}

Web Sitesi B (n={n_b}):
- Ortalama: {np.mean(site_b):.1f} saniye
- Standart Sapma: {np.std(site_b, ddof=1):.1f} saniye
- Normallik: {'Normal' if normal_b else 'Normal DeÄŸil'}

ğŸ“ˆ Ä°STATÄ°STÄ°KSEL TEST SONUÃ‡LARI (SÃœRELER):
---------------------------------
Test TÃ¼rÃ¼: {test_adi}
Test Ä°statistiÄŸi: {t_test['U'].iloc[0] if test_adi.startswith('Mann') else t_test['T'].iloc[0]:.3f}
p-deÄŸeri: {t_test['p-val'].iloc[0]:.4f}
Etki BÃ¼yÃ¼klÃ¼ÄŸÃ¼ (Cohen's d): {cohen_d:.3f}
Test GÃ¼cÃ¼ (1-Î²): {power:.2%}

ğŸ’¡ YORUM (SÃœRELER):
{'âœ…' if t_test['p-val'].iloc[0] < 0.05 else 'âŒ'} {'Fark istatistiksel olarak ANLAMLI' if t_test['p-val'].iloc[0] < 0.05 else 'Fark istatistiksel olarak ANLAMSIZ'}
(p = {t_test['p-val'].iloc[0]:.4f}, Î± = 0.05)

ğŸ” PRATÄ°K ETKÄ° (SÃœRELER):
Web Sitesi B, Web Sitesi A'ya gÃ¶re ortalama {np.mean(site_a)-np.mean(site_b):.1f} saniye daha hÄ±zlÄ±.
Bu fark, Cohen's d standartlarÄ±na gÃ¶re {'kÃ¼Ã§Ã¼k' if abs(cohen_d)<0.5 else 'orta' if abs(cohen_d)<0.8 else 'bÃ¼yÃ¼k'} bir etki bÃ¼yÃ¼klÃ¼ÄŸÃ¼ne sahip.

ğŸ˜Š KULLANICI MEMNUNÄ°YET SONUÃ‡LARI:
---------------------------------
Web Sitesi A:
- Ortalama Memnuniyet: {np.mean(memnuniyet_a):.1f}/10

Web Sitesi B:
- Ortalama Memnuniyet: {np.mean(memnuniyet_b):.1f}/10

Mann-Whitney U Testi:
- p-deÄŸeri: {memnuniyet_test['p-val'].iloc[0]:.4f}
- SonuÃ§: {'Memnuniyetler arasÄ±nda anlamlÄ± fark var' if memnuniyet_test['p-val'].iloc[0] < 0.05 else 'Memnuniyetler arasÄ±nda anlamlÄ± fark yok'}

ğŸ“Œ TAVSÄ°YELER:
1. {'Web Sitesi B aÃ§Ä±kÃ§a daha iyi performans gÃ¶steriyor.' if t_test['p-val'].iloc[0] < 0.05 else 'Performans farkÄ± istatistiksel olarak anlamlÄ± deÄŸil.'}
2. KullanÄ±cÄ± memnuniyeti Web Sitesi B'de {'Ã¶nemli Ã¶lÃ§Ã¼de daha yÃ¼ksek' if memnuniyet_test['p-val'].iloc[0] < 0.05 else 'benzer'}.
3. HÄ±z ve memnuniyet arasÄ±ndaki negatif korelasyon {'doÄŸrulanmÄ±ÅŸtÄ±r' if np.corrcoef(df['GÃ¶rev_SÃ¼resi'], df['Memnuniyet'])[0,1] < -0.3 else 'zayÄ±ftÄ±r'}.
""")
````

````python
import numpy as np
from scipy import stats

# Ã–rnek veriler (Web Sitesi A ve B iÃ§in performans sÃ¼releri ve memnuniyet puanlarÄ±)
# Bu verileri kendi gerÃ§ek verilerinizle deÄŸiÅŸtirin
siteA_performance = np.array([85, 78, 92, 65, 70, 88, 75, 82, 79, 91])  # saniye
siteB_performance = np.array([60, 55, 58, 63, 52, 59, 61, 57, 56, 54])  # saniye

siteA_satisfaction = np.array([3.2, 3.5, 3.0, 3.7, 3.1, 3.4, 3.3, 3.6, 3.2, 3.5])  # 5 Ã¼zerinden
siteB_satisfaction = np.array([4.8, 4.7, 4.9, 4.5, 5.0, 4.6, 4.8, 4.7, 4.9, 4.8])  # 5 Ã¼zerinden

# Performans iÃ§in t-testi
t_stat_perf, p_value_perf = stats.ttest_ind(siteA_performance, siteB_performance, equal_var=False)

# Memnuniyet iÃ§in t-testi
t_stat_satis, p_value_satis = stats.ttest_ind(siteA_satisfaction, siteB_satisfaction, equal_var=False)

# SonuÃ§larÄ± yazdÄ±r
print("PERFORMANS KARÅILAÅTIRMASI (Web Sitesi A vs B)")
print(f"Ortalama performans sÃ¼resi - A: {np.mean(siteA_performance):.1f} saniye")
print(f"Ortalama performans sÃ¼resi - B: {np.mean(siteB_performance):.1f} saniye")
print(f"Performans farkÄ±: {np.mean(siteA_performance)-np.mean(siteB_performance):.1f} saniye")
print(f"t-istatistiÄŸi: {t_stat_perf:.3f}, p-deÄŸeri: {p_value_perf:.5f}")

print("\nMEMNUNÄ°YET KARÅILAÅTIRMASI (Web Sitesi A vs B)")
print(f"Ortalama memnuniyet puanÄ± - A: {np.mean(siteA_satisfaction):.1f}")
print(f"Ortalama memnuniyet puanÄ± - B: {np.mean(siteB_satisfaction):.1f}")
print(f"Memnuniyet farkÄ±: {np.mean(siteB_satisfaction)-np.mean(siteA_satisfaction):.1f}")
print(f"t-istatistiÄŸi: {t_stat_satis:.3f}, p-deÄŸeri: {p_value_satis:.5f}")

# AnlamlÄ±lÄ±k deÄŸerlendirmesi
alpha = 0.05
print("\nSONUÃ‡:")
if p_value_perf < alpha:
    print("Performans sÃ¼releri arasÄ±ndaki fark istatistiksel olarak anlamlÄ±dÄ±r (p < 0.05).")
else:
    print("Performans sÃ¼releri arasÄ±ndaki fark istatistiksel olarak anlamlÄ± deÄŸildir.")

if p_value_satis < alpha:
    print("Memnuniyet puanlarÄ± arasÄ±ndaki fark istatistiksel olarak anlamlÄ±dÄ±r (p < 0.05).")
else:
    print("Memnuniyet puanlarÄ± arasÄ±ndaki fark istatistiksel olarak anlamlÄ± deÄŸildir.")
````

Bir ML projesine yatÄ±rÄ±m yapÄ±lmÄ±ÅŸ. ÃœrettiÄŸi tahminler neticesinde oluÅŸan gelir ile eski sistemin Ã¼rettiÄŸi gelirler karÅŸÄ±laÅŸtÄ±rÄ±lÄ±p anlamlÄ± farklÄ±lÄ±k olup olmadÄ±ÄŸÄ± test edilmek isteniyor.

Detaylar:

Model geliÅŸtirilmiÅŸ ve web sitesine entefre edilmiÅŸ.

````python
import numpy as np
import pandas as pd


#veri tipi 1 - excel , csv veya veritababÄ±ndan okuduÄŸumuzu varsayalÄ±m
#veri bu ÅŸekilde elimize geldiÄŸi iÃ§in veri tiplerini dÃ¶nÃ¼ÅŸtÃ¼rme iÅŸlemi yapmamÄ±z gerekiyor




A=pd.DataFrame([30,27,21,27,29,30,20,20,27,32,35,22,24,23,25,27,23,27,23,25,21,18,24,26,33,26,27,28,19,25])
B=pd.DataFrame([37,39,31,31,34,38,30,36,29,28,38,28,37,37,30,32,31,31,27,32,33,33,33,31,32,33,26,32,33,29])
A_B=pd.concat([A,B],axis=1)
A_B.columns=['A','B']
A_B.head()
````

````python
#veri tipi 2 - elimizdeki veri bu ÅŸekilde ise


#A ve B gruplarÄ±nÄ±n karÅŸÄ±laÅŸtÄ±rÄ±lmasÄ±
#A ve B gruplarÄ±nÄ±n ortalamalarÄ± arasÄ±nda istatistiksel olarak anlamlÄ± bir fark olup olmadÄ±ÄŸÄ±nÄ± test etmek istiyoruz.
#Bu test iÃ§in baÄŸÄ±msÄ±z iki Ã¶rneklem t testi kullanacaÄŸÄ±z.
#H0: M1 = M2 (iki grup ortalamalarÄ± arasÄ±nda istatistiksel olarak anlamlÄ± bir fark yoktur)
#H1: M1 != M2 (iki grup ortalamalarÄ± arasÄ±nda istatistiksel olarak anlamlÄ± bir fark vardÄ±r)


#A pandas DataFrame is created containing a single column of numeric values. These represent hypothetical income data (labeled "gelir").
A=pd.DataFrame([30,27,21,27,29,30,20,20,27,32,35,22,24,23,25,27,23,27,23,25,21,18,24,26,33,26,27,28,19,25])
B=pd.DataFrame([37,39,31,31,34,38,30,36,29,28,38,28,37,37,30,32,31,31,27,32,33,33,33,31,32,33,26,32,33,29])

#A ve A'nÄ±n grubu
GRUP_A =np.arange(len(A))   #Creating a Group Identifier , creates an array of integers from 0 to len(A)-1 (0 to 29 in this case, as A has 30 rows
GRUP_A=pd.DataFrame(GRUP_A)  # GRUP_A is converted to a DataFrame
GRUP_A[:]='A'               # All values in GRUP_A are set to 'A'
A=pd.concat([A,GRUP_A],axis=1)  # A and GRUP_A are concatenated along axis 1 (i.e. columns)

#B ve B'nÄ±n grubu
GRUP_B =np.arange(len(B))
GRUP_B=pd.DataFrame(GRUP_B)
GRUP_B[:]='B'
B=pd.concat([B,GRUP_B],axis=1)


#Combining A and B into a Single Dataset
#tum veri
AB=pd.concat([A,B])   # Both A and B are concatenated into a single DataFrame AB
AB.columns=['gelir','GRUP']    # renames the columns to gelir (income) and GRUP (group).
#AB[:10]
print(AB.head())
print(AB.tail())

"""
Hipotez Testi AdÄ±mlarÄ±  
"""
````

````python
import seaborn as sns

# Create the boxplot with hue instead of x
sns.boxplot(x='GRUP', y='gelir', hue='GRUP', data=AB, palette={'A': 'skyblue', 'B': 'lightcoral'}, dodge=False, legend=False);
````

````python
from scipy.stats import shapiro
````

````python
#shapiro fonsiyonu bizden sadece deÄŸiÅŸkenin kendisini bekler
#shapiro fonksiyonu varsayÄ±mÄ± test eder
print(shapiro(A_B.A))
print(f'p value : {shapiro(A_B.A)[1]} ,A daki veriler iÃ§in , H0 hipotezini reddemeyiz p-value>0.05 daÄŸÄ±lÄ±mÄ±mÄ±z normaldÄ±r.') #p value > 0.05 olduÄŸu iÃ§in H0 hipotezi reddedilemez, daÄŸÄ±lÄ±m normaldir
````

````python
print(shapiro(A_B.B)) #shapiro testi uyguluyoruz, bu fonsiyonun iÃ§erisine testini yapmak istediÄŸimiz deÄŸiÅŸkeni giriyoruz. sol taraftaki deÄŸer test istatistiÄŸini, saÄŸ taraftaki deÄŸer ise p-value deÄŸerini verir
print(f'p value : {shapiro(A_B.B)[1]} ,B deki veriler iÃ§in , H0 hipotezini reddemeyiz p-value>0.05 daÄŸÄ±lÄ±mÄ±mÄ±z normaldÄ±r.')   
````

# varyans homojenliÄŸi varsayÄ±mÄ± kontrolÃ¼

- **Varyans homojenliÄŸi testi**

**Ä°ki deÄŸiÅŸkeni karÅŸÄ±laÅŸtÄ±rdÄ±ÄŸÄ±mÄ±z iÃ§in vayans homojenliÄŸini kontorl ediyoruz**

- H0: Varyansalar homojendir.
- H1: Varyasnlar homojen deÄŸildir.

````python
from scipy import stats  #levene testi uygulamak iÃ§in gerekli stats modÃ¼lÃ¼nÃ¼ scipy'dan import edelim
stats.levene(A_B.A, A_B.B) #levene testi uyguluyoruz, bu fonsiyonun iÃ§erisine testini yapmak istediÄŸimiz iki deÄŸiÅŸkeni giriyoruz. sol taraftaki deÄŸer test istatistiÄŸini, saÄŸ taraftaki deÄŸer ise p-value deÄŸerini verir
````

````python
print(f'p value : {stats.levene(A_B.A,A_B.B)[1]} ,H0 dipotezini rededemeyiz.p>0.05 ,Varyans homojenliÄŸi varsayÄ±mÄ±nÄ± reddemeyiz.') 
````

## <span style ="color:red">**Hipotez Testi**</span>

````python
stats.ttest_ind(A_B['A'], A_B['B'], equal_var = True) #baÄŸÄ±msÄ±z iki Ã¶rneklem t testi uyguluyoruz, bu fonsiyonun iÃ§erisine testini yapmak istediÄŸimiz iki deÄŸiÅŸkeni ve varyans homojenliÄŸi varsayÄ±mÄ±nÄ± giriyoruz. sol taraftaki deÄŸer test istatistiÄŸini, saÄŸ taraftaki deÄŸer ise p-value deÄŸerini verir
````

`stats.ttest_ind` fonksiyonu, iki baÄŸÄ±msÄ±z Ã¶rneklem arasÄ±nda istatistiksel olarak anlamlÄ± bir fark olup olmadÄ±ÄŸÄ±nÄ± test etmek iÃ§in kullanÄ±lÄ±r. Bu fonksiyon, Student'un t-testini uygular.

Ä°ki ana parametre alÄ±r:

1. **Ä°lk grup verileri (`A_B['A']`):** Bu, karÅŸÄ±laÅŸtÄ±rma yapmak istediÄŸiniz birinci grup verileridir.
   
2. **Ä°kinci grup verileri (`A_B['B']`):** Bu, karÅŸÄ±laÅŸtÄ±rma yapmak istediÄŸiniz ikinci grup verileridir.

`equal_var` parametresi, varsayÄ±lan olarak `True` olarak ayarlanmÄ±ÅŸtÄ±r. Bu durumda, iki grup iÃ§in varyansÄ±n eÅŸit olduÄŸunu varsayar. EÄŸer `equal_var=False` olarak ayarlanÄ±rsa, varyanslarÄ±n eÅŸit olmadÄ±ÄŸÄ± kabul edilir ve Welch's t-testi uygulanÄ±r.

SonuÃ§lar arasÄ±nda bir t-test istatistiÄŸi ve p deÄŸeri bulunur. P deÄŸeri, iki grup arasÄ±ndaki farkÄ±n istatistiksel olarak anlamlÄ± olup olmadÄ±ÄŸÄ±nÄ± belirler. DÃ¼ÅŸÃ¼k bir p deÄŸeri (< 0.05 genellikle kullanÄ±lÄ±r) genellikle iki grup arasÄ±ndaki farkÄ±n istatistiksel olarak anlamlÄ± olduÄŸunu gÃ¶sterir.

Bu kod, `A_B` veri Ã§erÃ§evesindeki 'A' ve 'B' gruplarÄ± arasÄ±ndaki t-test istatistiÄŸini ve p deÄŸerini hesaplar.

````python
#A_B veri setinde A ve B gruplarÄ±
t_statistic, p_value = stats.ttest_ind(A_B['A'], A_B['B'], equal_var=True)
print(f"T-Statistic: {t_statistic:.4f}, P-Value: {p_value:.9f}")

print('p<0.05 ,H0 hipotezini reddettik yani ortalamalarÄ± eÅŸit deÄŸil')
````

**DolayÄ±sÄ± ile eski sistem ile yeni sistem arasÄ±nda gelirler gÃ¶z Ã¶nÃ¼ne alÄ±ndÄ±ÄŸÄ±nda istatistiki olarak anlamlÄ± bir farklÄ±lÄ±k vardÄ±r ve bu yÃ¼zden H0 reddedilir. Bu fark yeni sistemin lehinedir. OrtalamalarÄ± incelediÄŸimizde de zaten bunu gÃ¶zlemlemiÅŸtik.**

## <span style ="color:red">**Nonparametrik BaÄŸÄ±msÄ±z Ä°ki Ã–rneklem Testi**</span>


- **VarsayÄ±mlar gerÃ§ekleÅŸmediÄŸi zaman yapÄ±lÄ±r.**

`stats.mannwhitneyu` fonksiyonu, Mann-Whitney U testini uygular. Mann-Whitney U testi, iki baÄŸÄ±msÄ±z Ã¶rneklem arasÄ±nda medyan deÄŸerlerinin istatistiksel olarak farklÄ± olup olmadÄ±ÄŸÄ±nÄ± test etmek iÃ§in kullanÄ±lÄ±r. Bu test, Ã¶zellikle baÄŸÄ±msÄ±z Ã¶rneklem t-testinin varsayÄ±mlarÄ±nÄ±n karÅŸÄ±lanmadÄ±ÄŸÄ± durumlarda (Ã¶rneÄŸin, veriler normal daÄŸÄ±lmÄ±yorsa) kullanÄ±ÅŸlÄ±dÄ±r.

Fonksiyon, iki ana parametre alÄ±r:

1. **Ä°lk grup verileri (`A_B['A']`):** Bu, karÅŸÄ±laÅŸtÄ±rma yapmak istediÄŸiniz birinci grup verileridir.
   
2. **Ä°kinci grup verileri (`A_B['B']`):** Bu, karÅŸÄ±laÅŸtÄ±rma yapmak istediÄŸiniz ikinci grup verileridir.

SonuÃ§ olarak, Mann-Whitney U istatistiÄŸi ve p deÄŸeri dÃ¶ner. P deÄŸeri, iki grup arasÄ±ndaki medyan farkÄ±nÄ±n istatistiksel olarak anlamlÄ± olup olmadÄ±ÄŸÄ±nÄ± belirler. DÃ¼ÅŸÃ¼k bir p deÄŸeri (< 0.05 genellikle kullanÄ±lÄ±r) genellikle iki grup arasÄ±ndaki medyan farkÄ±nÄ±n istatistiksel olarak anlamlÄ± olduÄŸunu gÃ¶sterir.

````python
#varsayalÄ±m kigerÃ§ekleÅŸtirdiÄŸim varsayÄ±m testleri sonucu negatif oldu ve normal daÄŸÄ±lÄ±m ve varyans homojenliÄŸi varsayÄ±mlarÄ±nÄ±n saÄŸlandÄ±ÄŸÄ±nÄ± gÃ¶remedik

#Bu durumda non-parametrik baÄŸÄ±msÄ±z iki Ã¶rneklem t testi yapmamÄ±z gerekmektedir

stats.mannwhitneyu(A_B['A'], A_B['B']) #non-parametrik baÄŸÄ±msÄ±z iki Ã¶rneklem t testi uyguluyoruz, bu fonsiyonun iÃ§erisine testini yapmak istediÄŸimiz iki deÄŸiÅŸkeni giriyoruz. sol taraftaki deÄŸer test istatistiÄŸini, saÄŸ taraftaki deÄŸer ise p-value deÄŸerini verir
````

````python
# A_B veri Ã§erÃ§evesinde 'A' ve 'B' gruplarÄ±
u_statistic, p_value = stats.mannwhitneyu(A_B['A'], A_B['B'])

print(f"Mann-Whitney U Statistic: {u_statistic:.4f}, P-Value: {p_value:.9f}")


# Bu kod, `A_B` veri Ã§erÃ§evesindeki 'A' ve 'B' gruplarÄ± arasÄ±nda Mann-Whitney U istatistiÄŸini ve p deÄŸerini hesaplar.
````
